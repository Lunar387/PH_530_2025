## Gaussian beams



The tutorial expands the spatial modes in terms of Laguerre-Gaussian solutions, whose relevance is derived and explained in this section

### Paraxial Wave equation

Electromagnetic fields in an isotropic and uniform medium follow the scalar wave equation
\begin{equation}
  [\nabla^2 + k^2]\tilde{E}(x,y,z) = 0 \tag{1}
\end{equation}
where $ \tilde{E}(x,y,z) $ is the phasor amplitude of a field distribution that is sinusoidal in time. Assume the beam propagates along the $z$ direction, hence the primary spatial dependence of $\tilde{E}(x,y,z)$ is $e^{-ikz}$

Beams of practical interest generally have some transverse (here, $x$ and $y$ direction) variation of amplitude and phase, which change slowly with $z$ than $e^{ikz}$ variation for a well collimated beam.

Hence it is convenient to extract $e^{-ikz}$ out

\begin{equation}
  \tilde{E}(x,y,z) = \tilde{u}(x,y,z)e^{-ikz} \tag{2}
\end{equation}

Now,

\begin{equation}
  \left(\frac{\partial \tilde{E}}{\partial x}, \frac{\partial \tilde{E}}{\partial y}, \frac{\partial \tilde{E}}{\partial z} \right) = \left(\frac{\partial \tilde{u}}{\partial x}e^{-ikz}, \frac{\partial \tilde{u}}{\partial y}e^{-ikz}, \frac{\partial \tilde{u}}{\partial z}e^{-ikz} - ik \tilde{u}e^{-ikz} \right) \tag{3}
\end{equation}

\begin{equation}
  \left(\frac{\partial^2 \tilde{E}}{\partial x^2}, \frac{\partial^2 \tilde{E}}{\partial y^2}, \frac{\partial^2 \tilde{E}}{\partial z^2} \right) = \left(\frac{\partial^2 \tilde{u}}{\partial x^2}e^{-ikz}, \frac{\partial^2 \tilde{u}}{\partial y^2}e^{-ikz}, \frac{\partial^2 \tilde{u}}{\partial z^2}e^{-ikz} - 2ik \frac{\partial\tilde{u}}{\partial z}e^{-ikz} - k^2\tilde{u}e^{-ikz} \right) \tag{3}
\end{equation}

Using eq(3) in eq(1), we get

\begin{equation}
  (\nabla^2\tilde{u} - 2ik\partial_z\tilde{u} - k^2\tilde{u}^2 + k^2\tilde{u}^2)e^{-ikz} = 0
\end{equation}
\begin{equation}
  \nabla^2\tilde{u} - 2ik\partial_z\tilde{u} = 0 \tag{4}
\end{equation}

Now we make use of the paraxial wave approximation to the above equation ie, we neglect $\partial^2 \tilde{u}/\partial z^2$

\begin{equation}
  \frac{\partial^2 \tilde{u}}{\partial x^2} + \frac{\partial^2 \tilde{u}}{\partial y^2} -2ik\frac{\partial \tilde{u}}{\partial z} = 0 \tag{5}
\end{equation}

The above equation is called the paraxial wave equation

### Spherical waves and Fresnel approximation

A general solution to eq(1) for a uniform spherical wave diverging from a source point is given by

\begin{equation}
  \tilde{E}(\vec{r}, \vec{r_0}) = \frac{e^{-ik\rho (\vec{r}, \vec{r_0})}}{\rho (\vec{r}, \vec{r_0})} \tag{6}
\end{equation}

where $\rho(\vec{r}, \vec{r_0}) = \sqrt{(x-x_0)^2 + (y-y_0)^2 + (z-z_0)^2}$

If we wished to find the field distribution near the $z$-axis for a value of $z$ sufficiently far,

We can expand to first order in $z$,

\begin{equation}
  \rho(\vec{r}, \vec{r_0}) = z - z_0 + \frac{(x-x_0)^2 + (y-y_0)^2}{2(z-z_0)} \tag{7}
\end{equation}

which is called the Fresnel approximation. Using in eq(1), keeping upto quadratic in $z$ in the numerator and linear in denominator, and removing the $z$ phase to get $\tilde{u}$, we get

\begin{equation}
  \tilde{u}(x,y,z) = \frac{1}{z-z_0}\exp\left[-ik\frac{(x-x_0)^2 + (y-y_0)^2}{2(z-z_0)}\right] \tag{8}
\end{equation}

### Guassian spherical waves

Define $R'(z) = z - z_0 + R_0$ for a general beam with an initial radius of curvature $R_0$. Also note that the beam in eq(8) does not fall off in transverse directions. This is clearly unphysical. However the paraxial wave equation allows for complex values of $\vec{r_0}$. Exploiting this, we replace $z_0$ with $z_0 - \tilde{q}_0$. Doing these substitutions, we get

\begin{equation}
  \tilde{u}(x,y,z) = \frac{1}{\tilde{q}(z)}\exp\left[-ik\frac{x^2 + y^2}{2\tilde{q}(z)}\right] \tag{9}
\end{equation}

where $\tilde{q}(z) = z-z_0+\tilde{q}_0$, for $x_0, y_0 = 0$. Separating real and imaginary parts of $1/\tilde{q}(z)$

\begin{equation}
  \frac{1}{\tilde{q}(z)} = \frac{1}{R(z)} - \frac{i\lambda}{\pi w^2(z)} \tag{10}
\end{equation}

where $R(z)$ is the radius of curvature and $w(z)$ is the spot size, also called the beam waist. Note that this $R(z)$ cannot be derived using the equation of $R'(z)$, and must be derived using eq(10). Substituting in eq(9),

\begin{equation}
  \tilde{u}(x,y,z) = \frac{1}{\tilde{q}(z)}\exp\left[-ik\frac{x^2 + y^2}{2R(z)} - \frac{x^2 - y^2}{w^2(z)}\right] \tag{11}
\end{equation}

This is the lowest order solution to a larger family of solutions. The general solutions are Laguerre-Gaussian (LG) solutions, given by  

\begin{align}
  LG_{pl}(\rho, \phi, z) = \tilde{u}_{pl}(\rho, \phi, z) &= \sqrt{\frac{2p!}{\pi(p+|l|)!}}\left(\frac{1}{w(z)}\right)\left(\frac{\sqrt{2}\rho}{w(z)}\right)^{|l|} \\
  &\times \exp\left(-\frac{\rho^2}{w(z)^2}\right)L_p^{|l|}\left(\frac{2\rho^2}{w(z)^2}\right) \\
  &\times e^{i\left(l\phi + (2p+|l|+1)\arctan(\frac{z}{z_0}) + \frac{k\rho^2}{2R(z)}\right)} \tag{12}
\end{align}
where
\begin{equation}
  w(z) = w_0\sqrt{1+\left(\frac{z}{z_0}\right)^2} \qquad R(z) = z\left(1+\left(\frac{z}{z_0}\right)^2\right) \qquad z_0 = \frac{1}{2kw_0^2} \tag{13}
\end{equation}

are radius of the beam, curvature of the beam, the Rayleigh range. $w_0$ is the radius of the beam at the waist, $L^{|l|}_p$ are generalised Laguerre polynomials. The index $l$ is associated with $e^{il\phi}$, hence these modes are eigenmodes of the orbital angular momentum operator $\hat{L}$ with the eigenvalue $l\hbar$. The LG modes form a complete and orthogonal basis for a fixed beam waist and curvature.

## 2. Multimode quantum optics

A photon is described as an excitation of the vacuum state $|0 \rangle$. We have dropped the hats off operators in the following discussion.
\begin{equation}
  a^\dagger |0\rangle = |1\rangle
\end{equation}

where $a, a^\dagger$ are annihilation and creation operators respectively, satisfying $a^\dagger = \sqrt{n+1}|n+1\rangle$ and $a|n\rangle = \sqrt{n}|n-1\rangle$ ($n$ is the number of photons in the quantum state). A photon has various degrees of freedom such as polarisation $\sigma$, frequency (or time), and spatial modes (transverse position or wave vector). We label the creation/annihilation operators with these degrees of freedoms: $a_{\sigma, \omega, \mathbf{q}}, a^\dagger_{\sigma, \omega, \mathbf{q}}$. The polarisation $\sigma$ can be $\{-1, +1\}$, $\omega \ \epsilon \ \mathbb{R}$ and tranverse wave vector components $\mathbf{q} \ \epsilon \ \mathbb{R}^2$ where $\mathbf{k} = (\mathbf{q}, \sqrt{\omega^2/c^2 - |\mathbf{q}|^2})$.

One can go from the momentum to position space by doing a Fourier transform.
\begin{equation}
  a_{\sigma,\omega,\mathbf{r}_\perp} = \int d^{2q}\exp(-i\mathbf{q}\cdot\mathbf{r}_\perp)a_{\sigma,\omega,\mathbf{q}}
\end{equation}


A single photon can be a linear superposition of such degrees of freedom
\begin{equation}
  |1\rangle _{p, l} = a_{p, l}^\dagger |0\rangle = \sum_{\sigma}\int d\omega\int r_\perp^2 f_\sigma(\mathbf{r}_\perp, \omega)a^\dagger_{\sigma,\omega}(\mathbf{r}_\perp)
\end{equation}

Here $a^\dagger_{\sigma,\omega}(\mathbf{r}_\perp)$ is the creation operator that generates a photon of with frequency $\omega$ and polarisation at the transverse postion $\mathbf{r}_\perp$ specified by distance $z$.
We can create any general $N$ photon states using this formalism. We illustrate this using two examples as used in the tutorial, and compute their equal time first order correlation function $g^{(1)}(\mathbf{r}_\perp) = \langle\psi | a^\dagger(\mathbf{r}_\perp) a(\mathbf{r}_\perp)|\psi\rangle$

#### N identical photons (same polarisation and spatial modes)

\begin{equation}
  |N\rangle_{\sigma,p,l} = \frac{1}{\sqrt{N!}}(a_{\sigma,p,l}^\dagger)^N|0\rangle
\end{equation}

For calculating $g_1$, $a^\dagger(\mathbf{r}_\perp)$ can be expressed as a sum of mode operators $a^\dagger_{p,l}$ weighted by $LG_{p,l}(\mathbf{r}_\perp)$, due to orthonormality of the basis.

\begin{equation}
  a^\dagger(\mathbf{r}_\perp) = \sum_{p,l}LG_{p,l}(\mathbf{r}_\perp)a_{\sigma,p,l}^\dagger
\end{equation}

Thus,
\begin{equation}
  a(\mathbf{r}_\perp)|N\rangle_{\sigma,p,l} = \left(\sum_{p,l}LG^\dagger_{p,l}(\mathbf{r}_\perp)a_{\sigma,p,l}\right)|N\rangle_{\sigma,p,l}
  \\=\sqrt{N}LG_{p,l}^\dagger(\mathbf{r}_\perp)|N-1\rangle_{\sigma,p,l}
\end{equation}
Since the photons are identical, only the operator with the same $p,l$ as the $|N\rangle_{\sigma,p,l}$ state will give a non zero value $\sqrt{N}$. This implies

\begin{equation}
g^{(1)}(\mathbf{r}_\perp) = _{\sigma,p,l}\langle N | a^\dagger(\mathbf{r}_\perp) a(\mathbf{r}_\perp)|N\rangle_{\sigma,p,l}
\\=N|LG_{p,l}(\mathbf{r}_\perp)|^{2}
\end{equation}

The intensity measurement will display the typical shape of an LG mode.

#### N photons having different quantum states
\begin{equation}
\prod_{i=1}^{N} |1_{\sigma_i, p_i, l_i} \rangle := \prod_{i=1}^{N} \hat{a}^{\dagger}_{\sigma_i, p_i, l_i} |0\rangle
\end{equation}
Using the expansion of $a(\mathbf{r}_\perp)$ in terms of $a_{p,l}$ and operating it on $|\psi\rangle=\prod_{i=1}^{N} |1_{\sigma_i, p_i, l_i} \rangle$. The annihilation operator acts on the quantum state leaving only the terms with a photon in the state with the same $p,l$ values. The notation used here does not imply division with a ket, but it means that the the state in the denominator is replaced by the state in the numerator in the product to the right of it.

\begin{equation}
a(\mathbf{r}_\perp)\prod_{i=1}^{N} |1_{\sigma_i, p_i, l_i}\rangle=\left(\sum_{p,l}LG^\dagger_{p,l}(\mathbf{r}_\perp)a_{\sigma,p,l}\right)\prod_{i=1}^{N} |1_{\sigma_i, p_i, l_i}\rangle
\\=\left(\sum_{p_i,l_i}LG^\dagger_{p_i,l_i}(\mathbf{r}_\perp)\frac{|0_{\sigma_i, p_i, l_i}\rangle}{|1_{\sigma_i, p_i, l_i}\rangle}\prod_{j=1}^{N} |1_{\sigma_j, p_j, l_j}\rangle\right)
\end{equation}

where the notation $\frac{|0_{\sigma_i, p_i, l_i}\rangle}{|1_{\sigma_i, p_i, l_i}\rangle}$ means that in the state right to it, $|1_{\sigma_i, p_i, l_i}\rangle$ is replaced by $|0_{\sigma_i, p_i, l_i}\rangle$ and not division by a number state.

Since different number states are orthogonal, the only terms surviving will have the creation operator and annihilation operator having the same $p,l$ so that both the states in the inner product are the same, giving a blurred image.

\begin{equation}
 g^{(1)}(\mathbf{r}_\perp) = \sum_{i,j} \langle 1_{\sigma _i, p_i, l_i} | LG^\ast_{p_i,l_i}(\mathbf{r}_\perp) \delta_{p_i, p_j} \delta_{l_i,l_j} LG_{p_j,l_j}(\mathbf{r}_\perp) | 1_{\sigma_i, p_j, l_j} \rangle
 \\ = \sum_{i} |LG_{p_i,l_i}(\mathbf{r}_\perp)|^2
\end{equation}


#### Optical devices

Having described general multi photon states, we shift our focus to optical devices. A lossless linear device can be described by using a linear unitary transformation of the creation and annihilation operators

\begin{equation}
  a^{\dagger}_{\alpha} = \sum_{\beta} u_{\alpha\beta}a^{\dagger}_{\beta}
\end{equation}

The unitary is needed to keep the device lossless. The simplest example is the phase shifter, doing the following operation

\begin{equation}
  a^{\dagger}_{\alpha} = e^{i \theta}a^{\dagger}_{\alpha}
\end{equation}

Another example is the beamsplitter, a widely used optical device.

#### Action of Beamsplitter
The beamsplitter is a 2 input, 2 output device, made of a slab of material tilted at such an angle that some of the incident light is reflected and some transmitted. We can define $(a^\dagger, b^\dagger)$ to be the creation operators of the photons entering port $(a,b)$ and $(c^\dagger, d^\dagger)$ to be the creation operators of the photons exiting ports $(c, d)$. Assuming the beamsplitter is lossless (no. of photons conserved),

\begin{equation}
\begin{pmatrix}
\hat{a}^{\dagger} \\
\hat{b}^{\dagger}
\end{pmatrix}
=
\begin{pmatrix}
r^* & t \\
- t^* & r
\end{pmatrix}
\begin{pmatrix}
\hat{c}^{\dagger} \\
\hat{d}^{\dagger}
\end{pmatrix},
\end{equation}

Here, $r$ and $t$ are associated with the reflectivity and transmittivity of the device, and $|r|^2+|t|^2=1$. For incident fields with paraxial spatial modes, the coefficients only depend on the OAM modes as the index $l$ changes sign after getting reflected. Thus, for LG modes

\begin{equation}
\begin{pmatrix}
\hat{a}^{\dagger}_{l,p} \\
\hat{b}^{\dagger}_{-l,p}
\end{pmatrix}
=
\begin{pmatrix}
r^* & t \\
- t^* & r
\end{pmatrix}
\begin{pmatrix}
\hat{c}^{\dagger}_{-l,p} \\
\hat{d}^{\dagger}_{l,p}
\end{pmatrix},
\end{equation}

The BS is hence equivalent to a single qubit gate.
Now we only need to understand the action of a detector on quantum states to be able to implement generic quantum protocols.

#### Action of Detector
The action of any detector is found with the help of a measurement operator. When performing a measurement on a quantum system described by the density matrix $\rho$, the probability of detecting the photon is given by

\begin{equation}
\text{Probability} = \text{Trace} \left( \textstyle\hat{\prod} \hat{\rho} \right)
\end{equation}

where Trace() is the trace of the operator, and $\textstyle{\hat{\prod}}$ is the measurement operator.

A bucket detector is insensitive to the polarisation state, frequency or spatial mode of the incident photon. Hence, the action of a single photon bucket detector projects onto a mixed state given by

\begin{equation}
\hat{\prod}=|1\rangle \langle1|
\\=\sum_{\sigma}\int d^{2}r_{\perp}a^\dagger_{\sigma}(\mathbf{r}_\perp)|0\rangle \langle 0|a_{\sigma}(\mathbf{r}_\perp)
\end{equation}

If a single photon detector is sensitive to a specific spatial mode, e.g. $LG_{q,m}$ the action of the detector projects onto the desired spatial mode as shown by

\begin{equation}
\hat{\prod}=|1\rangle_{q,m} \langle1|
\\=\int d^{2}r_{\perp}d^{2}r^{'}_{\perp}LG(\mathbf{r}_\perp)a^\dagger_{\sigma}(\mathbf{r}_\perp)|0\rangle \langle 0|a_{\sigma}(\mathbf{r}_\perp)LG^{*}(\mathbf{r}_\perp)
\end{equation}

Hence, we now have all the necessary background to understand HOM effect after different detection conditions

##  3. Hong-Ou-Mandel Interference: Two-Photon Input State:

### Input State Description
We focus on the simplest case of two photons prepared in a product state, where each photon enters the beam splitter from a different port.

Let’s assume that the photon entering port 𝑎 is in the spatial mode ψ₁(𝐫⊥), and the photon entering port 𝑏 is in the spatial mode ψ₂(𝐫⊥). This means that we can express the input state as:

$$
|ψ₀⟩ = |ψ₁⟩_𝑎 ⊗ |ψ₂⟩_𝑏
$$

### Creation Operators
We denote $𝑎̂⁺_{α}$ as the creation operator for photons entering port 𝑎 in mode α, and $𝑏̂⁺_{β}$ as the creation operator for photons entering port 𝑏 in mode β.

If α and β correspond to transverse positions, we can write:

$$
|ψ₁⟩ = \int d^2𝐫⊥ ψ₁(𝐫⊥) \hat{a}_{𝐫⊥}^† |0⟩
$$

$$
|ψ₂⟩ = \int d^2𝐫⊥ ψ₂(𝐫⊥) \hat{b}_{𝐫⊥}^† |0⟩
$$

Alternatively, α and β can represent discrete indices associated with orthonormal sets of spatial modes (for example, ℓ and 𝑝).

For instance, if we decompose into Laguerre-Gaussian (LG) modes, we have:

$$
|ψ₁⟩ = \sum_{ℓ,p} \langle LG_{ℓ,p}|ψ₁\rangle \hat{a}_{ℓ,p}^† |0⟩
$$

$$
|ψ₂⟩ = \sum_{ℓ,p} \langle LG_{ℓ,p}|ψ₂\rangle \hat{b}_{ℓ,p}^† |0⟩
$$

Here, we use the notation:

$$
\langle LG_{ℓ,p}|ψ_{1,2}\rangle := \int d^2𝐫⊥ LG^*_{ℓ,p}(𝐫⊥) ψ_{1,2}(𝐫⊥)
$$

to denote the Hermitian product in the space of complex functions with integrable square.

### General Input State Formulation
Combining these expressions, we can write the full input state as:

$$
|ψ₀⟩ = |ψ₁⟩_𝑎 ⊗ |ψ₂⟩_𝑏 = \sum_{α,β} ψ_{1,α} ψ_{2,β} \hat{a}_α^† \hat{b}_β^† |0⟩
$$


### Action of the Beamsplitter
We consider, for simplicity, a 50:50 beamsplitter (BS) acting on the input modes as follows:

$$
\hat{a}_\alpha^† → \frac{1}{\sqrt{2}} (\hat{c}_\alpha^† + \hat{d}_\alpha^†)
$$

$$
\hat{b}_\beta^† → \frac{1}{\sqrt{2}} (\hat{c}_\beta^† - \hat{d}_\beta^†)
$$

Here, 𝑐 and 𝑑 refer to the output ports of the beamsplitter.

### Output State
The output state after the beamsplitter is given by:

$$
|ψ_{out}⟩ = \sum_{α,β} ψ_{1,α} ψ_{2,β}
\frac{1}{2} (\hat{c}_α^† + \hat{d}_α^†)(\hat{c}_β^† - \hat{d}_β^†) |0⟩
$$

### Measurement

### Coincidence Counts at Output Ports
In typical Hong-Ou-Mandel (HOM) experiments, we measure the coincidence counts at the output ports 𝑐 and 𝑑 of the beamsplitter.

To model this, we define the projection operator for port 𝑐 as:

$$
\hat{\Pi}_c = \sum_{γ} p_γ \hat{c}_γ^† |0⟩ ⟨0| \hat{c}_γ
$$

where $$ p_γ = 0, 1 $$, as required by the defining property of projection operators $$ \hat{\Pi}^2 = \hat{\Pi} $$ These coefficients allow us to distinguish between bucket, few-mode, and single-mode detectors.

### Coincidence Measurement Operator
The projector operator for coincidence measurements at ports 𝑐 and 𝑑 is given by:

$$
\hat{\Pi}_{cd} = \sum_{γ,γ'} p^{(c)}_γ p^{(d)}_{γ'}
\hat{c}_γ^† \hat{d}_{γ'}^† |0⟩ ⟨0| \hat{c}_γ \hat{d}_{γ'}
$$

The measurement output is then:

$$
R_{c,d} := Tr[\hat{\rho}_{out} \hat{\Pi}_{cd}]
= Tr[|ψ_{out}⟩ ⟨ψ_{out}| \hat{\Pi}_{cd}]
$$

Expanding this further:

$$
R_{c,d} = \sum_{γ,γ'} p^{(c)}_γ p^{(d)}_{γ'}
|⟨0| \hat{c}_γ \hat{d}_{γ'} |ψ_{out}⟩|^2
$$

Substituting the output state expression:

$$
R_{c,d} = \frac{1}{4} \sum_{γ,γ'} p^{(c)}_γ p^{(d)}_{γ'}
|ψ_{1,γ'} ψ_{2,γ} - ψ_{1,γ} ψ_{2,γ'}|^2
$$



## Example 1: Two Bucket Detectors
If both detectors in arms 𝑐 and 𝑑 can collect a large number of modes, then $$ p^{(c,d)}_γ = 1 $$ for each mode γ detectable by the sensors. The coincidence rate becomes:

$$
R_{c,d} = \frac{1}{4} \sum_{γ,γ'} |ψ_{1,γ'} ψ_{2,γ} - ψ_{1,γ} ψ_{2,γ'}|^2
$$

Expanding this expression:

$$
R_{c,d} = \frac{1}{2}
\left(
\sum_{γ} |ψ_{1,γ}|^2
\sum_{γ'} |ψ_{2,γ'}|^2
-
\sum_{γ} ψ^*_{1,γ} ψ_{2,γ}
\sum_{γ'} ψ^*_{2,γ'} ψ_{1,γ'}
\right)
$$

If all modes contained in \( ψ_1 \) and \( ψ_2 \) are detectable $$
\sum_{γ} |ψ_{1,γ}|^2 = 1 \quad \text{and} \quad \sum_{γ} |ψ_{2,γ}|^2 = 1
$$ this simplifies to:

$$
R_{c,d} = \frac{1}{2}(1 - |⟨ψ_1|ψ_2⟩|^2)
$$



## Example 2: Single-Mode Detectors
For single-mode detectors (e.g., SMFs or EMCCDs), we assume $$
p^{(c,d)}_\gamma = \delta_{\gamma,\eta(c,d)}
$$ The coincidence rate is:

$$
R_{c,d} = \frac{1}{4}
|ψ_{1,η_c} ψ_{2,η_d} - ψ_{1,η_d} ψ_{2,η_c}|^2
$$

Maximum HOM visibility can be achieved if the projection modes are identical and independent of the input modes.


## Example 3: Single-Mode Detector and Bucket Detector
In a hybrid configuration where one detector is single-mode and the other is a bucket detector $$ p^{(c)}_\gamma = \delta_{\gamma,\eta} $$ and $$ p^{(d)}_\gamma = 1, \quad \forall \gamma $$ we have:

$$
R_{c,d} = \frac{1}{4}
\sum_γ |ψ_{1, γ} ψ_{2, η} - ψ_{2, γ} ψ_{1, η}|^2
$$

Expanding further:

$$
R_{c,d} = \frac{1}{4}
\left(
|ψ_{1, η}|^2 + |ψ_{2, η}|^2 -
(\sum_γ ψ_{1, γ} ψ^*_{2, γ})(ψ_2 η)(ψ_1 η)^* + c.c.
\right)
$$

Using completeness $$ \sum_{γ} |γ\rangle \langle γ| = \hat{1} $$ this simplifies to:

$$
R_{c,d} = \frac{1}{4}
(|⟨η|ψ_1⟩|^2 + |⟨η|ψ_2⟩|^2 - [⟨ψ_2|ψ_1⟩ ⟨η|ψ_2⟩ ⟨ψ_1|η⟩ + c.c.])
$$



## Generalization to Correlated Input States
For non-separable input states $$ |ψ₀⟩ = \sum_{α,β} ψ_{α,β} \hat{a}_α^† \hat{b}_β^† |0⟩ $$ the coincidence rate becomes:

$$
R_{c,d} = \frac{1}{2} \sum_{γ,γ'} p_γ p'_{γ'}
|ψ_{γ,γ'} - ψ_{γ',γ}|^2
$$

For arbitrary transmissivity (unbalanced BS), the expression is:

$$
R_{c,d} = \sum_{γ,γ'} p^{(c)}_γ p^{(d)}_{γ'}
\left| |r|^2 ψ_{γ,γ'} - |t|^2 ψ_{γ',γ} \right|^2
$$

This highlights how imperfect visibility can arise due to differences in reflectivity and transmissivity of the BS.










## 4.  Explicit formulae for Laguerre-Gauss modes basis

From the previous sections, we observed that the coincidence rate in HOM experiments depends on the inner products between spatial modes $(\psi_1|\psi_2)$. The completeness and orthogonality of the Laguerre-Gauss modes allow us to always expand $\psi_{1,2}$ as an infinite series of LG functions. The inner product will thus become a superposition of inner products between different LG modes.

### Two Approaches

### 1. Same LG Basis
If $\psi_1$ and $\psi_2$ are decomposed in the same LG basis (same choice of waist $w_0$ and propagation distance $z$). Then, taking $\psi_{1,2} = \sum_{l,p} A_{l,p}^{(1,2)}LG_{l,p}\psi_1(\mathbf{r}_{\perp})$ :

$$
(\psi_1|\psi_2) = \sum_{\ell,\ell', p,p'} A^{(1)}_{\ell,p} A^{(2)}_{\ell',p'} (\ell, p | \ell', p')
$$

Using orthogonality:

$$
(\ell, p | \ell', p') = \delta_{\ell,\ell'} \delta_{p,p'}
$$

This simplifies to:

$$
(\psi_1|\psi_2) = \sum_{\ell, p} A^{(1)*}_{\ell,p} A^{(2)}_{\ell,p}
$$

### 2. Different LG Bases
When modes are expanded in different LG bases (different $w_0$ and $z$. Such a dituation is of interest when the modes incident on the beam splitter have different radii and wave-front curvature as shown in the figure), the inner product integral is:

$$
\mathcal{I}_{\ell,\ell', p, p'} = \int_0^{2\pi} d\phi \int_0^{\infty} \rho d\rho \text{LG}^*_{\ell,p}(\mu \rho, \phi; z) \text{LG}_{\ell',p'}(\mu' \rho, \phi; z')
$$

where

$$
\mu = \frac{2}{w^2(z)}, \quad \mu' = \frac{2}{w'^2(z')}
$$

The integral takes the form:

$$
\mathcal{I}_{\ell, \ell, p, p'} = A \int_0^{\infty} dx \ x^{|\ell|} e^{-\sigma x} L^{|\ell|}_p(\mu x) L^{|\ell|}_{p'}(\mu' x)
$$

with

$$
A = \frac{2^{|\ell|-1} \sqrt{2 p! \pi (p + |\ell|)! 2 p'! \pi (p' + |\ell|)!}}{w(z) w'(z')} e^{i \Delta \psi_G}
$$

and

$$
\sigma = \frac{1}{w^2(z)} + \frac{1}{w'^2(z')} + i \frac{k}{2} \left(-\frac{1}{R(z)} + \frac{1}{R'(z')}\right)
$$

where $\Delta\psi_G = (2p'+|l'| + 1)\arctan(\frac{z'}{z_0'}) - (2p+|l|+1)\arctan(\frac{z}{z_0})$.The solution involves the Hypergeometric Function $\,_2F_1$:

$$
(\ell, p | \ell', p') = A \frac{\Gamma(p + p' + |\ell| + 1)}{p! p'!} \frac{(\sigma - \mu')^{p'} (\sigma - \mu)^p}{\sigma^{p+p'+|\ell|+1}}
\,_2F_1\left(-p, -p'; -p - p' - |\ell|; \frac{\sigma(\sigma - \mu' - \mu)}{(\sigma - \mu')(\sigma - \mu)}\right)
$$

This formula is used to analyze experimental conditions, as variations in beam waist and propagation distance affect orthogonality and visibility in HOM experiments.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.special import gamma, hyp2f1, eval_genlaguerre
from scipy.integrate import quad
import time
```

```python
def lg_overlap(l, p, p_prime, w_ratio=1.0, z_ratio=0.0):
    """Implementation of Eq.(34) from the paper"""
    # Fixed beam parameters
    w1 = 1.0  # Normalized waist for mode 1
    w2 = w_ratio * w1  # Waist for mode 2
    z0 = 0.5 * w1**2  # Rayleigh range for mode 1
    z0_prime = 0.5 * w2**2  # Rayleigh range for mode 2
    z_prime = z_ratio * z0  # Propagation distance for mode 2

    # Calculate beam parameters at propagation distance z_prime
    w_z_prime = w2 * np.sqrt(1 + (z_prime/z0_prime)**2)
    R_prime = z_prime * (1 + (z0_prime/z_prime)**2) if z_prime != 0 else np.inf

    # Sigma parameter components
    sigma_real = (1/w1**2 + 1/w_z_prime**2)
    sigma_imag = (1/(2*np.inf) - 1/(2*R_prime))  # Mode 1 at waist (R = ∞)
    sigma = sigma_real + 1j*sigma_imag

    # Normalization factor (Eq.34)
    A = (2**l * np.sqrt(gamma(p+1)*gamma(p_prime+1)) /
         np.sqrt(gamma(p+l+1)*gamma(p_prime+l+1)))

    # Hypergeometric argument
    mu = 2/w1**2
    mu_prime = 2/w_z_prime**2
    denominator = (sigma - mu_prime) * (sigma - mu)

    if np.abs(denominator) < 1e-12:
        return 1.0 if p == p_prime else 0.0

    hypergeo_arg = (sigma * (sigma - mu_prime - mu)) / denominator

    # Full overlap calculation
    overlap = (
        A * gamma(p + p_prime + l + 1) / (gamma(p+1)*gamma(p_prime+1)) *
        (sigma - mu_prime)**p_prime * (sigma - mu)**p /
        sigma**(p + p_prime + l + 1) *
        hyp2f1(-p, -p_prime, -p-p_prime-l, hypergeo_arg)
    )

    return np.abs(overlap)**2
```

```python
# Parameters
l = 2
p_max = 6
p_values = np.arange(0, p_max+1)
waist_ratios = [0.5, 0.7, 1.0]
z_ratios = [0.0, 0.5, 1.0]
```

```python
# Plot waist variation, figure 3(a)
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
for i, w_ratio in enumerate(waist_ratios):
    data = np.zeros((7, 7))
    for p in p_values:
        for p_prime in p_values:
            data[p, p_prime] = lg_overlap(2, p, p_prime, w_ratio=w_ratio, z_ratio=0)

    sns.heatmap(data.T, cmap='viridis', annot=False, ax=axs[i],
                cbar_kws={'label': r'$|(l,p|l^{\prime},p^{\prime})|^2$'})
    axs[i].set_title(f'Waist Ratio: {w_ratio}')
    axs[i].set_xlabel('p')
    axs[i].set_ylabel('p\'')
plt.tight_layout()
plt.show()
```

```python
# Plot propagation distance variation, figure 3(b)
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
for i, z_ratio in enumerate(z_ratios):
    data = np.zeros((7, 7))
    for p in p_values:
        for p_prime in p_values:
            data[p, p_prime] = lg_overlap(2, p, p_prime, w_ratio=1.0, z_ratio=z_ratio)

    sns.heatmap(data.T, cmap='viridis', annot=False, ax=axs[i],
                cbar_kws={'label': r'$|(l,p|l^{\prime},p^{\prime})|^2$'})
    axs[i].set_title(f'z/z₀ Ratio: {z_ratio}')
    axs[i].set_xlabel('p')
    axs[i].set_ylabel('p\'')
plt.tight_layout()
plt.show()
```

```python
# Numerical integration formula implementation (Eq.32)
def lg_overlap_numerical(l, p, p_prime, w_ratio=1.0, z_ratio=0.0):
    """Numerical integration implementation of Eq.(32)"""
    # Fixed beam parameters
    w1 = 1.0  # Normalized waist for mode 1
    w2 = w_ratio * w1  # Waist for mode 2
    z0 = 0.5 * w1**2  # Rayleigh range for mode 1
    z0_prime = 0.5 * w2**2  # Rayleigh range for mode 2
    z_prime = z_ratio * z0  # Propagation distance for mode 2

    # Calculate beam parameters at propagation distance z_prime
    w_z_prime = w2 * np.sqrt(1 + (z_prime/z0_prime)**2)
    R_prime = z_prime * (1 + (z0_prime/z_prime)**2) if z_prime != 0 else np.inf

    # Sigma parameter components (Eq.33)
    sigma_real = (1/w1**2 + 1/w_z_prime**2)
    sigma_imag = (1/(2*np.inf) - 1/(2*R_prime))  # Mode 1 at waist (R = ∞)
    sigma = sigma_real + 1j*sigma_imag

    # Normalization factor (A from Eq.33)
    A = (2**(l-1) * np.sqrt(2 * gamma(p+1) * gamma(p_prime+1)) /
         (np.pi * np.sqrt(gamma(p+l+1) * gamma(p_prime+l+1))) *
         (1 / (w1 * w_z_prime)))

    # Define integrand function
    def integrand(x):
        return (x**l * np.exp(-sigma.real * x) *
                eval_genlaguerre(p, l, 2/w1**2 * x) *
                eval_genlaguerre(p_prime, l, 2/w_z_prime**2 * x))

    integral, _ = quad(integrand, 0, np.inf)
    return np.abs(A * integral)**2
```

```python
def time_evaluations(mode='waist', n_evaluations=100):
    """Time evaluations for different parameters"""
    ratios = np.linspace(0.5, 1, 10)  # Parameter variation range
    times_num = []
    times_anal = []

    l = 2
    p, p_prime = 3, 3  # Fixed mode indices for comparison

    for ratio in ratios:
        # Time numerical integration
        start = time.time()
        for _ in range(n_evaluations):
            if mode == 'waist':
                lg_overlap_numerical(l, p, p_prime, w_ratio=ratio)
            else:
                lg_overlap_numerical(l, p, p_prime, z_ratio=ratio)
        times_num.append((time.time() - start)/n_evaluations)

        # Time analytical solution
        start = time.time()
        for _ in range(n_evaluations):
            if mode == 'waist':
                lg_overlap(l, p, p_prime, w_ratio=ratio)
            else:
                lg_overlap(l, p, p_prime, z_ratio=ratio)
        times_anal.append((time.time() - start)/n_evaluations)

    return ratios, times_num, times_anal
```

```python
# Generate timing data
waist_ratios, num_times_w, anal_times_w = time_evaluations('waist')
z_ratios, num_times_z, anal_times_z = time_evaluations('z')
```

```python
# Create figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot waist variation timing
ax1.plot(waist_ratios, num_times_w, 'o-', label='Numerical Integration')
ax1.plot(waist_ratios, anal_times_w, 's-', label='Analytical Formula')
ax1.set_title('(c) Beam Waist Variation Timing')
ax1.set_xlabel('Beam Waist Ratio (w₂/w₁)')
ax1.set_ylabel('Average Time per Evaluation (s)')
ax1.legend()
ax1.grid(True)

# Plot propagation distance variation timing
ax2.plot(z_ratios, num_times_z, 'o-', label='Numerical Integration')
ax2.plot(z_ratios, anal_times_z, 's-', label='Analytical Formula')
ax2.set_title('(d) Propagation Distance Variation Timing')
ax2.set_xlabel('Propagation Distance Ratio (z/z₀)')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
