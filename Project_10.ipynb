## 2.2 - Performance Optimization via Temporal Filtering

## Overview of Temporal Filtering in QKD Systems

In Quantum Key Distribution (QKD) the imperfections in the system, such as **detector noise** (dark counts) and **photon arrival time fluctuations**, can degrade the quality of the key being generated. One way to improve the **performance** of QKD systems, particularly in **single-photon QKD**, is to use **temporal filtering**.

### Temporal Filtering Explained

**Temporal filtering** refers to the process of restricting the time window during which photon detection events are considered valid. This allows the system to discard photons detected outside a designated time window, effectively filtering out unwanted noise, such as dark counts from the detector.

- In the context of the paper, the authors show how this technique optimizes the QKD system's performance by **reducing the Quantum Bit Error Rate (QBER)** and increasing the **secure key rate**.

- Temporal filtering has a direct impact on the **Signal-to-Noise Ratio (SNR)**. By limiting the photon detection events to a small window, **noise events**, which are typically random, can be minimized, resulting in a cleaner signal. However, as the acceptance time window narrows, some useful photon detection events might also be discarded, which reduces the **sifted key fraction**.

- The key goal is to find the optimal **trade-off** between reducing the **QBER** and maintaining a high **sifted key fraction**, thereby maximizing the **secure key rate**.

In this section, we will explain how temporal filtering impacts QKD performance and how the **QBER**, **sifted key fraction**, and **secure key rate** are influenced by this technique.




## Quantum Bit Error Rate (QBER) and Temporal Filtering

### What is QBER?

**Quantum Bit Error Rate (QBER)** is a crucial parameter in QKD systems that measures the fraction of errors in the sifted key, which arises from imperfections in the detection process or channel noise. The QBER gives us an idea of how much noise and error are introduced during the key exchange, and it directly affects the security and efficiency of the final key.

### QBER in the Context of Temporal Filtering

In the paper, the authors focus on the **H-polarization channel** for calculating the **QBER**. The error contribution in the H-polarization channel is calculated by:

$$ QBER_H(\Delta t, t_c) = \frac{N_V}{N_H + N_V} $$

Where:
- $N_H$ is the number of photon detections in the **H** polarization.
- $N_V$ is the number of photon detections in the **V** polarization.
- $ \Delta t $ is the width of the **acceptance time window**.
- $ t_c $ is the center time of the pulse.

The QBER is calculated by considering all photon events that are detected within the time window $\Delta t $, centered around $ t_c $.

As **temporal filtering** is applied, the system reduces noise, especially from **dark counts**, which do not correspond to real photon events. By narrowing the time window, the system focuses on the **signal photons**, resulting in a reduction of the QBER.

## Trade-offs in Temporal Filtering

### Signal-to-Noise Ratio (SNR) and Sifted Key Fraction

As we reduce the acceptance time window **$ \Delta t $**, we improve the **Signal-to-Noise Ratio (SNR)** because we effectively remove **dark counts** (false detections caused by detector noise). However, the cost is the loss of **valid photon detections**, which means that we **sift fewer bits** for the final key.

The **sifted key fraction** is defined as the fraction of valid photon detection events that contribute to the final key. As $ \Delta t $ shrinks, some valid signal photons are discarded, which reduces the fraction of usable detections.

The paper discusses that the optimal QKD performance is achieved when we balance:
1. **Low QBER**: Ensuring fewer errors in the key generated.
2. **High sifted key fraction**: Maximizing the fraction of photon detections that are useful for generating the final key.


### Photon Statistics and Security Analysis for Sub-Poissonian Quantum Light Sources

For **sub-Poissonian quantum light sources**, the multiphoton probability \( p_m \) is given by:

$$ p_m \leq \frac{1}{2} \mu^2 g^{(2)}(0) $$

Where:
- $g^{(2)}(0)$: Second-order photon autocorrelation function.
- $\mu$: Mean photon number per pulse.

Reducing $ g^{(2)}(0)$ helps minimize multiphoton events, improving security by reducing the risk of **photon number splitting attacks**.

### Temporal Filtering and $ g^{(2)}(0)$

Temporal filtering of photon detection events improves **antibunching** and **security**. As the acceptance window $ \Delta t $ narrows, $g^{(2)}(0)$ decreases, indicating better photon purity.

## Real-Time Photon Statistics Monitoring in Quantum Key Distribution (QKD)



This section focuses on monitoring the photon statistics of the single-photon source (SPS) in real-time during quantum key distribution (QKD) to ensure security and detect potential attacks. The key metric used is the second-order autocorrelation function g (2)(0), which quantifies the probability of multiphoton emission events. A low g (2)(0) (close to 0) indicates high single-photon purity, while higher values suggest unwanted multiphoton emissions that could compromise security.

### 1- Why Monitor g (2)(0) in Real-Time?

In QKD, an eavesdropper (Eve) could exploit multiphoton pulses to gain partial key information via photon number splitting (PNS) attacks.
Real-time monitoring ensures that the photon statistics remain sub-Poissonian (ideal for QKD) and detects anomalies (e.g., unexpected increases in g (2)(0)) that might indicate tampering.

### 2-How g (2)(0) is Measured:

Photon Detection Events:
The receiver (Bob) records timestamps of photon detections across all four polarization channels (H, V, D, A).
Each detection event is logged with high precision (1 ps resolution in this experiment).
Screenshot 2025-03-30 at 3.51.24 PM.png
Correlation Analysis:
A histogram of time differences (τ) between all possible pairs of detection events is constructed.
This is called the g (2)(τ) histogram (see Fig. 2c in the paper).
Key Regions in the Histogram:
Zero-Delay Peak (τ=0):
Represents coincidences where two photons arrived at the same time (or within the detector’s timing resolution).
For an ideal single-photon source, this peak should be suppressed (since two photons should not be emitted simultaneously).
Side Peaks (τ≠0):
Represent uncorrelated photon detections (e.g., from dark counts or re-excitation of the quantum emitter).
These peaks serve as a reference for random coincidences.
Calculating g (2)(0):
The formula used is:

$g^{2}(0)=\frac{Coincidences at τ=0}{Average Conincidences at τ≠0}$


Numerically:

$N_{c=0}$= Counts in the zero-delay peak.

$N_{side}$= Average counts in side peaks.
Then, $g^{2}(0)= \frac{Nc=0}{Nside}	$


Interpreting g(2)(0):

* g (2)(0)=0 → Perfect single-photon source (no multiphoton emission).
*g(2)(0)< 0.5 → Good single-photon source (sub-Poissonian).
*g (2)(0)=1 → Coherent laser light (Poissonian statistics).
*g (2)
(0)>1 → Classical light (e.g., thermal source).

#### Why This Matters for QKD:

If  g(2)(0) is too high, multiphoton pulses exist, which could be exploited in photon number splitting (PNS) attacks.
By monitoring g (2)(0) in real-time, the system can detect security breaches (e.g., an attacker injecting extra photons).


### 3. Finite-Key Effects and Accumulation Time

In QKD, finite-key effects arise because real-world implementations cannot collect an infinite number of photons. The accumulation time (how long data is collected before processing) affects both security and key generation rate.


#### Accumulation Time Trade-offs:
* Short accumulation time (e.g., 10 s):Faster key generation.but Large uncertainty in  g(2)(0) (e.g., ±43% in Fig. 4a), making security analysis unreliable.
* Long accumulation time (e.g., 360 s):more Accurate g (2)(0)(±6% error), stronger security.Slower key rate, longer delays before key distillation.

* Optimal block size:
60 s blocks: ~1.47 Mbit sifted key,  g(2)(0) error ~16% → Suitable for finite-key analysis.
360 s blocks: ~8.93 Mbit, error ~6% → Near-asymptotic security (ignoring finite-size effects).

###4 Real-Time Security Implications:
If g (2)(0) deviates from the expected value, it could indicate:
Source instability: Drift in SPS performance (e.g., due to temperature fluctuations).
Eavesdropping: An attacker injecting photons or manipulating the channel to alter photon statistics.
By comparing g (2) (0) at Alice (source) and Bob (receiver), discrepancies (e.g., lower g (2)(0) at Bob due to PNS attacks) can be detected.

## Simulation Methodology for Optimizing QKD Performance via Temporal Filtering

The simulations aimed to evaluate how temporal filtering of single-photon pulses could enhance Quantum Key Distribution (QKD) performance under varying experimental conditions. By modeling different combinations of pulse lifetimes and noise levels, the study identified optimal filtering strategies to maximize the secret key rate.


The simulation results, presented in Figure 5 as two-dimensional heatmaps, revealed how the secret key fraction varies with different temporal filter settings. For systems with **low noise and short pulse durations**, the optimal filtering strategy involved using a narrow time window of about 1 ns centered on the pulse peak, yielding a modest but consistent 2.5% improvement in secret key rate.
In contrast, systems operating with **high noise levels** showed dramatically greater benefits from temporal filtering. The high noise, short pulse scenario demonstrated a remarkable 184.5% improvement in secret key rate when using an asymmetric narrow filter window positioned slightly after the pulse peak. This configuration effectively excluded noise-dominated portions of the signal while preserving most of the genuine single-photon events.


The simulations with **longer pulse durations** revealed different optimization patterns. For low-noise systems with long pulses, a moderately wide filter window of about 2 ns centered on the pulse peak provided optimal results, improving the secret key rate by 6%. The most challenging scenario - high noise combined with long pulses - required an extremely narrow filter window of just 0.25 ns positioned slightly before the pulse peak to achieve a 148.3% improvement in secret key rate. These results clearly demonstrate that temporal filtering can significantly enhance QKD performance across all tested conditions, with particularly dramatic improvements in noisy environments.

This research demonstrates how carefully controlling photon detection timing can optimize quantum communication performance. The optimal temporal filtering strategy depends on two key factors: environmental noise levels and photon pulse duration.

## Simulations

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve
from scipy.stats import poisson
from scipy.optimize import minimize_scalar

class QKDOptimizer:
    """
    A class to simulate and optimize QKD performance with single-photon sources
    based on the Kupko et al. paper (npj Quantum Information, 2020).
    """

    def __init__(self, pulse_width=1.5, detector_jitter=0.5, rep_rate=80e6):
        """
        Initialize the QKD optimizer with system parameters.

        Parameters:
        - pulse_width: Radiative lifetime of quantum dot (ns)
        - detector_jitter: Temporal resolution of detectors (ns)
        - rep_rate: Pulse repetition rate (Hz)
        """
        self.pulse_width = pulse_width  # ns
        self.detector_jitter = detector_jitter  # ns
        self.rep_rate = rep_rate  # Hz

        self.dark_count_rate = 10e3  # Hz (from paper)
        self.p_dc = self.dark_count_rate / rep_rate  # Dark count probability per pulse

        # Optical imperfections parameters (from paper)
        self.q = 0.004  # Error from optical imperfections
        self.detection_efficiency = 0.1  # Typical value

        # Generate synthetic pulse shape
        self.generate_pulse_shape()

    def generate_pulse_shape(self):
        """Generate synthetic single-photon pulse shape (exponential decay convolved with Gaussian)"""
        # Time array
        t = np.linspace(-5, 10, 1500)  # ns

        # Exponential decay
        exp_decay = np.exp(-t/self.pulse_width)
        exp_decay[t < 0] = 0  # Zero before t=0

        # Gaussian detector response
        sigma = self.detector_jitter / (2*np.sqrt(2*np.log(2)))  # Convert FWHM to sigma
        gaussian = np.exp(-t**2 / (2*sigma**2))

        # Convolve to get final pulse shape
        self.pulse = convolve(exp_decay, gaussian, mode='same')
        self.pulse = self.pulse / np.max(self.pulse)  # Normalize
        self.time_axis = t - t[np.argmax(self.pulse)]  # Center at t=0

    def apply_temporal_filter(self, delta_t, t_c):
        """
        Apply temporal filter to the pulse and calculate filtered parameters.

        Parameters:
        - delta_t: Acceptance time window width (ns)
        - t_c: Center of acceptance window (ns)

        Returns:
        - filtered_signal: Fraction of signal within window
        - qber: Quantum bit error rate
        - snr: Signal-to-noise ratio
        """
        # Window bounds
        t_start = t_c - delta_t/2
        t_end = t_c + delta_t/2

        # Calculate signal within window
        window_mask = (self.time_axis >= t_start) & (self.time_axis <= t_end)
        signal_in_window = np.sum(self.pulse[window_mask])
        total_signal = np.sum(self.pulse)
        filtered_signal = signal_in_window / total_signal

        # Calculate QBER (from paper equation 1)
        p_signal = filtered_signal * self.detection_efficiency
        p_click = p_signal + self.p_dc
        qber = (self.q * p_signal) / p_click + (self.p_dc/2) / p_click

        # Calculate SNR
        snr = p_signal / (self.p_dc * delta_t / 12.5)  # Normalized to full window

        return filtered_signal, qber, snr

    def secret_key_rate(self, delta_t, t_c, g2=0.089, mu=0.0043):
        """
        Calculate the secret key rate based on temporal filtering parameters.

        Parameters:
        - delta_t: Acceptance time window width (ns)
        - t_c: Center of acceptance window (ns)
        - g2: g^(2)(0) value
        - mu: Mean photon number per pulse

        Returns:
        - Secret key rate fraction
        """
        # Get filtered parameters
        f, qber, _ = self.apply_temporal_filter(delta_t, t_c)

        # Calculate terms for secret key rate (simplified from paper)
        beta = 1 - 0.5 * mu**2 * g2  # Single photon fraction
        r_e = 1 - qber  # Simplified compression function
        h_e = -qber * np.log2(qber) - (1-qber) * np.log2(1-qber) if qber > 0 else 0  # Binary entropy

        # Secret key rate (equation 2 from paper)
        s = 0.5 * f * (beta * r_e - 1.2 * h_e)  # Using f(e)=1.2 from refernces

        return s

    def plot_pulse_and_filter(self, delta_t=2.5, t_c=0):
        """Plot the pulse shape and temporal filter window"""
        plt.figure(figsize=(10, 6))

        # Plot pulse shape
        plt.plot(self.time_axis, self.pulse, label='Single-photon pulse')

        # Plot filter window
        t_start = t_c - delta_t/2
        t_end = t_c + delta_t/2
        window_mask = (self.time_axis >= t_start) & (self.time_axis <= t_end)
        plt.fill_between(self.time_axis[window_mask], 0, self.pulse[window_mask],
                        color='red', alpha=0.3, label='Acceptance window')

        plt.xlabel('Time (ns)')
        plt.ylabel('Normalized intensity')
        plt.title('Single-photon pulse shape with temporal filter')
        plt.legend()
        plt.grid(True)
        plt.show()

    def plot_2d_secret_key_rate(self, delta_t_range=(0.1, 12.5), t_c_range=(-6.25, 6.25),
                              resolution=50, g2=0.089, mu=0.0043):
        """
        Create a 2D plot of secret key rate as a function of window width and center.

        Parameters:
        - delta_t_range: Tuple of (min, max) for window width (ns)
        - t_c_range: Tuple of (min, max) for window center (ns)
        - resolution: Number of points in each dimension
        - g2: g^(2)(0) value
        - mu: Mean photon number per pulse
        """
        # Create grid of parameters
        delta_t_vals = np.linspace(delta_t_range[0], delta_t_range[1], resolution)
        t_c_vals = np.linspace(t_c_range[0], t_c_range[1], resolution)

        # Calculate secret key rate for each combination
        secret_rates = np.zeros((len(delta_t_vals), len(t_c_vals)))

        for i, dt in enumerate(delta_t_vals):
            for j, tc in enumerate(t_c_vals):
                secret_rates[i, j] = self.secret_key_rate(dt, tc, g2, mu)

        # Create the plot
        plt.figure(figsize=(10, 8))

        # Use pcolormesh for the 2D plot
        mesh = plt.pcolormesh(t_c_vals, delta_t_vals, secret_rates,
                             shading='auto', cmap='viridis')
        plt.colorbar(mesh, label='Secret key rate fraction')

        plt.xlabel('Window center $t_c$ (ns)')
        plt.ylabel('Window width $Δt$ (ns)')
        plt.title('Secret key rate as function of temporal filter parameters')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()

        return delta_t_vals, t_c_vals, secret_rates

    def simulate_different_conditions(self):
        """Simulate four different conditions as in Figure 5 of the paper"""
        # Store original parameters
        orig_pulse_width = self.pulse_width
        orig_dark_count_rate = self.dark_count_rate

        # Create figure with 4 subplots
        fig, axs = plt.subplots(2, 2, figsize=(14, 12))
        plt.subplots_adjust(hspace=0.3, wspace=0.3)

        # Condition 1: Low noise and long pulse (1.5 ns)
        self.pulse_width = 1.5
        self.dark_count_rate = 10e3  # 10 kHz
        self.generate_pulse_shape()
        self.plot_secret_key_2d(axs[0,0], fig, "a) Low noise, long pulse (1.5 ns)")

        # Condition 2: High noise and short pulse (0.5 ns)
        self.pulse_width = 0.5
        self.dark_count_rate = 300e3  # 300 kHz
        self.generate_pulse_shape()
        self.plot_secret_key_2d(axs[0,1], fig, "b) High noise, short pulse (0.5 ns)")

        # Condition 3: Low noise and short pulse (0.5 ns)
        self.pulse_width = 0.5
        self.dark_count_rate = 10e3  # 10 kHz
        self.generate_pulse_shape()
        self.plot_secret_key_2d(axs[1,0], fig, "c) Low noise, short pulse (0.5 ns)")

        # Condition 4: High noise and long pulse (1.5 ns)
        self.pulse_width = 1.5
        self.dark_count_rate = 300e3  # 300 kHz
        self.generate_pulse_shape()
        self.plot_secret_key_2d(axs[1,1], fig, "d) High noise, long pulse (1.5 ns)")

        # Restore original parameters
        self.pulse_width = orig_pulse_width
        self.dark_count_rate = orig_dark_count_rate
        self.generate_pulse_shape()

        plt.show()

    def plot_secret_key_2d(self, ax, fig, title):
        """Create 2D plot of secret key rate for a specific condition"""
        # Parameter ranges
        delta_t_vals = np.linspace(0.1, 12.5, 50)
        t_c_vals = np.linspace(-5, 5, 50)

        # Calculate secret key rate for each combination
        secret_rates = np.zeros((len(delta_t_vals), len(t_c_vals)))

        for i, dt in enumerate(delta_t_vals):
            for j, tc in enumerate(t_c_vals):
                secret_rates[i, j] = self.secret_key_rate(dt, tc)

        # Create the plot
        mesh = ax.pcolormesh(t_c_vals, delta_t_vals, secret_rates,
                            shading='auto', cmap='viridis')
        fig.colorbar(mesh, ax=ax, label='Secret key rate fraction')

        # Mark the optimal point
       # opt_dt, opt_tc, max_rate = self.optimize_window()
       # ax.scatter(opt_tc, opt_dt, color='red', s=100,
       #           label=f'Optimal: Δt={opt_dt:.2f}ns')

        ax.set_xlabel('Window center $t_c$ (ns)')
        ax.set_ylabel('Window width $Δt$ (ns)')
        ax.set_title(title)
        ax.legend()
        ax.grid(True, alpha=0.3)

# Example usage
if __name__ == "__main__":
    # Initialize with parameters similar to the paper
    qkd = QKDOptimizer(pulse_width=1.5, detector_jitter=0.5)

    # Plot the pulse shape with example filter window
    qkd.plot_pulse_and_filter(delta_t=2.5, t_c=0)

    # Create 2D plot of secret key rate vs filter parameters (similar to Fig 2d in paper)
    qkd.plot_2d_secret_key_rate()

    # Create Figure 5 equivalent
    qkd.simulate_different_conditions()
```

# 3- Discussion

### Key Achievement

* Demonstrated that temporal filtering (adjusting detection time windows) significantly improves QKD performance when using single-photon sources (SPS).

### Experimental Validation

* Tested on a solid-state quantum dot SPS with polarization encoding (BB84 protocol).

* Showed optimization can:

  * Maximize secret key rates

  * Extend maximum tolerable loss (up to 38 dB, critical for long-distance QKD).

### Security Advantages

* Real-time monitoring of photon statistics (𝑔²(0)) detects tampering:

  * Ensures single-photon emission (𝑔²(0) < 1).

  * Counters photon-number-splitting attacks.

* Forces attackers to use SPS (harder to exploit than lasers).

### Broad Applicability

* Method works for:

  * Measurement-device-independent QKD (more secure against detector attacks).

  * Quantum repeaters (long-distance networks).

* Enables QKD certification by verifying photon statistics.

### Limitations & Future Work

  * Finite-key effects: Temporal filtering reduces raw key size, requiring better error correction.

  * Efficiency: Must improve SPS brightness (especially for telecom wavelengths).

  * Side-channel attacks: Additional protections needed for full implementations.

# 4- Results

### Single-Photon Source

* Quantum dot (QD) at 918 nm, pulsed at 80 MHz.

* Measured stats: Antibunching 𝑔²(0) = 0.089, mean photons/pulse μ = 0.0043.

### QKD Setup
* BB84 protocol with 4 polarizations (H/V/D/A).

* Detectors: Timing jitter 500–600 ps, dark counts 10–300 kHz.

### Key Formulas

* QBER
$$ QBER = \frac{q \cdot p_{signal}}{p_{click}} + \frac{p_{dc}}{2 \cdot p_{click}} $$

* Secret Key Rate (S)
$$ S = \frac{p_{click}}{2} \left[ \left( 1 - \frac{1}{2} \mu^2 g^{(2)}(0) \right) r(e) - f(e) h(e) \right] $$

### Optimization
* Temporal filtering: Adjusted time window (Δ𝑡 = 1–12.5 ns) to maximize key rate.

* Security: Real-time 𝑔²(0) monitoring to detect attacks.

### Key Results

* Achieved QBER ≈ 0.48% after optimization.

* Tolerable channel loss up to 38 dB with filtering.
