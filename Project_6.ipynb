## **1. Introduction**

### 1.1) What is Quantization of Light?

#### Classical View of Light
- In classical physics, light is described as a **continuous electromagnetic wave**, as per **Maxwell's equations**.

#### Limitations of Classical Theory
- Some phenomena, like the **photoelectric effect**, could not be explained by classical theory.
- This indicated a deeper underlying structure to the nature of light.

#### Quantum Perspective
- In quantum mechanics, the **electromagnetic field is quantized**.
- This means **energy is not continuous**, but exists in **discrete packets** called **photons**.

#### Photon Energy
- Each photon carries energy given by the **Planck relation**:
  
  $$
  E = h \nu
  $$
  
  where $( \nu )$ is the frequency of light, and  (h)  is Planck’s constant.

#### Motivation for Quantization
- Quantization explains why **only light above a certain frequency** can eject electrons in the **photoelectric effect**, regardless of intensity.

#### Harmonic Oscillator Analogy
- Each **mode** of the electromagnetic field behaves like a **quantum harmonic oscillator**.
- It has **discrete energy levels**, unlike the **continuous energy** of a classical oscillator.

#### Why Start with a Single-Mode Field?
- To understand quantized fields, we begin with a **single mode** of light confined in a cavity.
- This **simplifies the problem** and builds physical intuition.

#### Role of the Cavity
- By placing **conducting plates** (boundaries), we trap light inside and force it to form **standing waves**.
- These are **discrete modes** with fixed frequency, direction, and polarization.

#### Benefit of Confinement
- This controlled setup allows us to treat the field as a **single quantum harmonic oscillator**.
- It is a **foundational step** toward understanding more **complex quantum optical systems**.

### 1.2) Motivation

The study of quantum multimode fields is motivated by both **fundamental** and **technological** interests. Multimode field quantization offers:

- A deeper understanding of superposition, entanglement, and decoherence in realistic quantum systems.
- New approaches to scalable and reconfigurable quantum networks by coupling modes through nonlinear optics.
- A framework for describing quantum fluctuations that underlie noise, sensitivity, and measurement precision in quantum systems.
- A unified platform to explore various degrees of freedom—spatial, spectral, temporal, and polarization—enabling multidimensional control of light.

By investigating multimode quantum fields, researchers aim to enhance the capabilities of **quantum computing**, **secure communication**, **quantum sensing**, and **fundamental tests of quantum mechanics**.



## 2. Field Quantization

### 2.1) Quantization of a single-mode field

We begin with the  case of a radiation field confined to a one-dimensional cavity along the $z$-axis with perfectly conducting walls at $z = 0$ and $z = L$, as shown in *the below figure.*

The electric field must vanish on the boundaries and will take the form of a standing wave. We assume there are no sources of radiation, i.e., **no currents or charges nor any dielectric media in the cavity**. The field is assumed to be polarized along the $x$-direction, $E(x,t) = \mathbf{e}_x E_x (z,t)$, where $\mathbf{e}_x$ is a unit polarization vector.

Maxwell’s equations without sources are, in SI units,

$$
\nabla \times \mathbf{E} = -\frac{\partial \mathbf{B}}{\partial t} \tag{2.1}
$$
$$
\nabla \times \mathbf{B} = \mu_0 \varepsilon_0 \frac{\partial \mathbf{E}}{\partial t} \tag{2.2}
$$

$$
\nabla \cdot \mathbf{B} = 0 \tag{2.3}
$$

$$
\nabla \cdot \mathbf{E} = 0 \tag{2.4}
$$

A single-mode field satisfying Maxwell’s equations and the boundary conditions is given by

$$
E_x (z,t) = \left( \frac{2\omega^2}{V \varepsilon_0} \right)^{1/2} q(t) \sin(kz)  \tag{2.5}
$$

where $\omega$ is the frequency of the mode and $k$ is the wave number related to the frequency according to $\omega = c k$. The boundary condition at $z = L$ yields the allowed frequencies $\omega_k = c (\pi n / L)$, $n = 1, 2, \dots$.

We assume that $\omega$ in Eq. (2.5) is one of these frequencies and ignore the rest for now. In Eq. (2.5), $V$ is the effective volume of the cavity and $q(t)$ is a time-dependent factor having the dimension of length. As we shall see, $q(t)$ will act as a canonical position.

The magnetic field in the cavity, from Eq. (2.5) and Eq. (2.2), is $\mathbf{B}(y,t) = \mathbf{e}_y B_y(z,t)$, where

$$
B_y (z,t) = \left( \frac{\mu_0}{\varepsilon_0} \right)^{1/2} \left( \frac{2\omega^2}{V \varepsilon_0} \right)^{1/2} \dot{q}(t) \cos(kz)  \tag{2.6}
$$

Here, $q(t)$ will play the role of a canonical momentum for a “particle” of unit mass, i.e., $p(t) = \dot{q}(t)$.

The classical field energy, or Hamiltonian $H$, of the single-mode field is given by

$$
H = \frac{1}{2} \int_V \left[ \varepsilon_0 E_x^2 (z,t) + \frac{1}{\mu_0} B_y^2 (z,t) \right] dV
$$

$$
= \frac{1}{2} \int_V \varepsilon_0 E_x^2 (z,t) dV + \frac{1}{2} \int_V \frac{1}{\mu_0} B_y^2 (z,t) dV  \tag{2.7}
$$

From Eqs. (2.5) and (2.6), it is straightforward to show that

$$
H = \frac{1}{2} (\dot{q}^2 + \omega^2 q^2) \tag{2.8}
$$

from which it is apparent that a single-mode field is formally equivalent to a classical harmonic oscillator of unit mass, where the electric and magnetic field appear as position and momentum-like state factors, by the use of an appropriate position coordinate.

#### Quantization:-

We replace the classical variables $q$ and $p$ with operators $\hat{q}$ and $\hat{p}$ that satisfy the canonical commutation relation:

$$
[\hat{q}, \hat{p}] = i \hbar \tag{2.9}
$$

Then the quantized electric and magnetic field operators become:

$$
\hat{E}_x(z, t) = \left( \frac{2 \omega^2}{V \varepsilon_0} \right)^{1/2} \hat{q}(t) \sin(kz) \tag{2.10}
$$

$$
\hat{B}_y(z, t) = \left( \frac{\mu_0}{k} \cdot \frac{2 \omega^2}{V} \right)^{1/2} \dot{\hat{q}}(t) \cos(kz) \tag{2.11}
$$

The quantized Hamiltonian becomes:

$$
\hat{H} = \frac{1}{2} \left( \hat{p}^2 + \omega^2 \hat{q}^2 \right) \tag{2.12}
$$

---

#### Creation and Annihilation Operators :-

We define the annihilation and creation operators as:

$$
\hat{a} = \left( \frac{1}{2 \hbar \omega} \right)^{1/2} (\omega \hat{q} + i \hat{p}) \tag{2.13}
$$

$$
\hat{a}^\dagger = \left( \frac{1}{2 \hbar \omega} \right)^{1/2} (\omega \hat{q} - i \hat{p}) \tag{2.14}
$$

### Inverting the Equations

Add and subtract the above expressions to find $ \hat{q} $ and $ \hat{p} $:

#### Position Operator $ \hat{q} $:

$$
\hat{q} = \left( \frac{\hbar}{2 \omega} \right)^{1/2} (\hat{a} + \hat{a}^\dagger)\tag{2.14.1}
$$

#### Momentum Operator $ \hat{p} $:

$$
\hat{p} = i \left( \frac{\hbar \omega}{2} \right)^{1/2} (\hat{a}^\dagger - \hat{a})\tag{2.14.2}
$$

---

- *Position operator* in terms of $ \hat{a} $ and $ \hat{a}^\dagger $:

  $$
  \boxed{
  \hat{q} = \left( \frac{\hbar}{2 \omega} \right)^{1/2} (\hat{a} + \hat{a}^\dagger)
  }\tag{2.14.3}
  $$

- *Momentum operator* in terms of $ \hat{a} $ and $ \hat{a}^\dagger $:

  $$
  \boxed{
  \hat{p} = i \left( \frac{\hbar \omega}{2} \right)^{1/2} (\hat{a}^\dagger - \hat{a})\tag{2.14.4}
  }
  $$

Electric and magnetic field operators become:

$$
\hat{E}_x(z, t) = E_0 (\hat{a} + \hat{a}^\dagger) \sin(kz) \tag{2.15}
$$

$$
\hat{B}_y(z, t) = B_0 (\hat{a} - \hat{a}^\dagger) \cos(kz) \tag{2.16}
$$

where:

- $E_0 = \left( \frac{\hbar \omega}{\varepsilon_0 V} \right)^{1/2}$  
- $B_0 = \left( \frac{\mu_0}{k} \cdot \frac{\varepsilon_0 \hbar \omega^3}{V} \right)^{1/2}$

These represent the field amplitudes "per photon".

The operators $\hat{a}$ and $\hat{a}^\dagger$ satisfy:

$$
[\hat{a}, \hat{a}^\dagger] = 1 \tag{2.17}
$$

So the Hamiltonian becomes:

$$
\hat{H} = \hbar \omega \left( \hat{a}^\dagger \hat{a} + \frac{1}{2} \right) \tag{2.18}
$$

---

#### Time Evolution in Heisenberg Picture :-

Using Heisenberg’s equation:

$$
\frac{d \hat{O}}{dt} = \frac{i}{\hbar} [\hat{H}, \hat{O}] \tag{2.19}
$$

Applying to $\hat{a}$:

$$
\frac{d \hat{a}}{dt} = -i \omega \hat{a} \tag{2.20}
$$
$$
\Rightarrow \hat{a}(t) = \hat{a}(0) e^{-i \omega t} \tag{2.21}
$$

And for the creation operator:

$$
\hat{a}^\dagger(t) = \hat{a}^\dagger(0) e^{i \omega t} \tag{2.22}
$$

Define number operator:

$$
\hat{n} = \hat{a}^\dagger \hat{a} \tag{2.23}
$$

Let $|n\rangle$ be the number (Fock) state:

$$
\hat{H} |n\rangle = \hbar \omega \left(\hat{n} + \frac{1}{2} \right) |n\rangle \tag{2.24}
$$

Apply $\hat{a}^\dagger$:

$$
\hat{H} (\hat{a}^\dagger |n\rangle) = (E_n + \hbar \omega)(\hat{a}^\dagger |n\rangle) \tag{2.25}
$$

Apply $\hat{a}$:

$$
\hat{H} (\hat{a} |n\rangle) = (E_n - \hbar \omega)(\hat{a} |n\rangle) \tag{2.26}
$$

Ground state $|0\rangle$ satisfies:

$$
\hat{a} |0\rangle = 0 \tag{2.27}
$$

Hence, zero-point energy:

$$
\hat{H} |0\rangle = \frac{1}{2} \hbar \omega |0\rangle \tag{2.28}
$$

General energy levels:

$$
E_n = \hbar \omega \left(n + \frac{1}{2} \right), \quad n = 0, 1, 2, \dots \tag{2.29}
$$

### 2.2 Quantum fluctuations of a single-mode field

The number state $\lvert n \rangle$ is a state of well-defined energy but it is not a state of well-defined electric field since

$$
\langle n | \hat{E}_x(z, t) | n \rangle = E_0 \sin(kz) \left[ \langle n | \hat{a} | n \rangle + \langle n | \hat{a}^\dagger | n \rangle \right] = 0 \tag{2.37}
$$

i.e. the mean field is zero. But the mean of the square of this field, which contributes to the energy density, is not zero:

$$
\langle n | \hat{E}_x^2(z, t) | n \rangle = E_0^2 \sin^2(kz) \langle n | \hat{a}^{\dagger 2} + \hat{a}^2 + \hat{a}^\dagger \hat{a} + \hat{a} \hat{a}^\dagger | n \rangle \tag{2.38}
$$

$$
= E_0^2 \sin^2(kz) \langle n | \hat{a}^{\dagger 2} + \hat{a}^2 + 2 \hat{a}^\dagger \hat{a} + 1 | n \rangle \tag{2.39}
$$

$$
= 2 E_0^2 \sin^2(kz) (n + 1)
$$

The fluctuations in the electric field may be characterized by the **variance**

$$
\langle (\Delta \hat{E}_x(z, t))^2 \rangle = \langle \hat{E}_x^2(z, t) \rangle - \langle \hat{E}_x(z, t) \rangle^2 \tag{2.40}
$$

or by the **standard deviation**

$$
\Delta E_x = \sqrt{ \langle (\Delta \hat{E}_x(z, t))^2 \rangle } \tag{2.41}
$$

which is sometimes referred to as the **uncertainty of the field**.  
For the number state $\lvert n \rangle$, even when $n = 0$, the field has fluctuations—the so-called **vacuum fluctuations**.

Now the number states $\lvert n \rangle$ are taken to represent a state of the field containing $n$ photons. Yet as we have seen, the average field is zero. This is all in accordance with the uncertainty principle because the **number operator** $\hat{n}$ does not commute with the electric field:

$$
[\hat{n}, \hat{E}_x] = E_0 \sin(kz) (\hat{a}^\dagger - \hat{a}) \tag{2.42}
$$

Thus $\hat{n}$ and $\hat{E}_x$ are complementary quantities for which their respective uncertainties obey the inequality:

$$
\Delta n \, \Delta E \geq E_0 |\sin(kz)| \cdot |\langle \hat{a}^\dagger - \hat{a} \rangle| \tag{2.43}
$$

For a number state $\lvert n \rangle$, the right-hand side vanishes, but $\Delta n = 0$ as well.  
If the field were accurately known, then the number of photons would be uncertain.

There is a connection here to the notion of the **phase** of the electric field.  
In classical physics, the amplitude and phase of a field can be simultaneously well-defined. This is **not** so in quantum mechanics.

In fact, the history of the concept of a quantum phase operator is long and contentious, and we shall deal with this issue at length later. For now, we simply take a heuristic point of view, for which the phase is in some sense **complementary** to number, much in the way that time is complementary to energy.

By analogy to the time–energy uncertainty relation, there should be a **number–phase uncertainty relation** of the form:

$$
\Delta n \, \Delta \phi \geq 1 \tag{2.44}
$$

From this, one could argue that for a well-defined (accurately known) phase, the photon number is uncertain, whereas for a well-defined photon number, the phase is uncertain and, in fact, the phase is randomly distributed over the range:

$$
0 < \phi < 2\pi
$$

### 2.3 Quadrature operators for a single-mode field

When we explicitly include the time dependence of the electric field operator we have:

$$
\hat{E}_x(z, t) = E_0 (\hat{a} e^{-i\omega t} + \hat{a}^\dagger e^{i\omega t}) \sin(kz) \tag{2.45}
$$

We now introduce the so-called **quadrature operators**:

$$
\hat{X}_1 = \frac{1}{2} (\hat{a} + \hat{a}^\dagger) \tag{2.46}
$$

$$
\hat{X}_2 = \frac{1}{2i} (\hat{a} - \hat{a}^\dagger) \tag{2.47}
$$

In terms of these, the field operator may be recast as:

$$
\hat{E}_x(t) = 2 E_0 \sin(kz) [\hat{X}_1 \cos(\omega t) + \hat{X}_2 \sin(\omega t)] \tag{2.48}
$$

It is evident that $\hat{X}_1$ and $\hat{X}_2$ are associated with field amplitudes oscillating **out of phase by $90^\circ$** (and hence are in quadrature).

Note that $\hat{X}_1$ and $\hat{X}_2$ are essentially the **position** and **momentum** operators (from Eqs. 2.13 and 2.14), but scaled to be **dimensionless**.

They satisfy the **commutation relation**:

$$
[\hat{X}_1, \hat{X}_2] = \frac{i}{2} \tag{2.49}
$$

from which it follows that:

$$
\langle (\Delta \hat{X}_1)^2 \rangle \langle (\Delta \hat{X}_2)^2 \rangle \geq \frac{1}{4} \tag{2.50}
$$

For the number states:

$$
\langle n | \hat{X}_1 | n \rangle = \langle n | \hat{X}_2 | n \rangle = 0 \tag{2.51}
$$

but

$$
\langle n | \hat{X}_1^2 | n \rangle = \frac{1}{4} \langle n | \hat{a}^2 + \hat{a}^{\dagger 2} + 2 \hat{a}^\dagger \hat{a} + 1 | n \rangle = \frac{1}{4}(2n + 1) \tag{2.52}
$$

and similarly

$$
\langle n | \hat{X}_2^2 | n \rangle = \frac{1}{4}(2n + 1) \tag{2.53}
$$

Thus, for a number state, the uncertainties in both quadratures are the same. Furthermore, the vacuum state ($n = 0$) **minimizes the uncertainty product** since:

$$
\langle (\Delta \hat{X}_1)^2 \rangle = \langle (\Delta \hat{X}_2)^2 \rangle = \frac{1}{4} \tag{2.54}
$$

The quanta of the single-mode cavity field are excitations of energy in discrete amounts of $\hbar \omega$. These quanta, universally referred to as **photons**, are not localized particles (in field theory, there is no position operator for photons) but are spread out over the entire mode volume.

This is in sharp contrast to the view of photons as "corpuscles" of light as in the old quantum theory.

## 2.4) Multimode Fields

The results from the single-mode case can be extended to multimode radiation fields, meaning fields that support multiple modes of oscillation. Here, we assume:

1. The fields exist in free space.
2. There are no sources (e.g., charges or currents).
3. The Maxwell equations in free space still hold.

These conditions ensure that the field behaves in a pure radiation state without external influences.

#### Mathematical Formulation

To describe the multimode radiation field, we use the vector potential $A(\mathbf{r}, t)$, which satisfies the wave equation:

$$
\nabla^2 A - \frac{1}{c^2} \frac{\partial^2 A}{\partial t^2} = 0 \tag{2.55}
$$

This equation describes how the vector potential propagates in free space. It is the classical wave equation for electromagnetic waves, stating that the second spatial derivative equals the second time derivative scaled by $1/c^2$, where $c$ is the speed of light.

Additionally, the Coulomb gauge condition is imposed:

$$
\nabla \cdot A(\mathbf{r}, t) = 0 \tag{2.56}
$$

This condition simplifies calculations by eliminating unphysical degrees of freedom from the vector potential.

#### Electric and Magnetic Fields from the Vector Potential

From the vector potential $A(\mathbf{r}, t)$, we can derive the electric and magnetic fields:

**Electric Field:**

$$
E(\mathbf{r}, t) = - \frac{\partial A(\mathbf{r}, t)}{\partial t} \tag{2.57}
$$

This equation shows that the electric field is obtained by taking the time derivative of the vector potential with a negative sign.

**Magnetic Field:**

$$
B(\mathbf{r}, t) = \nabla \times A(\mathbf{r}, t) \tag{2.58}
$$

This means that the magnetic field is given by the curl of the vector potential. These relations follow directly from Maxwell’s equations and are common in classical electromagnetism.

#### Modeling Free Space as a Large Cavity

To make calculations more manageable, the text introduces a cubic cavity of side length $L$ with perfectly reflecting walls. The reasoning behind this assumption is:

1. The cavity is chosen to be large compared to anything inside (e.g., atoms).
2. The cavity is much larger than the wavelength of the electromagnetic fields.
3. Any physical results derived using this model should remain valid even as we take the limit $L \to \infty$.

This approach is common in physics because imposing boundary conditions in a finite system simplifies the analysis while still capturing the essential physics.

#### Imposing Periodic Boundary Conditions

To properly describe the allowed field modes, periodic boundary conditions are applied. Specifically, in the x-direction, the plane wave solutions must satisfy:

$$ e^{i k_x x} = e^{i k_x (x + L)} \tag{2.59} $$

This means that the wavefunction (or field amplitude) at position $x$ must be the same as at $x + L$, enforcing a repeating pattern.

From this condition, we derive the allowed values for the wavevector $k_x$:

$$ k_x = \frac{2\pi}{L} m_x, \quad m_x = 0, \pm1, \pm2, \dots \tag{2.60} $$

This equation tells us:

1. The wavevector $k_x$ is quantized in discrete steps of $2\pi / L$.
2. The integer $m_x$ determines the mode number.
3. The same quantization applies to the other spatial dimensions (y and z).

Since the field is in free space, we later take $L \to \infty$, which allows $k_x$ to take continuous values rather than being strictly quantized.

#### Quantization of Wavevector Components

In a cubic cavity of side length $L$, the allowed wavevectors in the y- and z-directions follow the same quantization rule as in the x-direction:

$$ k_y = \left(\frac{2\pi}{L}\right) m_y, \quad m_y = 0, \pm1, \pm2, \dots $$
$$ k_z = \left(\frac{2\pi}{L}\right) m_z, \quad m_z = 0, \pm1, \pm2, \dots $$

This comes from applying periodic boundary conditions, ensuring that a wave function repeats identically after a translation by $L$.



The total wavevector $k$ is then:

$$ \mathbf{k} = \frac{2\pi}{L} (m_x, m_y, m_z) \tag{2.61} $$

The magnitude of the wavevector is related to the frequency of the wave by the dispersion relation:

$$ k = \frac{\omega_k}{c} \tag{2.62} $$

where $c$ is the speed of light and $\omega_k$ is the frequency associated with the wavevector $k$.

#### Counting the Number of Modes in a Volume

Each set of integers $(m_x, m_y, m_z)$ corresponds to a distinct mode of the field. Since the integers can take positive and negative values, the number of possible modes is large but countable. Instead of treating free space as having a continuous spectrum of wavevectors, we approximate it by summing over discrete modes.

The total number of modes within small intervals $\Delta m_x, \Delta m_y, \Delta m_z$ is given by:

$$ \Delta m = \Delta m_x \Delta m_y \Delta m_z = 2 \left( \frac{L}{2\pi} \right)^3 \Delta k_x \Delta k_y \Delta k_z \tag{2.63} $$

#### Approximating Mode Density in Continuous Space


For a sufficiently large cavity, we approximate the discrete mode count using a differential form:

$$ dm = 2 \left( \frac{V}{8\pi^3} \right) dk_x dk_y dk_z \tag{2.64} $$

where we set $V = L^3$, the volume of the cubic cavity.

Now, we express the wavevector in spherical coordinates:

$$ \mathbf{k} = k ( \sin\theta \cos\phi, \sin\theta \sin\phi, \cos\theta )\tag{2.65} $$

In spherical coordinates, the volume element $dk_x dk_y dk_z$ transforms into:

$$ dm = 2 \left( \frac{V}{8\pi^3} \right) k^2 dk d\Omega \tag{2.66} $$

where:

$$ d\Omega = \sin\theta d\theta d\phi $$

is the solid angle element, and $k^2 dk$ arises from the Jacobian of the transformation from Cartesian to spherical coordinates.

Using the relation $k = \frac{\omega_k}{c}$, we rewrite this in terms of frequency:

$$ dm = 2 \left( \frac{V}{8\pi^3} \right) \frac{\omega_k^2}{c^3} d\omega_k d\Omega \tag{2.67} $$

#### Integrating Over the Solid Angle

To obtain the total number of modes in a spherical shell between $k$ and $k+dk$, we integrate over all solid angles:

$$ \left\{ \text{number of modes in all directions in range } k \text{ to } k+dk \right\} = V \frac{k^2}{\pi^2} dk = V \rho_k dk \tag{2.68}$$

where:

$$ \rho_k = \frac{k^2}{\pi^2} $$

is the density of states per unit volume in $k$-space.

### Field Quantization and Mode Density

#### Mode Density in Terms of Frequency
The mode density, which represents the number of available modes per unit volume, is given by:

$$\rho_k dk = \frac{k^2}{\pi^2} dk \tag{2.69}$$

 integrating over all solid angles gives:

$$\text{Number of modes in the range } \omega_k \text{ to } \omega_k + d\omega_k = \frac{V \omega_k^2}{\pi^2 c^3} d\omega_k \tag{2.70}$$

which is rewritten using mode density as:

$$V \rho(\omega_k) d\omega_k, \quad \text{where } \rho(\omega_k) = \frac{\omega_k^2}{\pi^2 c^3} \tag{2.71}$$

This tells us that the number of modes available increases proportionally to the square of the frequency.

#### Plane Wave Expansion of the Vector Potential
The vector potential $A(r,t)$ can be represented as a superposition of plane waves:

$$A(r,t) = \sum_{k,s} e_{k,s} \left[A_{k,s}(t) e^{i k \cdot r} + A_{k,s}^{*}(t) e^{-i k \cdot r}\right] \tag{2.72}$$

where:
- $A_{k,s}(t)$ is the complex amplitude of the field.
- $e_{k,s}$ are real polarization vectors.
- The summation over $k$ runs over all possible wavevectors.
- $s$ represents the two independent polarization states of the electromagnetic field.

#### Orthogonality Conditions for Polarizations
For an electromagnetic wave, the polarization vectors $e_{k,1}$ and $e_{k,2}$ must be orthogonal:

$$e_{k,s} \cdot e_{k,s'} = \delta_{s,s'} \tag{2.73}$$

Also, from the gauge condition

$$k \cdot e_{k,s} = 0$$

This condition ensures that the polarization vectors are perpendicular to the wavevector, which is a fundamental requirement for transverse waves.

The polarization vectors form a right-handed coordinate system with the wavevector:

$$e_{k,1} \times e_{k,2} = \frac{k}{|k|} = \hat{k} \tag{2.74}$$

This ensures consistency in the definition of polarization directions.

#### Continuous Integral Form of the Mode Sum
In free space, the summation over discrete wavevectors is replaced by an integral:

$$\sum_k \rightarrow \frac{V}{\pi^2} \int k^2 dk \tag{2.75}$$

This replaces summation by an integral over wavevector magnitudes, making the field representation continuous in k-space.

#### Harmonic Oscillator Representation of Mode Amplitudes
The evolution of the complex field amplitudes $A_{k,s}(t)$ follows the harmonic oscillator equation:

$$\frac{d^2 A_{k,s}}{dt^2} + \omega_k^2 A_{k,s} = 0 \tag{2.76}$$

which has the general solution:

$$A_{k,s}(t) = A_{k,s} e^{-i \omega_k t} \tag{2.77}$$

where the initial condition is set as $A_{k,s}(0) = A_{k,s}$.

#### Electric and Magnetic Field Components
Using Maxwell’s equations, the electric and magnetic fields are obtained from the vector potential:

**Electric Field:**

$$E(r,t) = i \sum_{k,s} \omega_k e_{k,s} \left[A_{k,s} e^{i (k \cdot r - \omega_k t)} - A_{k,s}^{*} e^{-i (k \cdot r - \omega_k t)}\right] \tag{2.78}$$

**Magnetic Field:**

$$B(r,t) = \frac{i}{c} \sum_{k,s} \omega_k (k \times e_{k,s}) \left[A_{k,s} e^{i (k \cdot r - \omega_k t)} - A_{k,s}^{*} e^{-i (k \cdot r - \omega_k t)}\right] \tag{2.79}$$

These expressions show how the quantized electromagnetic fields evolve in space and time.

#### Energy of the Electromagnetic Field

The total energy of the electromagnetic field is given by:

$$ H = \frac{1}{2} \int_V \left( \varepsilon_0 \mathbf{E} \cdot \mathbf{E} + \frac{1}{\mu_0} \mathbf{B} \cdot \mathbf{B} \right) dV. \tag{2.80} $$

This equation states that the total field energy is composed of electric field energy and magnetic field energy integrated over a volume $V$.

#### Periodic Boundary Conditions and Mode Expansion

Since we are dealing with a quantized field in a finite volume, we impose periodic boundary conditions to simplify the mathematics:

$$ \int_0^L e^{i k_x x} dx = \begin{cases} L, & k_x = 0, \\ 0, & k_x \neq 0. \end{cases}\tag{2.81} $$

This ensures that only specific discrete wavevectors $k$ contribute to the field modes, enforcing a discrete mode structure in a finite-sized cavity.

This result extends similarly to the $y$- and $z$-directions, leading to the general result:

$$ \int_V e^{i(\mathbf{k} - \mathbf{k'}) \cdot \mathbf{r}} dV = \delta_{\mathbf{k}, \mathbf{k'}} V.\tag{2.82} $$

This equation ensures orthogonality of modes: different wavevectors $k$ are independent and only identical wavevectors contribute to the integral.

\mathbf{E} \cdot \mathbf{E} = \sum_{\mathbf{k}s} \sum_{\mathbf{k}'s'} \mathbf{E}_{\mathbf{k}s} \cdot \mathbf{E}_{\mathbf{k}'s'}\tag{2.87}

Expanding the dot product:

$$
\begin{align}
\mathbf{E}_{\mathbf{k}s} \cdot \mathbf{E}_{\mathbf{k}'s'} &= -\omega_k \omega_{k'} \hat{\epsilon}_{\mathbf{k}s} \cdot \hat{\epsilon}_{\mathbf{k}'s'} \Bigg[ A_{\mathbf{k}s} A_{\mathbf{k}'s'} e^{i[(\mathbf{k}+\mathbf{k}') \cdot \mathbf{r} - (\omega_k + \omega_{k'}) t]} \\
&\quad - A_{\mathbf{k}s} A^*_{\mathbf{k}'s'} e^{i[(\mathbf{k}-\mathbf{k}') \cdot \mathbf{r} - (\omega_k - \omega_{k'}) t]} \\
&\quad - A^*_{\mathbf{k}s} A_{\mathbf{k}'s'} e^{-i[(\mathbf{k}-\mathbf{k}') \cdot \mathbf{r} - (\omega_k - \omega_{k'}) t]} \\
&\quad + A^*_{\mathbf{k}s} A^*_{\mathbf{k}'s'} e^{-i[(\mathbf{k}+\mathbf{k}') \cdot \mathbf{r} - (\omega_k + \omega_{k'}) t]} \Bigg]
\end{align}\tag{2.88}
$$


After integration, oscillatory terms average out, leaving only terms for $\mathbf{k} = \mathbf{k}'$ and $s = s'$.

---

*Apply periodic boundary conditions*

$$
\int_V e^{i(\mathbf{k} - \mathbf{k}') \cdot \mathbf{r}} dV = \delta_{\mathbf{k}, \mathbf{k}'} V\tag{2.89}
$$

Thus, integrating over the volume:

$$
\int_V \mathbf{E} \cdot \mathbf{E} dV = 2 \sum_{\mathbf{k}, s} \omega_k^2 |A_{\mathbf{k}s}(t)|^2 V - R\tag{2.90}
$$

where $R$ collects the oscillatory terms.

---



Substituting into $H_E$:

$$
H_E = \frac{1}{2} \varepsilon_0 \int_V \mathbf{E} \cdot \mathbf{E} dV = \frac{1}{2} \varepsilon_0 \left( 2 \sum_{\mathbf{k}, s} \omega_k^2 |A_{\mathbf{k}s}(t)|^2 V - R \right)\tag{2.91}
$$

Simplifies to:

$$
H_E = \varepsilon_0 V \sum_{\mathbf{k}, s} \omega_k^2 A_{\mathbf{k}s}(t) A_{\mathbf{k}s}^*(t) - R\tag{2.92}
$$


The remaining term is given by:

$$
R = \frac{1}{2} \varepsilon_0 \sum_{\mathbf{k}s \neq \mathbf{k}'s'} \omega_k \omega_{k'} (\hat{\epsilon}_{\mathbf{k}s} \cdot \hat{\epsilon}_{\mathbf{k}'s'}) \left[ A_{\mathbf{k}s}(t) A_{\mathbf{k}'s'}(t) + A^*_{\mathbf{k}s}(t) A^*_{\mathbf{k}'s'}(t) \right]\tag{2.93}
$$


$$ \frac{1}{2} \int_V \varepsilon_0 \mathbf{E} \cdot \mathbf{E} dV = \varepsilon_0 V \sum_{k,s} \omega_k^2 A_{ks}(t) A_{ks}^*(t) - R.\tag{2.94}
$$

Here:

- $A_{ks}(t)$ are the mode amplitudes.
- $\omega_k = |k|c$ is the frequency of each mode.
- $s$ denotes the polarization state.

The term $R$ corrects for mode correlations and is given by:

$$ R = \frac{1}{2} \varepsilon_0 V \sum_{k,s} \omega_k^2 e_{ks} \cdot e_{-ks} \left[ A_{ks}(t) A_{-ks}(t) + A_{ks}^*(t) A_{-ks}^*(t) \right].\tag{2.95} $$

The presence of $R$ accounts for possible dependencies between opposite wavevectors.

#### Magnetic Field Contribution Using Vector Identities

To compute the contribution from the magnetic field, we use a vector identity:

$$ (\mathbf{A} \times \mathbf{B}) \cdot (\mathbf{C} \times \mathbf{D}) = (\mathbf{A} \cdot \mathbf{C}) (\mathbf{B} \cdot \mathbf{D}) - (\mathbf{A} \cdot \mathbf{D}) (\mathbf{B} \cdot \mathbf{C}).\tag{2.96} $$

Applying this identity, we obtain:

$$ (\mathbf{k} \times e_{ks}) \cdot (\mathbf{k} \times e_{ks'}) = \delta_{s,s'}.\tag{2.97} $$

And for opposite wavevectors:

$$ (\mathbf{k} \times e_{ks}) \cdot (-\mathbf{k} \times e_{-ks}) = - e_{ks} \cdot e_{-ks}.\tag{2.98} $$

Using these, we derive the magnetic field energy contribution:

$$ \frac{1}{2} \int_V \frac{1}{\mu_0} \mathbf{B} \cdot \mathbf{B} dV = \varepsilon_0 V \sum_{k,s} \omega_k^2 A_{ks}(t) A_{ks}^*(t) + R.\tag{2.99} $$

#### Total Energy of the Field

Adding both the electric and magnetic energy contributions, the total energy simplifies to:

$$ H = 2 \varepsilon_0 V \sum_{k,s} \omega_k^2 A_{ks}(t) A_{ks}^*(t). \tag{2.100}$$

Or equivalently:

$$ H = 2 \varepsilon_0 V \sum_{k,s} \omega_k^2 A_{ks} A_{ks}^*. \tag{2.101}$$

This represents the total field energy as a sum over individual mode contributions.

#### Introduction of Canonical Variables for Quantization

To move towards quantization of the field, we introduce canonical variables $q_{ks}$ and $p_{ks}$, analogous to a harmonic oscillator:

$$ A_{ks} = \frac{1}{2 \omega_k} (\varepsilon_0 V)^{1/2} \left[ \omega_k q_{ks} + i p_{ks} \right].\tag{2.102} $$

$$ A_{ks}^* = \frac{1}{2 \omega_k} (\varepsilon_0 V)^{1/2} \left[ \omega_k q_{ks} - i p_{ks} \right].\tag{2.103} $$

These variables reformulate the field modes into a form suitable for quantization, where $q_{ks}$ and $p_{ks}$ will later become quantum mechanical operators.

#### where:

$q_{k,s}$ is the generalized coordinate, representing the real part of the field amplitude.

$p_{k,s}$ is the conjugate momentum, representing the imaginary part of the field amplitude.

These variables transform the field into a set of harmonic oscillators.

#### Why Introduce $q_{k,s}$ and $p_{k,s}$?

They allow us to quantize the field using the same methods as for a quantum harmonic oscillator.

They provide a way to express the field modes in a familiar oscillator form.

They lead to commutation relations that define quantum mechanical behavior.

$$

The electric field operator is then

$$
\hat{\mathbf{E}}(\mathbf{r}, t) = i \sum_{k,s} \left( \frac{\hbar \omega_k}{2 \varepsilon_0 V} \right)^{1/2} \mathbf{e}_{ks} \left[ \hat{a}_{ks} e^{i(\mathbf{k} \cdot \mathbf{r} - \omega_k t)} - \hat{a}^\dagger_{ks} e^{-i(\mathbf{k} \cdot \mathbf{r} - \omega_k t)} \right]. \tag{2.126}
$$

while the magnetic field operator is

$$
\hat{\mathbf{B}}(\mathbf{r}, t) = \left( \frac{i}{c} \right) \sum_{k,s} \left( \frac{\hbar \omega_k}{2 \varepsilon_0 V} \right)^{1/2} \left( \hat{\mathbf{k}} \times \mathbf{e}_{ks} \right) \left[ \hat{a}_{ks} e^{i(\mathbf{k} \cdot \mathbf{r} - \omega_k t)} - \hat{a}^\dagger_{ks} e^{-i(\mathbf{k} \cdot \mathbf{r} - \omega_k t)} \right], \tag{2.127}
$$

where $( \hat{\mathbf{k}} = \mathbf{k}/|\mathbf{k}| )$. The annihilation and creation operators appearing are to be understood as Heisenberg picture operators evaluated At time ( t = 0 ), as in the single-mode case, the time-dependent annihilation operator for a free field is given by

$$
\hat{a}_{ks}(t) = \hat{a}_{ks}(0) e^{-i\omega_k t}. \tag{2.128}
$$

Thus, the electric field operator, for instance, can be written as

$$
\hat{\mathbf{E}}(\mathbf{r}, t) = i \sum_{k,s} \left( \frac{\hbar \omega_k}{2\varepsilon_0 V} \right)^{1/2} \mathbf{e}_{ks} \left[ \hat{a}_{ks}(t) e^{i\mathbf{k} \cdot \mathbf{r}} - \hat{a}^\dagger_{ks}(t) e^{-i\mathbf{k} \cdot \mathbf{r}} \right]. \tag{2.129}
$$

Sometimes this field is written as

$$
\hat{\mathbf{E}}(\mathbf{r}, t) = \hat{\mathbf{E}}^{(+)}(\mathbf{r}, t) + \hat{\mathbf{E}}^{(-)}(\mathbf{r}, t), \tag{2.130}
$$

where

$$
\hat{\mathbf{E}}^{(+)}(\mathbf{r}, t) = i \sum_{k,s} \left( \frac{\hbar \omega_k}{2\varepsilon_0 V} \right)^{1/2} \mathbf{e}_{ks} \hat{a}_{ks}(t) e^{i \mathbf{k} \cdot \mathbf{r}}, \tag{2.131}
$$

and

$$
\hat{\mathbf{E}}^{(-)}(\mathbf{r}, t) = \left[ \hat{\mathbf{E}}^{(+)}(\mathbf{r}, t) \right]^\dagger. \tag{2.132}
$$

$(\hat{\mathbf{E}}^{(+)})$ is called the **positive frequency part** of the field, as it contains all terms that oscillate as $( e^{-i\omega t} ) for ( \omega > 0 )$, while $(\hat{\mathbf{E}}^{(-)})$ is called the **negative frequency part**. The former is essentially a collective annihilation operator, while the latter is a collective creation operator.

## 3. Applications of Multimode Quantum Fields

Quantum multimode fields underpin a wide array of emerging applications:

1. **Quantum Imaging**: Exploiting spatial mode correlations to achieve resolution and sensitivity beyond classical limits.
2. **Quantum Information and Computing**: Utilizing high-dimensional multimode entanglement for scalable quantum processors and robust quantum key distribution.
3. **Time/Frequency Domain Applications**: Harnessing temporal and spectral modes for time-bin encoding, frequency comb generation, and ultrafast quantum signal processing.
4. **Quantum Sensing and Metrology**: Enhancing precision measurement devices, including atomic clocks and gravitational wave detectors, by managing multimode quantum noise.
5. **Quantum Communication**: Enabling high-capacity, low-loss transmission through multimode fibers and free-space links using entangled photons across many modes.
6. **Laser Technology**: Understanding and engineering multimode quantum properties to improve coherence, beam quality, and noise control in laser systems.
