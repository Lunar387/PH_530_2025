### **Quantum Information Science and Secure Communication**  

Quantum information science is an emerging field that leverages the unique properties of quantum systems to enhance both **communication** and **computation** beyond what is possible with classical physics. One key area of application is **cryptography**, where traditional public-key methods rely on the assumed difficulty of solving problems like **prime factorization**. However, with quantum computers, these problems can be solved efficiently, making classical cryptographic methods potentially vulnerable.  

In contrast, **quantum cryptography** offers a secure way to communicate by using the principles of quantum mechanics. By transmitting a secret key using **single photons**, any attempt at eavesdropping disturbs the quantum system, making it detectable. The level of potential eavesdropping can be measured through **error rates**, and a process called **privacy amplification** can further reduce any leaked information, ensuring the security of the key.

However, some methods that use weak coherent light pulses can, in rare cases, emit multiple photons, which could allow an eavesdropper to extract information without being detected. This vulnerability can be addressed using **true single-photon sources** or **entangled photon pairs**, which may enable secure communication over longer distances.  

While many experiments have studied **entangled photon pairs** in the context of Bell’s inequalities and quantum correlations, there had been no explicit demonstrations of their use in a cryptographic protocol. However, Ekert’s quantum key distribution protocol (E91) was a major breakthrough, showing that entangled photons can be used to establish a **completely secure cryptographic key** between two distant parties. This highlights the potential of quantum mechanics in revolutionizing secure communication.

## **Experimental Setup**  

The experiment uses **spontaneous parametric down-conversion (SPDC)** to generate photon pairs. A **351.1 nm Argon-ion laser** pumps two **Beta Barium Borate (BBO) crystals**, whose optic axes are **perpendicular** to each other. This setup ensures that **45° polarized photons** can be down-converted in either crystal with equal probability.

The generated photons are sent to **Alice and Bob**, who analyze them using **Liquid Crystal Waveplates (LCWs)** to randomly choose one of four measurement bases. A **half-wave plate (HWP) at 22.5°** rotates the polarization into the **diagonal basis** before measurement. A **polarizing beam splitter (PBS)** then separates the photons based on polarization, directing them to **silicon avalanche photodiodes (Si-APDs)** for detection.


### Mathematical Setup

Due to high spatial overlap and coherence, photon pairs are created in a **maximally entangled state** since the probability of down-conversion is equal in both crystals. In our version of the **Ekert protocol**, Alice and Bob each receive one photon from an **entangled pair** in the state:

$$
|\phi^+\rangle = (|H_1 H_2\rangle + |V_1 V_2\rangle)/\sqrt{2}
$$
where \( H \) and \( V \) represent **horizontal** and **vertical** polarization, respectively. To analyze the entangled state, Alice and Bob measure their photons in the following **rotated bases**:

$$
\left( |H_1\rangle + e^{i\alpha} |V_1\rangle \right), \quad \left( |H_2\rangle + e^{i\beta} |V_2\rangle \right)
$$

where the angles **\( \alpha \) and \( \beta \)** are randomly chosen from:

$$
\alpha_1 = 45^\circ, \quad \alpha_2 = 90^\circ, \quad \alpha_3 = 135^\circ, \quad \alpha_4 = 180^\circ
\$$

$$
\beta_1 = 0^\circ, \quad \beta_2 = 45^\circ, \quad \beta_3 = 90^\circ, \quad \beta_4 = 135^\circ
$$


### Given Entangled State
The Bell state given is:

$$
|\phi^+\rangle = \frac{|H_1 H_2\rangle + |V_1 V_2\rangle}{\sqrt{2}}
$$

This means that if Alice and Bob measure in the **same basis**, their results will always be correlated.

---

### Measurement in the Given Basis
Alice and Bob measure in the basis:

$$
| \psi_A \rangle = |H_1\rangle + e^{i\alpha} |V_1\rangle
$$
$$
| \psi_B \rangle = |H_2\rangle + e^{i\beta} |V_2\rangle
$$

We normalize these:

$$
| \psi_A \rangle = \frac{|H_1\rangle + e^{i\alpha} |V_1\rangle}{\sqrt{2}}
$$
$$
| \psi_B \rangle = \frac{|H_2\rangle + e^{i\beta} |V_2\rangle}{\sqrt{2}}
$$

Their orthogonal states (which detect the opposite polarization) are:

$$
| \psi_A^\perp \rangle = \frac{|H_1\rangle - e^{i\alpha} |V_1\rangle}{\sqrt{2}}
$$
$$
| \psi_B^\perp \rangle = \frac{|H_2\rangle - e^{i\beta} |V_2\rangle}{\sqrt{2}}
$$

---

### Computing Probabilities
Now, we compute the probability that Alice gets $ | \psi_A \rangle $ and Bob gets $ | \psi_B \rangle $, which is:

$$
P_{12}(\alpha, \beta) = |\langle \psi_A, \psi_B | \phi^+ \rangle|^2
$$

Expanding the inner product:

$$
\langle \psi_A, \psi_B | \phi^+ \rangle = \frac{1}{\sqrt{2}} ( \langle \psi_A | H_1 \rangle \langle \psi_B | H_2 \rangle + \langle \psi_A | V_1 \rangle \langle \psi_B | V_2 \rangle )
$$

Since:

$$
\langle \psi_A | H_1 \rangle = \frac{1}{\sqrt{2}}, \quad \langle \psi_A | V_1 \rangle = \frac{e^{-i\alpha}}{\sqrt{2}}
$$

$$
\langle \psi_B | H_2 \rangle = \frac{1}{\sqrt{2}}, \quad \langle \psi_B | V_2 \rangle = \frac{e^{-i\beta}}{\sqrt{2}}
$$

we get:

$$
\langle \psi_A, \psi_B | \phi^+ \rangle = \frac{1}{2} (1 + e^{-i(\alpha + \beta)})
$$

Taking the absolute square:





$$
= \frac{1 + \cos(\alpha + \beta)}{4}
$$

Similarly, for $ P_{12'}(\alpha, \beta) $,



$$
= \frac{1 - \cos(\alpha + \beta)}{4}
$$

---

### Result
$$
P_{12}(\alpha, \beta) = P_{1'2'}(\alpha, \beta) = \frac{1 + \cos(\alpha + \beta)}{4}
$$

$$
P_{12'}(\alpha, \beta) = P_{1'2}(\alpha, \beta) = \frac{1 - \cos(\alpha + \beta)}{4}
$$
After performing their measurements, Alice and Bob publicly disclose **which bases were used**, but they do not reveal their measurement results. This allows them to compare correlations while preserving the secrecy of their key.

For the entangled state $ |\phi^+\rangle $, the probabilities of coincidence detection between Alice’s **detector 1** (or **detector 10**, which detects photons with orthogonal polarization) and Bob’s **detector 2** (or **detector 20**) are given by:

$$
P_{12}^{\alpha, \beta} = \frac{1}{4} (1 + \cos (\alpha + \beta))
$$

$$
P_{1,2'}^{\alpha, \beta} = \frac{1}{4} (1 - \cos (\alpha + \beta))
$$

These probabilities describe the expected quantum correlations between Alice and Bob’s measurements, which are used to test **Bell’s inequalities** and ensure the security of the quantum key distribution process.



```python
import numpy as np

def calculate_probabilities(alpha, beta):
    """Calculate both probabilities based on alpha and beta values."""
    P12 = (1 + np.cos(np.radians(alpha + beta))) / 4
    P1020 = (1 - np.cos(np.radians(alpha + beta))) / 4
    return P12, P1020

# Define possible values of alpha and beta
alpha_values = {1: 45, 2: 90, 3: 135, 4: 180}
beta_values = {1: 0, 2: 45, 3: 90, 4: 135}

# User input for alpha and beta
print("Choose α (1: 45°, 2: 90°, 3: 135°, 4: 180°)")
alpha_choice = int(input("Enter choice for α (1-4): "))

print("Choose β (1: 0°, 2: 45°, 3: 90°, 4: 135°)")
beta_choice = int(input("Enter choice for β (1-4): "))

# Validate input
if alpha_choice in alpha_values and beta_choice in beta_values:
    alpha = alpha_values[alpha_choice]
    beta = beta_values[beta_choice]

    P12, P12bar = calculate_probabilities(alpha, beta)

    print(f"\nResults for α = {alpha}° and β = {beta}°:")
    print(f"P_12(α, β)   = {P12:.4f}")
    print(f"P_12bar(α, β) = {P12bar:.4f}")
else:
    print("Invalid choice. Please enter a number between 1 and 4.")
```

With a typical coincidence rate of 5000 per second, the probability of detecting at least one photon pair within a 1 ms window is 99%. However, there's also a 96% chance of detecting multiple pairs in the same window. Here is how the calculation goes:




- **Coincidence Rate** = 5000 coincidences per second  
  → Alice and Bob detect 5000 photon pairs per second that pass the coincidence test.

- **Collection Time Window** = 1 ms (0.001 s)  
  → The experiment collects photon pairs over a duration of 1 millisecond.

- **Probability of Detecting at Least One Pair in 1 ms** = 99%  
  → There is a 99% chance that at least one photon pair is detected in each 1 ms window.

#### Estimating the Probability

The probability of detecting at least one photon pair in a 1 ms time window follows the **Poisson distribution**, which describes independent events occurring in a fixed time period.

### **Poisson Distribution Formula**
$$
P(n; \lambda) = \frac{\lambda^n e^{-\lambda}}{n!}
$$
where:
- $ P(n; \lambda) $ is the probability of detecting \( n \) events in the time interval.
- $ \lambda $ is the expected number of events in that interval.
- $ n $ is the number of detected events.

### **Applying to This Case**
- Expected coincidences per 1 ms:
  $$
  \lambda = 5000 \times 0.001 = 5
  $$
  → On average, 5 photon pairs are expected in 1 ms.

- Probability of detecting **zero** photon pairs (\( n = 0 \)):

  $$
  P(0;5) = \frac{5^0 e^{-5}}{0!} = e^{-5} \approx 0.0067 \text{ (0.67%)}
  $$

  → There is only a **0.67% chance** of detecting no photon pairs in 1 ms.

- Probability of detecting **at least one** photon pair:

  $$
  P(n \geq 1) = 1 - P(0) = 1 - e^{-5} \approx 0.993 \approx 99\%
  $$

  → This confirms that in each 1 ms window, **the probability of detecting at least one photon pair is 99%**.



```python
import math

def poisson_probability(lmbda, k):
    """Compute the Poisson probability for k occurrences given mean rate lmbda."""
    return (math.exp(-lmbda) * (lmbda ** k)) / math.factorial(k)

def cumulative_poisson(lmbda, k_min):
    """Compute the cumulative probability of at least k_min photons."""
    return 1 - sum(poisson_probability(lmbda, k) for k in range(k_min))

# Given data
coincidence_rate = 5000  # per second
time_window = 1e-3  # 1 ms
lmbda = coincidence_rate * time_window  # Expected photon pairs in 1 ms

# Prompt user for type of probability calculation
print("Choose an option:")
print("1. Probability of at least one photon")
print("2. Probability of more than one photon")
print("3. Probability of exactly one photon")
print("4. Probability of exactly k photons (user-specified k)")

choice = int(input("Enter your choice (1-4): "))

if choice == 1:
    result = cumulative_poisson(lmbda, 1)
    print(f"Probability of at least one photon: {result:.2%}")

elif choice == 2:
    result = cumulative_poisson(lmbda, 2)
    print(f"Probability of more than one photon: {result:.2%}")

elif choice == 3:
    result = poisson_probability(lmbda, 1)
    print(f"Probability of exactly one photon: {result:.2%}")

elif choice == 4:
    k = int(input("Enter the number of photons (k): "))
    result = poisson_probability(lmbda, k)
    print(f"Probability of exactly {k} photons: {result:.2%}")

else:
    print("Invalid choice. Please enter a number between 1 and 4.")

```

### Preventing Eavesdropping by Filtering Extra Events

During each collection window, the **phase setting remains constant**, meaning multiple photon pairs could be detected. If multiple pairs were used, an **eavesdropper (Eve)** could potentially gain extra information by analyzing these additional events.

### **How This Issue Was Avoided**
To prevent this security risk, only the **first detected photon pair** in each time window was kept, while the rest were discarded. This ensures that only a **single measurement** per phase setting is used.

### **Why This Ensures Security**
- Alice and Bob's measurement systems are assumed to be **completely isolated**, meaning no external influence can alter their measurement settings.
- An eavesdropper cannot introduce additional photons to manipulate the system or extract information about **Alice's (α) and Bob's (β) phase settings**.
- Since only **one photon pair per measurement setting** is used, the system remains secure even without rapid switching of settings.

By enforcing this **one-pair-per-window** rule, the protocol ensures that quantum key distribution remains **resistant to eavesdropping**, preserving the integrity of the key exchange.

# **Data Distribution in Quantum Cryptography**
In quantum cryptographic protocols, the raw data obtained from measurements is divided into three main categories:

1. **Cryptographic Key Contribution (25%)**: Only **one-fourth (1/4)** of the total data is used to generate the actual cryptographic key. These data points correspond to measurement outcomes where Alice and Bob's basis choices match and can be used securely for key distribution.

2. **Bell Inequality Testing (50%)**: **Half (1/2)** of the data is used for verifying the violation of Bell's inequalities. This ensures that the system exhibits quantum correlations rather than classical behavior, confirming security against eavesdropping.

3. **Unused Data (25%)**: The remaining **one-fourth (1/4)** of the data is discarded. These may correspond to cases where measurements are incompatible or unnecessary for key generation.

The following pie chart illustrates this division:


```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_data_distribution(total_data=10000):
    """
    Simulates the division of data based on the statement:
    - 1/4 contributes to the raw cryptographic key
    - 1/2 is used for testing Bell's inequalities
    - 1/4 is not used
    """
    key_data = total_data // 4  # 25% of the data
    bell_test_data = total_data // 2  # 50% of the data
    unused_data = total_data - (key_data + bell_test_data)  # Remaining 25%

    return key_data, bell_test_data, unused_data

def plot_data_distribution(key_data, bell_test_data, unused_data):
    """Plots the data distribution using a pie chart."""
    labels = ['Cryptographic Key (1/4)', "Bell's Inequality Test (1/2)", 'Unused Data (1/4)']
    sizes = [key_data, bell_test_data, unused_data]
    colors = ['lightblue', 'lightcoral', 'lightgrey']

    plt.figure(figsize=(6,6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
    plt.title('Data Distribution in Quantum Cryptography')
    plt.show()

# Simulate and plot
total_data = 10000
key_data, bell_test_data, unused_data = simulate_data_distribution(total_data)
plot_data_distribution(key_data, bell_test_data, unused_data)
```

## Residual Information in Privacy Amplification

Using these theorems ( CHARLES H. BENNETT et al),

### Theorem 1: Equitable Functions and Information Separation

Let *N* be the length of the originally transmitted bit string and let *K < N* be the safety parameter used for error detection. Let $\pi : \{0,1\}^N \to \{0,1\}^N $ be a randomly chosen permutation.

Define two equitable functions:
$$
f : \{0,1\}^N \to \{0,1\}^K
g : \{0,1\}^N \to \{0,1\}^{N-K}
$$
These functions are given by:

$$
    f(x) = \pi(x) \mod 2^K, \quad g(x) = \pi(x) \div 2^K.
$$

Then, knowledge of $\pi$ (hence of $f$ and $g$) yields **no information on** $g(x)$, except that it is of length $N-K$.

### Theorem 2: Shannon Information Bound

**Theorem 2**



$$
    I(Z;YFG) \leq \log(1 + 2^{-S}) < 2^{-S}/\ln 2.
$$



### Theorem 3: Equitable Function Information Bound

**Theorem 3**

Let *N, R,* and *S* be as above. Let $ g: \{0,1\}^N \to \{0,1\}^R $ be any fixed equitable function. Then, the expected amount of information given on $ g(x) $ by $ f(x) $, $ f $, and $ g $ is less than $ 2^{-S}/\ln 2 $ bits.

### Parameter Definitions

- \(N\) : Length of the originally transmitted bit string.
- \(K\) : Safety parameter used for error detection.
- \(S\) : Any nonnegative integer smaller than $N - K$.
- \(R\) : The remaining length of the final string, defined as $R = N - K$.
- $\pi$ : A randomly chosen permutation mapping $\{0,1\}^N$ onto itself.
- \(F\) : A function mapping $\{0,1\}^N$ to $\{0,1\}^K$, defined as $f(x) = \pi(x) \mod 2^K$.
- \(G\) : A function mapping $\{0,1\}^N$ to $\{0,1\}^{N-K}$, defined as $g(x) = \pi(x) \div 2^K$.
- \(X\) : The original bit string of length $N$.
- \(Y\) : The dependent random variable defined by $Y = f(X)$.
- \(Z\) : The dependent random variable defined by $Z = g(X)$.
- \(I(Z; YFG)\) : The mutual information between $Z$ and the tuple $(Y,F,G)$, representing the residual information available to an eavesdropper.

## Bell Parameters

In particular, the Bell parameters measured in this experiment are as follows:


$$S = -E(\alpha_1, \beta_1) + E(\alpha_1, \beta_3) + E(\alpha_3, \beta_1) + E(\alpha_3, \beta_3),$$



$$ S' = E(\alpha_2, \beta_2) + E(\alpha_2, \beta_4) + E(\alpha_4, \beta_2) - E(\alpha_4, \beta_4),
$$

where the correlation function is defined as:

$$
E(\alpha, \beta) = \frac{R_{12}(\alpha, \beta) + R_{1'2'}(\alpha, \beta) - R_{12'}(\alpha, \beta) - R_{1'2}(\alpha, \beta)}{R_{12}(\alpha, \beta) + R_{1'2'}(\alpha, \beta) + R_{12'}(\alpha, \beta) + R_{1'2}(\alpha, \beta)},
$$

where the \(R\)'s are the various coincidence counts between Alice and Bob detectors.

1. *Local Realistic Theories (Classical Limit):*  
   - Predicted upper bound: \( |S| $\leq$ 2\) and \( |S'| $\leq$2 \).  
   - Theories assuming *local realism* cannot exceed this limit.
   
   
   ##  Derivation of Bell’s Inequality
   

 Bell argued that if a more complete physical description were to remain compatible with local realism, the measurement outcomes $A$ and $B$ would need to be determined by functions of the form:

$$
A(\mathbf{a}, \lambda) = \pm 1, \quad B(\mathbf{b}, \lambda) = \pm 1.
$$

Here, $A$ and $B$ are the measurement outcomes at each detector given detector settings $\mathbf{a}$ and $\mathbf{b}$, and $\lambda$ is shorthand for any other properties of the particles prepared at the source that would affect the measurement outcomes at $A$ and $B$. (In the realist picture, $\lambda$ corresponds to a set of hidden variables that determine the outcomes.)

### Bell’s Assumptions:

As Bell emphasized, the forms of $A(\mathbf{a}, \lambda)$ and $B(\mathbf{b}, \lambda)$  incorporate locality. In the standard quantum mechanical description, a locality condition is imposed such that:

- The detector at $\mathbf{a}$ only depends on its own detector setting $\mathbf{a}$ and is independent of the detector setting $\mathbf{b}$.
- The outcome at detector $B$ does not depend on $\mathbf{a}$ and the outcome at $A$ does not depend on $\mathbf{b}$.

For a particular value of $\lambda$, the expectation value is given by:

$$
E(\mathbf{a}, \mathbf{b}) = \int d\lambda \, \rho(\lambda) A(\mathbf{a}, \lambda) B(\mathbf{b}, \lambda).
$$

### Derivation of Bell-CHSH Parameter:

For a local-realist theory, the Bell-CHSH parameter $S$ is expressed as:

$$
S = E(\mathbf{a}, \mathbf{b}) + E(\mathbf{a'}, \mathbf{b}) - E(\mathbf{a}, \mathbf{b'}) - E(\mathbf{a'}, \mathbf{b'}).
$$

Expanding in terms of hidden variables:

$$
S = \int d\lambda \rho(\lambda) [A(\mathbf{a}, \lambda)B(\mathbf{b}, \lambda) + A(\mathbf{a'}, \lambda)B(\mathbf{b}, \lambda) -A(\mathbf{a}, \lambda)B(\mathbf{b'}, \lambda) +A(\mathbf{a'}, \lambda)B(\mathbf{b'}, \lambda)].
$$

### Simplification for Specific Cases:

Let us consider the expression inside the integral for a particular value of $\lambda$:

- Since $A(\mathbf{a}, \lambda)$ and $B(\mathbf{b}, \lambda)$ can only take values $\pm 1$, the quantity inside the brackets becomes:

$$
 \int d\lambda \rho(\lambda)\left[A(\mathbf{a}, \lambda)[B(\mathbf{b}, \lambda) - B(\mathbf{b'}, \lambda)] + A(\mathbf{a'}, \lambda)[B(\mathbf{b}, \lambda) + B(\mathbf{b'}, \lambda)] \right].$$
Suppose for a particular value of  $\lambda$ ,these functions happens to be $\ B(\mathbf{b}, \lambda)$=+1 and $\ B(\mathbf{b'}, \lambda)$= -1,then we have


$$[
\left[ B(b, \lambda_i) - B(b', \lambda_i) \right] = +2, \quad \left[ B(b, \lambda_i) + B(b', \lambda_i) \right] = 0
]$$


By using the fact that $p(\lambda)$ is normalised, and knowing that $\left|A(\mathbf{a}, \lambda)\right| = \left|B(\mathbf{b}, \lambda)\right| = 1$, this simplifies to:

$$
|S| \leq 2.
$$

### Quantum Mechanical Prediction:

Using quantum mechanics, one can derive that the maximum possible value of $S$ is:

$$
S = 2 \sqrt{2} \approx 2.828.
$$

### Proof

   - $A(\theta_1) = \pm 1$: Measurement outcome for Photon 1 at angle $\theta_1$.  
   - $B(\theta_2) = \pm 1$: Measurement outcome for Photon 2 at angle $\theta_2$.  

   $$
   E(\theta_1, \theta_2) = \langle A(\theta_1)B(\theta_2) \rangle
   $$
   - Computed using probabilities of joint outcomes:
   $$
   E = P(+,+|\theta_1,\theta_2) + P(-,-|\theta_1,\theta_2) - P(+,-|\theta_1,\theta_2) - P(-,+|\theta_1,\theta_2).
   $$

---

## *Quantum Mechanical Framework*

### *Entangled Photon State*
The two photons are prepared in the *Bell state*:
$$
|\Psi\rangle = \frac{1}{\sqrt{2}} \left( |H\rangle_1 |H\rangle_2 + |V\rangle_1 |V\rangle_2 \right),
$$
- $|H\rangle$: Horizontal polarization state.  
- $|V\rangle$: Vertical polarization state.  

### *Measurement Operators*
The electric field operator for polarization at angle $\theta$:  
$$
\hat{a}{1\theta} = \hat{a}{1H} \cos\theta + \hat{a}_{1V} \sin\theta \quad \text, \\
\hat{a}{2\theta} = \hat{a}{2H} \cos\theta + \hat{a}_{2V} \sin\theta \quad \text.
$$

---

## *Joint Detection Probability*
The probability of detecting both photons at angles $\theta_1$ and $\theta_2$:  
$$
P(\theta_1, \theta_2) = \langle \Psi | \hat{a}{1\theta_1}^\dagger \hat{a}{2\theta_2}^\dagger \hat{a}{2\theta_2} \hat{a}{1\theta_1} | \Psi \rangle.
$$

### *Simplification*
Substituting $|\Psi\rangle$ and the operators:  
$$
P(\theta_1, \theta_2) = \frac{1}{2} \cos^2(\theta_1 - \theta_2).
$$

---

## *Correlation Function*
Using $P(\theta_1, \theta_2)$, the correlation becomes:
$$
E(\theta_1, \theta_2) = \frac{1}{2} \cos^2 (\theta_1 - \theta_2) + \frac{1}{2} \cos^2 (\theta_1 - \theta_2) - \frac{1}{2} \sin^2 (\theta_1 - \theta_2) - \frac{1}{2} \sin^2 (\theta_1 - \theta_2) $$


### *Derivation Steps*
$$
\begin{aligned}
E &= \cos^2(\theta_1 - \theta_2) - \sin^2(\theta_1 - \theta_2) \\
&= \cos[2(\theta_1 - \theta_2)] \quad \text{(using } \cos^2 x - \sin^2 x = \cos 2x).
\end{aligned}
$$

---

## *Bell Parameter $S$*
The CHSH-Bell parameter is defined as:
$$
S = |E(\theta_1, \theta_2) - E(\theta_1, \theta_2')| + |E(\theta_1', \theta_2) + E(\theta_1', \theta_2')|.
$$

### *Optimal Angles*
Maximal violation occurs at:  
$$
\theta_1 = 0, \quad \theta_2 = \frac{3\pi}{8}, \quad \theta_1' = -\frac{\pi}{4}, \quad \theta_2' = \frac{\pi}{8}.
$$


Substitute angles into $E(\theta_i, \theta_j) = \cos[2(\theta_i - \theta_j)]$:  

Thus:  
$$
S  = 2\sqrt{2}.
$$

Since this value exceeds the classical bound of 2, it indicates a violation of Bell’s inequality and confirms that nature does not obey local realism, thereby supporting the phenomenon of quantum entanglement.

# Quantum Nondemolition (QND) Eavesdropping

Based on "Eavesdropping using quantum-nondemolition measurements" by Werner & Milburn (Phys. Rev. A 47, 639).

---

## 1. Objective

Investigate how an eavesdropper using a *QND measurement* on an entangled photon affects *quantum cryptographic security*.

---

## 2. Entangled Photon State

The two-photon entangled polarization state is:

$$
|\psi_0\rangle = \frac{1}{\sqrt{2}} \left( |x\rangle_1 |y\rangle_2 + |y\rangle_1 |x\rangle_2 \right)
$$

- Photon 1 → Channel 1 → Detector D1  
- Photon 2 → Channel 2 → Detector D2  
- $x$ and $y$: Orthogonal polarization modes  

---

## 3. Detection Probabilities

Each photon passes through a polarizer:

- $\theta_1$: Before D1  
- $\theta_2$: Before D2  

Joint detection probability:

$$
P(D_1, D_2 | \theta_1, \theta_2) = \frac{1}{2} |\sin(\theta_1 + \theta_2)|^2
$$

---

## 4. Bell Inequality (Clauser-Horne Form)

Bell test combination:

$$
S = P(\theta_1, \theta_2) - P(\theta_1, \theta_2') + P(\theta_1', \theta_2) + P(\theta_1', \theta_2') - P(\theta_1', -) - P(-,\theta_2') \leq 0
$$

Quantum mechanics predicts a violation under certain angle settings.

---

## 5. QND Measurement Setup

The eavesdropper attempts a *non-destructive polarization measurement* using:

- A *Kerr medium*  
- A *probe beam* (coherent state)  
- *Homodyne detection* of the probe  

---

The interaction Hamiltonian for the two modes is:

$$
H_I = \hbar \chi n_s n_p
$$

Where:

- $\hbar$: Reduced Planck constant  
- $\chi$: Nonlinear coupling constant  
- $n_s$: Photon number operator for the signal mode  
- $n_p$: Photon number operator for the probe mode
---

## 6. Kerr Interaction

The probe quantity measured quadrature amplitude :

$$
\hat{Y} = \frac{1}{2} (\hat{b}^\dagger + \hat{b})
$$

- $\hat{b}$: Probe annihilation operator
- $\hat{b}^\dagger$: Probe creation operator


The probe undergoes a *phase shift* if a photon is present.

SNR of measurement:

$$
\text{SNR} = \frac{|<c_0> - <c_1>|}{(\Delta c_0 ^2 + \Delta c_1 ^2)^{1/2}}
$$

- <𝑐0>−<𝑐1> is the average  separation of the two peaks
- $\Delta c_{0,1} ^2 $ are the variances of each of the two peakes

- when the probe is prepared in a coherent state or the signal has only one photon the SNR is maximized at   
$$
  \chi = \pi
$$
 with
$$
SNR_{max} = 2^{2/3}\alpha
$$
- $\alpha$ coherent probe amplitude
---

## 7. State After QND Interaction

Joint state becomes:

$$
|\Psi\rangle = \frac{1}{\sqrt{2}} \left( |x\rangle_1 |y\rangle_2 |\alpha e^{-i\phi} \rangle_p + |y\rangle_1 |x\rangle_2 |\alpha \rangle_p \right)
$$

$\phi$ is proportional to the nonlinearity and length of the Kerr medium.

Reduced (polarization) density matrix:

### (11): Mixed State After QND Measurement

After the QND interaction and tracing over the probe field, the polarization part of the state becomes a mixed state. The off-diagonal terms (interference) are modulated by the overlap between the shifted and unshifted probe states:

$$
\rho' = \frac{1}{2} \Big( |x\rangle_{11} \langle x| \otimes |y\rangle_{22} \langle y| + |y\rangle_{11} \langle y| \otimes |x\rangle_{22} \langle x| + |x\rangle_{12} \langle x| \otimes |y\rangle_{21} \langle y| \langle \alpha | \alpha e^{-i\phi} \rangle + |x\rangle_{21} \langle x| \otimes |y\rangle_{12} \langle y| \langle \alpha e^{-i\phi} | \alpha \rangle \Big)
$$

---

## 8. Equation (12): Decoherence Factor $\Gamma$

The decoherence factor $\Gamma$ quantifies the reduction of off-diagonal coherence due to the distinguishability of the probe states:

$$
\Gamma = \exp \left( -|\alpha - \alpha e^{-i\phi}|^2 \right)
$$

This term governs how much the interference is lost due to the QND measurement.

---

## 9. Equation (13): Simplified $\Gamma$ for $\phi = \pi$

When the Kerr-induced phase shift is $\phi = \pi$ (maximum distinguishability between probe states), the expression simplifies to:

$$
\Gamma = \exp \left( -4|\alpha|^2 \right)
$$

Thus, a larger probe amplitude $|\alpha|$ causes stronger decoherence (smaller $\Gamma$).

---

## 10. Equation (14): Modified Joint Detection Probability

The joint probability of detecting both photons at detectors $D_1$ and $D_2$, given polarizer angles $\theta_1$ and $\theta_2$, becomes:

$$
P(D_1, D_2 \mid \theta_1, \theta_2) = \frac{1}{2} \left( \sin^2(\theta_1 + \theta_2) + \frac{\Gamma^{1/2} - 1}{2} \sin(2\theta_1) \sin(2\theta_2) \right)
$$

The second term depends on $\Gamma$ and reflects how the QND measurement has disturbed the original quantum correlation.

---

## 11. Equation (15): Bell Inequality Violation Parameter $S$

The Bell parameter $S$ indicates how strongly the state violates Bell's inequality. After the QND measurement, it is reduced to:

$$
S = \frac{1}{2} \left[ \frac{1}{\sqrt{2}} \left( 1 + \Gamma^{1/2} \right) - 1 \right]
$$

As $\Gamma \to 0$, the violation disappears; as $\Gamma \to 1$, the maximum quantum violation is restored.

## 12. Key Insight

Even a QND measurement *disturbs the entangled state* by entangling it with the probe field. The resulting decoherence ($\Gamma < 1$) leads to:

- *Reduced entanglement*  
- *Weakened Bell violation*  
- *Detectable eavesdropping*  

---

## 13. Practical Limitations

- Real Kerr nonlinearities are weak  
- For $\phi = \pi$, interaction length ≈ $10^6$ km  
- Additional issues: losses, noise, phase drifts  

---

## Conclusion

> The more information the eavesdropper gains via QND, the more coherence is lost — making them easier to catch through Bell tests.

Quantum cryptography remains secure against such attacks due to *fundamental quantum limits* on measurement and information gain.


# Quantum Nondemolition (QND) Eavesdropping in Quantum Cryptography

## 1. Introduction
The second eavesdropping strategy examined was a **Quantum Nondemolition (QND) Measurement**.  
QND measurements of optical photon number and polarization are **not currently feasible**.  
This fact makes current quantum cryptographic implementations secure, even when weak optical pulses are used (with an average photon number per pulse less than 1) .  

**Objective:**  
- Ensure quantum cryptography is secure against **all possible eavesdropping strategies**.
- Test cryptographic systems under **various attack models**.

---

## 2. Effect of QND Measurement

### Theoretical Effect
QND measurements introduce a **random phase** between the eigenstates of the measurement.  
This phase randomness arises due to the entanglement of the eigenstates with the readout quantum system.

### Simulation
To **simulate the effect** of a QND measurement:
- Insert a birefringent element in Bob’s path.
- The birefringent element separates the extraordinary and ordinary components of the photon wave packet by more than the **coherence length** (~140 mm, determined by the interference filters before the detectors).
- This introduces a **completely random phase** between the polarization components, mimicking a QND measurement.

### Mathematical Description
The effect of the eavesdropper is to apply a **projective measurement**:

$$
|x\rangle \langle x| + e^{i\phi_j} |x'\rangle \langle x'|
$$

Where:
- $\phi_j$ is the **random phase**.

## 4. Optimal Bases and Error Rates

### Case 1: Optimal Bases (Same Plane)
- Optimal bases for eavesdropping lie **in the same plane** (on the Poincaré sphere) as the bases used by Alice and Bob.
- Error probability:  
$$
\text{Error probability per intercepted bit} = 25\%
$$
- Bell parameter:  
$$
|S| = \sqrt{2}
$$

### Case 2: Random Bases (Orthogonal Plane)
- If the eavesdropper does **not know the plane** of the measurement bases and uses random measurements in an orthogonal plane:
    - Error probability increases to:
    $$
    \text{Error probability per bit} = 32.5\%
    $$
    - Bell parameter drops to:
    $$
    |S| = \frac{1}{\sqrt{2}}
    $$

## 5. Security Strategy: Using Multiple Planes

### Magnifying the Eavesdropper’s Effect
To improve security, Alice and Bob can:
- Choose measurement bases from at least **two (preferably three) orthogonal planes**.
- This “magnifies” the effect of an eavesdropper, increasing the **bit error rate (BER)** beyond 25%, making detection easier.

### Six-State Protocol
- Quantitative investigations support this strategy, known as the **“six-state” protocol** .
- Using six-state bases enhances security by:
    - Increasing the error rate caused by eavesdropping.
    - Making the presence of an eavesdropper more detectable.


---
## Six-State QKD Protocol


### 1. Overview

The **six-state protocol** is an extension of the **BB84 quantum key distribution protocol**, using **three mutually unbiased bases (MUBs)** instead of two, making eavesdropping more detectable.

---

### 2. The Three Bases

The six quantum states come from three mutually unbiased bases:

**Z-basis (Computational)**:  
  $|0\rangle,\quad |1\rangle$

**X-basis (Diagonal)**:  
  $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle), \quad |-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$

**Y-basis (Circular)**:  
  $|+i\rangle = \frac{1}{\sqrt{2}}(|0\rangle + i|1\rangle), \quad |-i\rangle = \frac{1}{\sqrt{2}}(|0\rangle - i|1\rangle)$

These six states form the **six-state protocol**.

---

### 3. Protocol Steps

1. **Alice** randomly picks one of the three bases (X, Y, or Z) and sends one of the two possible states.
2. **Bob** also randomly selects a basis and measures the incoming qubit.
3. Alice and Bob publicly compare their chosen bases.
4. They keep only the bits where their bases match (this happens with probability $1/3$).
5. Perform **error estimation**, **reconciliation**, and **privacy amplification** to extract a secure key.

---

### 4. Key Features

**Increased security**: More difficult for an eavesdropper to guess the basis.
**Better error detection**: Due to using 3 MUBs.
**Drawback**: Lower raw key rate (only $1/3$ of the bits are kept, vs. $1/2$ in BB84).

---

### 5. Why More Secure?

Eavesdropping becomes easier to detect because:
In BB84, Eve has a $50\%$ chance of guessing the correct basis.
In the six-state protocol, Eve only has a $33\%$ chance.
The increased symmetry and use of all three MUBs helps detect **coherent and individual attacks** more effectively.

---

### Conclusion:  
The six-state QKD protocol offers improved security at the cost of slightly lower efficiency. It is a robust alternative to BB84, especially against more sophisticated eavesdropping strategies.

---

## 6. Eavesdropper Measuring a Fraction of Photons

### Strategy: Partial Interception
- An eavesdropper can measure only a **fraction** of the photons to reduce the induced error rate.
- This increases the measured Bell parameter \( S \), but reduces her knowledge of the cryptographic key.
- For instance:
    - If Eve measures **less than 58.6%** of the photons in the optimal basis:
        - \( S > 2 \)
        - \( $\text{BER}$ < 15\% \)
    - However, Eve’s knowledge of the key will be **less than Bob’s**, allowing Alice and Bob to use **privacy amplification** to generate a secret key .

---

## 7. Conclusion
- QND eavesdropping disturbs the entangled state by introducing **decoherence**, reducing the Bell violation.
- **Multiple basis measurements** and **six-state protocols** magnify the effects of eavesdropping, improving detection.
- **Partial interception** reduces Eve’s information while maintaining security through privacy amplification.

**Quantum cryptography remains secure against QND and other physically possible eavesdropping strategies.**

This paper explores **quantum cryptography**—a way to securely share secret information using the laws of **quantum physics**, which are **impossible to hack** using conventional or even quantum computers. The authors specifically implemented the **Ekert Protocol**, a **quantum key distribution (QKD)** method that uses **entangled photon pairs** to share a secret key between two people (**Alice and Bob**) and detect any eavesdropper (**Eve**) trying to spy on them.

## Ekert Protocol  
Unlike older systems that used weak laser pulses, this protocol uses **entangled photons**, which are **deeply connected in a quantum way**. If one is measured, the other's result is **instantly affected**—no matter the distance. This connection **can't be faked or copied** without being detected. The Ekert Protocol checks for spying by testing **Bell's inequality**, a quantum rule that must be **violated if the photons are truly entangled**. If an eavesdropper interferes, this violation **decreases**, alerting **Alice and Bob**.

## Experiment Setup  
- **Photon Generation**: Entangled photons were created using **spontaneous parametric down-conversion** in special crystals.  
- **Measurement Setup**:  
  - Alice and Bob each got one photon and **randomly measured its polarization** using different settings.  
- **Key Creation**: When their settings **matched**, they used those results to form a **secret key**.  
- **Security Check**: When settings **didn't match**, they used the results to **test Bell's inequality**—this helps check for **eavesdropping**.  

They only kept the **first detection event** in every time window to avoid counting **multiple pairs**, which could leak **extra information to Eve**.  

The authors tested two possible strategies an **eavesdropper (Eve)** might use:  

- **Intercept-and-resend**: Eve **measures a photon** and sends a **replacement to Bob**. This introduces **errors**.  
- **Quantum nondemolition (QND) attack**: A theoretical method where Eve **measures without destroying** the photon. Since this isn't possible yet, they **simulated** it using **special crystals**.  

They found that in both strategies, **Eve’s presence could be detected**, because:  

1. The **error rate increases**, and  
2. The **Bell inequality violation decreases**.  

The experiment also showed that if Eve guesses the **wrong measurement basis** (wrong "plane"), her **error rate increases** from **25% to 32.5%**, making her even **easier to detect**.  

## Security Test  
- We generated **over 24,000 raw key bits** in a few **10-minute runs**.  
- After **error correction** and **privacy amplification**, they got **15,444 secure bits**.  
- The observed **bit error rate (BER)** was **low: about 3%**.  
- The **Bell parameter |S| was ~2.66**, clearly **violating the classical limit (2)**, meaning **entanglement remained strong and Eve wasn't interfering**.  

## Future Potential  
Even though **Alice and Bob were on the same optical table** in this test, the setup can be **scaled to hundreds of meters or even to satellites**, enabling **global quantum communication**.  

## Conclusion  
This was **one of the first experiments** to use **entangled photons** in a **full quantum cryptography protocol**.  

It successfully demonstrated that **eavesdropping can be detected**.   

This setup **lays the groundwork** for **real-world secure quantum communication over long distances**.
