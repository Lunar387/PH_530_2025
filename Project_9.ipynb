## What does this paper look at?

The aim of the paper is to advance the performance of **heralded single-photon sources (HSPS)** by benchmarking and designing a HSPS with maximum purity and at the same time having a high brightness.
It first derives a theoretical limit for how pure a HSPS can be. Purity is given by the **(heralded) second-order autocorrelation function at zero delay** $g^{(2)}_{h}(0)$. In order to be able to measure $g^{(2)}_{h}(0)$, they transform it from being in terms of $\mu$ (mean number of photon pairs produced) into being in terms of the **Coincidence-to-Accidental Ratio ($CAR$)**, because $\mu$ is not directly measurable, while $CAR$ is.
They then go on and design an experimental setup for a HSPS which reaches this theoretical limit of purity at a high brightness.

## What is a Heralded Single-Photon Source (HSPS)?

A heralded single photon source uses a nonlinear medium to split an input photon (pump photon) into an entangled photon-pair. These two photons are then immediately spatially separated into different paths via optical filtering components. This enables to "herald" (herald here means "to indicate") the existence of one photon in the pair, by detecting the other.

## What is single-photon purity?

Single-photon purity refers to how close the output state is to a single-photon number state.
This means, in a perfectly pure single-photon source, you always get one output photon and never two or more.
For a general single-photon source, purity is quantified by the \textbf{second-order autocorrelation function at zero delay} $g^{(2)}(0)$, which was derived and defined in class by:

\begin{equation}
    g^{(2)}(\tau) = \frac{\langle I(t) I(t+\tau) \rangle}{\langle I(t) \rangle^{2}} \quad\implies\quad g^{(2)}(0) = \frac{\langle I(t)^{2} \rangle}{\langle I(t) \rangle^{2}}
\end{equation}

In other words, $g^{(2)}(0)$ is the likelihood of measuring two photons at the same time, and therefore $g^{(2)}(0) = 0$ means there is zero probability of having two photons in the output channel, hence a perfectly pure single photon source.
Now for a heralded single-photon source, this becomes the \textbf{heralded second-order autocorrelation function at zero delay} $g^{(2)}_{h}(0)$.

Given you have detected a photon in the "Heralding" arm, $g^{(2)}_{h}(0)$ intuitively gives you the answer to the question: "Given that I know a photon should now be in the Heralded arm, how often do I accidentally see two photons instead of exactly one?"
In other words, after a successful detection in the "Heralding" arm, $g^{(2)}_{h}(0)$ tells you the probability of having exactly one photon versus accidentally having two or more.
 $g^{(2)}_{h}(0) = 0$ here means that there is zero probability of having two photons in the "Heralded" channel. Whether this limit can be achieved will be derived later.

To measure $g^{(2)}_{h}(0)$, a Hanbury-Brown-Twiss (HBT) setup can be introduced in the "Heralded" channel:

Now, $g^{(2)}_{h}(0)$ can be written in terms of creation ($\hat{a}^{\dagger}$) and annihilation ($\hat{a}$) operators as:

\begin{equation}
    g^{(2)}_{h}(0) = \frac{\langle \hat{a}_{1}^{\dagger}\hat{a}_{2}^{\dagger}\hat{a}_{3}^{\dagger}\hat{a}_{3}\hat{a}_{2}\hat{a}_{1} \rangle}{\langle \hat{a}_{1}^{\dagger}\hat{a}_{2}^{\dagger}\hat{a}_{2}\hat{a}_{1}  \rangle\langle \hat{a}_{1}^{\dagger}\hat{a}_{3}^{\dagger}\hat{a}_{3}\hat{a}_{1}  \rangle} \langle \hat{a}_{1}^{\dagger}\hat{a}_{1}  \rangle
\end{equation}

Where:

- $\langle\hat{a}_{1}^{\dagger}\hat{a}_{1}  \rangle$ is the total number of clicks in detector 1.
- $\langle \hat{a}_{1}^{\dagger}\hat{a}_{2}^{\dagger}\hat{a}_{2}\hat{a}_{1}  \rangle$ is the total number of clicks in detector 2, given detector 1 clicks.
- $\langle \hat{a}_{1}^{\dagger}\hat{a}_{3}^{\dagger}\hat{a}_{3}\hat{a}_{1}  \rangle$ is the total number of clicks in detector 3, given detector 1 clicks.
- $\langle \hat{a}_{1}^{\dagger}\hat{a}_{2}^{\dagger}\hat{a}_{3}^{\dagger}\hat{a}_{3}\hat{a}_{2}\hat{a}_{1} \rangle$ is the total number of times detectors 1 \& 2 click simultaneously given detector 1 clicks.


Conceptionally, this means: "given detector 1 fired (herald), what is the normalized probability that detectors 2 and 3 both register a photon at the same time (meaning two photons were in the heralded arm)?". If always only one photon is present in the heralded arm per herald, detectors 2 and 3 will never click together, making $g^{(2)}_{h}(0) \to 0$.

## What is the Coincidence-to-Accidental Ratio (CAR)?

The Coincidence-to-Accidental Ratio (CAR) is a metric used to characterize the quality and performance of photon-pair sources and is defined by:

\begin{equation}
    CAR = \frac{\text{coincidence counts}}{\text{accidental counts}}
\end{equation}

In our case, a coincidence count means both detectors (Heralded and Heralding) click simultaneously or within the same time window, indicating a pair was generated. To measure the coincidence counts, you simply count how many times both detectors click together within this window.

An accidental count is when the detectors click within the timing window but not due to the same pair. You can measure the accidental counts by shifting one detector's timing window by a delay much greater than the photon correlation time. Because no genuine photon pairs survive such large delays (true photon pairs are simultaneous), all coincidences you measure now are purely accidental.

Therefore, CAR essentially measures the signal-to-noise ratio of the pair source: a high CAR means that almost all clicks are true pair events with very few accidental coincidences.

### Various Types of Single-Photon Sources:

1. **On-Demand Single-Photon Sources (Quantum Emitters):**
<p>
 These use single quantum systems that emit one photon at a time when excited. Examples include single atoms or ions, quantum dots, nitrogen-vacancy (NV) centers in diamond, and other single emitters in solid-state devices​. These sources can, in principle, produce a photon “on demand” (at a chosen time) by external excitation. They tend to exhibit sub-Poissonian photon statistics (antibunching) and can achieve low \( g^2(0) \) values. However, they often require challenging conditions (like cryogenic temperatures or precise fabrication), and their repetition rates might be limited.
</p>


2. **Heralded Single-Photon Sources (Parametric Sources):**
<p>
   These are probabilistic sources based on nonlinear optical processes, where photon pairs are created spontaneously and one photon signals (heralds) the presence of the other. The two most common processes are spontaneous parametric down-conversion (SPDC) (a second-order nonlinear process, \( \chi^2 \)) and spontaneous four-wave mixing (SFWM) (a third-order nonlinear process, \( \chi^3 \)).
</p>
<p>
   In SPDC, a pump photon in a nonlinear crystal splits into a pair of lower-energy photons (signal and idler); in SFWM, two photons from a pump are annihilated to create two new photons (signal and idler) in a nonlinear medium. These sources are heralded in the sense that detection of one photon of the pair implies the existence of its partner.
</p>
<p>
   Heralded sources are not on-demand (the pair creation is probabilistic), but when a “herald” click occurs, one knows a single photon is present in the other channel. They have advantages of operating at room temperature and being naturally compatible with photonic integration (e.g., waveguides).
</p>
<p>
   Many implementations use integrated photonic waveguides or fibers to generate photon pairs, allowing high repetition rates and easy coupling to fiber networks.
</p>


3. **Attenuated Laser (Faint Laser Source):**
<p>
 This is a simple but non-ideal method where a laser beam is heavily attenuated so that the average photon number per pulse is much less than 1 (e.g., 0.1 or so). This yields a high probability of either 0 or 1 photon per pulse and a very low probability of 2 or more. However, an attenuated laser does not produce true single-photon states – it produces a weak coherent state with Poissonian statistics.
   </p>
<p>
   For example, an average of 0.1 photons per pulse gives about 9% chance of one photon and 1% chance of two or more​. Such a source has \( g^2(0) = 1 \) indicating no antibunching (it’s essentially a scaled-down classical source). While useful in some tests, this doesn’t meet the strict single-photon criteria needed for quantum applications that demand near-zero multi-photon probability.
</p>

### Spontanous Four Wave Mixing (SFWM):

Four wave mixing is a photon-photon scattering process, during which two photons from a relatively high-intensity beam, called pump, scatter through the third order nonlinearity of the materialto generate two daughter photons, called signal and idler photons, respectively. The frequencies of the daughter photons are symmetrically displaced from the pump frequency, satisfying the energy conservation relation:


\begin{equation}
\omega_s + \omega_i = 2\omega_p
\end{equation}


where p,s,i denotes the pump/signal/idler frequency, respectively.
The daughter photons also form a time energy entangled state, in the sense that the two particlew wave function can not be factorised into products of single particle wave functions:

\begin{equation}
\Psi(\omega_s, \omega_i) \neq X(\omega_s) \cdot Y(\omega_i)
\end{equation}

The only drawback of the SFWM is the process of spontaneous Raman scattering (SRS), which inevitably occrs in anythird order nonlinear medium and generates uncorrelated photons into the detection bands, leading to a degradation in the quality of the generated entanglement. SFWM is termed “spontaneous” because the generation of the new photons is initiated by quantum vacuum fluctuations rather than a seeded input – it happens probabilistically when the pump is present. When a pulse goes through the waveguide, there is a small probability that an SFWM event occurs, yielding one signal-idler photon pair. The probability per pulse is tuned by the pump power. The spontaneous nature means multiple pairs can sometimes occur in one pulse (especially if the pump power is high), which is a source of impurity as we’ll discuss, but with careful design (and low enough per-pulse probability) one pair per pulse dominates.

### Mathematical Framework:

Just as in reference $[4]$ https://journals.aps.org/pra/abstract/10.1103/PhysRevA.72.033801 of our paper, we take the standard approach of modern quantum optics, i.e., finding out the interaction Hamiltonian and calculating the evolution of the state vector using the Schrödinger picture. We accomplish the first task by seeking
connections with the well-known classical FWM theory in optical fibers . The coupled classical-wave equations for
the pump, signal, and idler fields are:

\begin{equation}
  \begin{aligned}
  \frac{\partial A_p}{\partial z} &= i \gamma |A_p|^2 A_p, \\
  \frac{\partial A_s}{\partial z} &= i \gamma \left(2 |A_p|^2 A_s + A_p^2 A_i^* e^{-i \Delta k z} \right), \\
  \frac{\partial A_i}{\partial z} &= i \gamma \left(2 |A_p|^2 A_i + A_p^2 A_s^* e^{-i \Delta k z} \right),
  \end{aligned}
\end{equation}

This set of equations was obtained from and derived in the book: Govind P. Agrawal, Nonlinear Fiber Optics, 3rd ed. (Academic Press, New York, 2001).

They then use the quantum correspondence principle to replace classical fields $A_j$ with the quantum field operators $\hat{E}_{j}^{(+)}$ in order to get the quantum equations of motion:

\begin{equation}
  \begin{aligned}
  \frac{\partial \hat{E}_p^{(+)}}{\partial z} &= i \eta \hat{E}_p^{(-)} \hat{E}_p^{(+)} \hat{E}_p^{(+)}, \\
  \frac{\partial \hat{E}_s^{(+)}}{\partial z} &= i \eta \left( 2 \hat{E}_p^{(-)} \hat{E}_p^{(+)} \hat{E}_s^{(+)} + \hat{E}_i^{(-)} \hat{E}_p^{(+)} \hat{E}_p^{(+)} \right), \\
  \frac{\partial \hat{E}_i^{(+)}}{\partial z} &= i \eta \left( 2 \hat{E}_p^{(-)} \hat{E}_p^{(+)} \hat{E}_i^{(+)} + \hat{E}_s^{(-)} \hat{E}_p^{(+)} \hat{E}_p^{(+)} \right),
  \end{aligned}
\end{equation}

where $\hat{E}_{j}^{(+)} = \sqrt{\frac{\hbar \omega_{j}}{2\epsilon_{0}V_{Q}}} \hat{a}_{j}$, $\hat{E}_{j}^{(-)} = \sqrt{\frac{\hbar \omega_{j}}{2\epsilon_{0}V_{Q}}} \hat{a}_{j}^{\dagger}$.


The interaction hamiltonian is then constructed in such a way, so that it leads to the equations above via the Heisenberg equation of motion for the field operators: $i\hbar(\frac{\partial \hat{E}}{\partial t}) = [\hat{E},H_{I}]$. This results in the following form of the interaction hamiltonian:

\begin{equation}
  \begin{aligned}
  \hat{H}_I &= \beta\epsilon_0 \chi^{(3)} \int_{V} dV \left(
  \hat{E}_p^{(-)} \hat{E}_p^{(-)} \hat{E}_p^{(+)} \hat{E}_p^{(+)}
  + 2 \hat{E}_s^{(-)} \hat{E}_i^{(-)} \hat{E}_p^{(+)} \hat{E}_p^{(+)} \right. \\
  &\left. + 2 \hat{E}_p^{(-)} \hat{E}_p^{(-)} \hat{E}_s^{(+)} \hat{E}_i^{(+)}
  + 4 \hat{E}_p^{(-)} \hat{E}_p^{(+)} \hat{E}_s^{(-)} \hat{E}_s^{(+)} \right. \\
  &\left. + 4 \hat{E}_p^{(-)} \hat{E}_p^{(+)} \hat{E}_i^{(-)} \hat{E}_i^{(+)}
  \right)
  \end{aligned}
\end{equation}

## Derivation of SFWM output state - (Lars)

The two-photon output state of the SFWM is calculated by using first-order perturbation theory on the above interaction hamiltonian:

\begin{equation}
|\Psi\rangle = |0\rangle + \frac{1}{i\hbar} \int_{-\infty}^{\infty} \hat{H}_{I}(t) \, dt |0\rangle
\end{equation}

From this expression we can see, that only the following $FPS$ terms in the interaction hamiltonian survive, because all other terms vanish when acting on the vacuum state $|0\rangle$. Hence the hamiltonian becomes:

\begin{equation}
\hat{H}_{\text{FPS}} = \alpha\epsilon_0 \chi^{(3)} \int_{V} dV \left(
\hat{E}_s^{(-)} \hat{E}_i^{(-)} \hat{E}_p^{(+)} \hat{E}_p^{(+)} + \text{H.c.}
\right)
\end{equation}

Therefore the output state vector is given by:

\begin{equation}
|\Psi\rangle = |0\rangle + \frac{1}{i\hbar} \int_{-\infty}^{\infty} \hat{H}_{\text{FPS}}(t) \, dt |0\rangle
\end{equation}

When inserting the explicit expressions for the various electric field operators appearing in the hamiltonian, we arrive in the final expression of the output state:

\begin{equation}
|\Psi\rangle = |0\rangle + \sum_{k_s, k_i} F(k_s, k_i) \hat{a}_{k_s}^\dagger \hat{a}_{k_i}^\dagger |0\rangle
\end{equation}

\begin{equation}
F(k_s, k_i) = g \int_{-L}^{0} dz \, \frac{1}{\sqrt{1 - i k'' \sigma_p^2 z}}
\exp \left[ -\frac{i k'' \sigma_p^2 z}{4} (\omega_s - \omega_i + \delta)^2
- \frac{(\omega_s + \omega_i)^2}{4 \sigma_p^2} - i \gamma P_p z \right]
\end{equation}

\begin{equation}
g = \frac{\sqrt{2} \chi^{(3)} P_p}{i \epsilon_0 V_Q n^3 \omega_p \sigma_p}
\end{equation}

This is the first-order perturbed quantum state generated by four-wave mixing (SFWM). It is a superposition of the vacuum state and a two-photon state. The two-photon state contains one signal photon at $k_{s}$ and on idler photon at $k_{i}$, created by the operators $\hat{a}_{k_s}^\dagger \hat{a}_{k_i}^\dagger$.

The function $F(k_s, k_i)$ is the joint spectral amplitude (JSA) — it captures the probability amplitude for generating a signal-idler photon pair with specific momenta or frequencies.


## Derivation of Heralded Second-Order Autocorrelation Function at zero delay in Hanbury-Brown-Twiss setup - (Deeksha)


For a heralded single photon source (HSPS), the heralded second-order correlation function is:

\begin{equation}
    g^{(2)}_{h}(0) = \frac{\langle \hat{a}_{1}^{\dagger}\hat{a}_{2}^{\dagger}\hat{a}_{3}^{\dagger}\hat{a}_{3}\hat{a}_{2}\hat{a}_{1} \rangle}{\langle \hat{a}_{1}^{\dagger}\hat{a}_{2}^{\dagger}\hat{a}_{2}\hat{a}_{1}  \rangle\langle \hat{a}_{1}^{\dagger}\hat{a}_{3}^{\dagger}\hat{a}_{3}\hat{a}_{1}  \rangle} \langle \hat{a}_{1}^{\dagger}\hat{a}_{1}  \rangle
\end{equation}

where the subscripts denote the detectors at the heralding $(1)$ or heralded (2, 3) side. As measurements between the heralding side and the heralded are independent, operators $\hat{a}_1$ and, $\hat{a}_2$,  $\hat{a}_3$ are mutually commutative. Therefore, $g^{(2)}_h(0)$ can be rewritten as:

$$
g^{(2)}_h(0) = \frac{\langle \hat{n}_s \hat{n}_i (\hat{n}_i - 1) \rangle}{\langle \hat{n}_s \hat{n}_i \rangle^2} \langle \hat{n}_s \rangle
$$

where $\hat{n}$ represents the photon number operator, the angle brackets represent the expected value, and the subscripts $s$ and  $i$ refer to the signal photons used for heralding and the idler photons being heralded, respectively.

Due to the simultaneity of photon pairs generated through SFWM, the photon pair state can be written as:

$$
|\Phi \rangle = \sum_{n=0}^{\infty} P(n) |n, n \rangle_{s,i}
$$

where  $P(n)$ is the probability of generating $n$ photon pairs per pulse.

Then, we can calculate the second-order correlation function of the heralded single-photon source $g^{(2)}_h(0)$ under any distribution by taking Equation (12) into Equation (11), and we obtain:

$$
g^{(2)}_h(0) = \frac{\langle \Phi | \hat{n}_s \hat{n}_i (\hat{n}_i - 1) | \Phi \rangle}
{\langle \Phi | \hat{n}_s \hat{n}_i | \Phi \rangle^2 \langle \Phi | \hat{n}_s | \Phi \rangle}
= \frac{\sum n^2 (n-1) P(n)}{\left( \sum n^2 P(n) \right)^2} \sum n P(n)
$$

Therefore, for SFWM sources with Poissonian distribution, \( P(n) = \frac{e^{-\mu} \mu^n}{n!} \), its \( g^{(2)}_h(0) \) can be expressed as:

$$
g^{(2)}_h(0) = \frac{\mu^2(\mu + 2)}{[\mu(\mu + 1)]^2} = 1 - \frac{1}{(\mu + 1)^2}
$$

\begin{equation}
    g^{(2)}_{h}(0) = \frac{\langle \hat{a}_{1}^{\dagger}\hat{a}_{2}^{\dagger}\hat{a}_{3}^{\dagger}\hat{a}_{3}\hat{a}_{2}\hat{a}_{1} \rangle}{\langle \hat{a}_{1}^{\dagger}\hat{a}_{2}^{\dagger}\hat{a}_{2}\hat{a}_{1}  \rangle\langle \hat{a}_{1}^{\dagger}\hat{a}_{3}^{\dagger}\hat{a}_{3}\hat{a}_{1}  \rangle} \langle \hat{a}_{1}^{\dagger}\hat{a}_{1}  \rangle
\end{equation}


## Derivation of theoretical limit of Heralded Second-Order Autocorrelation Function at zero delay - (Deeksha)

When the source generates distinguishable correlated pairs,then the photon pairs follow Poissonian distribution $P(\mu, x)$.

$$P(\mu, x) = \frac{e^{-\mu} \mu^x}{x!}$$

Let $\alpha$ be the collection efficiency of detectors, then the probability of detection of $x$ photon pairs is given by
$$
P_{\text{detection}} = 1 - (1 - \alpha)^x
$$

Accidental Coincidence Rate:
$$
R_\text{acc} = (\sum_{x=0}^{\infty} P(\mu, x) \left[ 1 - (1 - \alpha)^x \right])^2
$$

Coincidence Rate of Correlated Photons:
$$
R_\text{true} = \sum_{x=0}^{\infty} P(\mu, x) \left[ 1 - (1 - \alpha)^x \right]^2
$$

COincidence to Accidence Photon Counts Ratio:
$$
CAR = \frac{R_\text{true}}{R_\text{acc}}
$$

When the collection efficiency $\alpha << 1$ the  probability of detection of $x$ photons is approximated to $\alpha x$.

Evaluating the coincidence and accidental rates under the approximation, we get

$$
R_\text{acc}  = (\sum_{x=0}^{\infty} P(\mu, x) x \alpha)^2 = (\mu \alpha)^2
$$

$$
R_\text{true} =  \sum_{x=0}^{\infty} P(\mu, x) x^2 \alpha ^2 = \alpha^2 \mu (\mu +1)
$$

Thus, CAR for distinguishable correlated photon pairs is given by
$$
 CAR = 1 + \frac{1}{\mu}
$$

Now equating the two equations we get relation between $g^{(2)}_{h}(0)$ and $CAR$ both of which can be evaluated experimenatlly.

\begin{equation}
    g^{(2)}_{h}(0) = \frac{2 CAR - 1}{CAR^{2}}
\end{equation}


```python
import numpy as np
import matplotlib.pyplot as plt

CAR = np.logspace(1, 4, 500)
g2_h = (2 * CAR - 1) / (CAR ** 2)

fig, ax = plt.subplots(figsize=(6, 5))
ax.plot(CAR, g2_h, color='red', linewidth=2, label='Coherent pump limit')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(10, 1e4)
ax.set_ylim(1e-3, 1)
ax.set_xlabel("CAR", fontsize=12)
ax.set_ylabel(r"$g^{(2)}_h(0)$", fontsize=12)
ax.annotate("Coherent\npump limit", xy=(200, 0.008), xytext=(40, 0.01),
            arrowprops=dict(facecolor='red', arrowstyle="->"),
            fontsize=10, color='black', ha='center')
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```

<h2>A. SL, ML and Optical Injection Locking</h2>

<p>The <strong>ML</strong> operates in continuous-wave (CW) mode, emitting a stable optical signal at 1550.52 nm. It provides a clean frequency reference to control and stabilize the <strong>Slave Laser (SL)</strong> through optical injection locking.</p>

<p>The SL is a semiconductor distributed feedback (DFB) laser that generates 2.5 GHz, 70ps optical pulses at 1550.52 nm. It operates using a technique called <strong>gain switching</strong>, where the laser is driven with a high-frequency electrical modulation to force it into short-pulse generation. If the SL were left to operate freely, it could exhibit significant timing jitter and frequency instability.</p>

<h3>1. How the SL Operates?</h3>
<ul>
  <li>The SL is initially below its lasing threshold, meaning it does not emit light.</li>
  <li>A high-frequency electrical current is applied to the SL.</li>
  <li>This modulation quickly pushes the laser above and below threshold, causing rapid bursts of optical emission.</li>
  <li>The result is a train of short optical pulses instead of a continuous beam.</li>
</ul>

<h3>B. Optical Injection Locking</h3>
<p>This technique, where a high-power slave laser is stabilized using a low-noise master laser, is called <strong>Optical Injection Locking</strong>. This ensures that the slave laser operates at exactly the frequency of the injected master laser, reducing noise and improving coherence.</p>

<h2>B. Amplification of Gain-Switched Pulses using an EDFA</h2>

<p>The optical pulse generated from SL has relatively low power, so they are amplified using an <strong>Erbium-Doped Fibre Amplifier (EDFA)</strong>.</p>

<ol>
  <li>EDFA is a fiber-based optical amplifier that boosts the signal power without converting it to an electrical signal.</li>
  <li>It consists of erbium-doped optical fiber pumped by a 980 nm or 1480 nm laser diode. When the 1550 nm pulses pass through the erbium-doped fiber, the erbium ions amplify them via stimulated emission.</li>
</ol>

<p>The EDFA is used because it provides:</p>
<ul>
  <li>Low noise amplification (important for quantum optics).</li>
  <li>High gain (~20–30 dB), sufficient for SFWM in the waveguide.</li>
  <li>Broadband operation, making it suitable for different wavelength channels.</li>
</ul>

<h3><strong>Problem: EDFA Also Introduces Noise</strong></h3>
<p>EDFA introduces amplified spontaneous emission (ASE) noise, which consists of:</p>
<ul>
  <li>Unwanted photons across a broad spectrum.</li>
  <li>Background noise that degrades single-photon purity.</li>
</ul>

<p>To remove this noise, two optical filters are used.</p>


<h2>C. Optical Filtering: Fiber Bragg Grating (FBG) and Tunable Bandpass Filter (BPF)</h2>

<p>Optical filtering is a crucial step in ensuring that only the desired wavelengths are utilized while eliminating unwanted broadband noise. In this setup, two filtering components are used: a <strong>Fiber Bragg Grating (FBG)</strong> and a <strong>Tunable Bandpass Filter (BPF)</strong>. Together, they provide highly selective wavelength filtering, ensuring that only the required pump wavelength reaches the system.</p>

<h3>1. Fiber Bragg Grating (FBG)</h3>
<p>A Fiber Bragg Grating (FBG) is an optical component consisting of a periodic variation in the refractive index along the length of an optical fiber. This periodic structure acts as a wavelength-selective reflector, allowing only a narrow range of wavelengths to be reflected while transmitting others.</p>

<h4>Role of FBG in the Setup</h4>
<ul>
  <li>The FBG used here has a Full-Width at Half-Maximum (FWHM) of 0.4 nm, meaning it allows only a very narrow range of wavelengths to pass through.</li>
  <li>It effectively filters out unwanted broadband Amplified Spontaneous Emission (ASE) noise that is generated along with the pump pulses. The desired pump wavelength of 1550.52 nm is selectively retained while all other unwanted wavelengths are blocked. This step is essential for ensuring that only a clean and stable pump wavelength is available for Spontaneous Four-Wave Mixing (SFWM).</li>
</ul>

<h3>2. Tunable Bandpass Filter (BPF)</h3>
<p>Following the FBG, a Tunable Bandpass Filter (BPF) is used to further refine the spectral purity of the pump light.</p>

<h4>Role of BPF in the Setup</h4>
<ul>
  <li>The BPF has a narrow passband of 0.3 nm, meaning it allows an even more precise selection of the pump wavelength while rejecting any remaining noise. It provides additional filtering to remove any residual ASE noise that might have passed through the FBG.</li>
  <li>Since it is tunable, it allows for fine adjustments in the filtered wavelength, offering flexibility in selecting the exact pump frequency required for the experiment.</li>
  <li>This ensures that the light entering the waveguide is highly monochromatic and free from unwanted spectral components.</li>
</ul>

<h3>3. Filtering Performance</h3>
<p>The combined effect of the FBG and BPF results in an exceptionally high level of optical filtering:</p>
<ul>
  <li>The system achieves an optical isolation of ≥106 dB, meaning that noise photons are suppressed by a factor of 10¹⁰. This significant reduction in noise ensures that the pump laser remains highly stable and spectrally pure.</li>
  <li>By eliminating unwanted spectral components, the setup enhances the efficiency of Spontaneous Four-Wave Mixing (SFWM) and improves overall system performance.</li>
</ul>


<h2>D. Active Pump Power Stabilization</h2>

<p>Even after optical filtering, stabilizing the pump power is essential to maintain the reliability of the Spontaneous Four-Wave Mixing (SFWM) process. Any fluctuation in pump power can have a direct impact on the system’s performance, affecting both the rate of photon-pair generation and the consistency of single-photon detection. These fluctuations can lead to variations in the photon emission rate, reducing coherence and affecting experimental outcomes.</p>

<h3>1. Key Components of Power Stabilization</h3>
<p>To address this issue, an active power stabilization system is implemented using three key components:</p>

<h3>(a) Optical Power Meter (PM) with a 10:90 Tap Beam Splitter</h3>
<p>The first step in power stabilization involves real-time monitoring of the pump power using an optical power meter (PM).</p>

<h4>Role of the PM and Beam Splitter</h4>
<ul>
  <li>A 10:90 tap beam splitter is introduced after the optical filtering stage. 10% of the pump power is directed toward an optical power meter (PM), which continuously measures the power level.</li>
  <li>90% of the pump power continues toward the waveguide, where it is used for SFWM.</li>
  <li>The PM provides real-time feedback on power fluctuations, which is essential for active stabilization.</li>
</ul>

<h3>(b) Variable Optical Attenuator (VOA)</h3>
<p>Once the power meter detects fluctuations, a Variable Optical Attenuator (VOA) is used to actively regulate the pump power.</p>

<h4>Role of the VOA in Power Stabilization</h4>
<ul>
  <li>The VOA dynamically adjusts the intensity of the pump beam based on feedback from the power meter.</li>
  <li>If the PM detects excess power, the VOA reduces the intensity to prevent fluctuations that could affect photon-pair generation.</li>
  <li>If the PM detects a drop in power, the VOA allows more light through, compensating for the loss.</li>
  <li>This feedback loop ensures that the pump power remains at a constant, stable level.</li>
</ul>


<h2>E. Polarization Control for Efficient SFWM</h2>

<h3>1. Why is Polarization Important?</h3>
<p>Before the pump enters the silicon spiral waveguide, it is necessary to align its polarization to the waveguide’s mode.</p>
<p>The silicon spiral waveguide supports only the <strong>TE (transverse electric)</strong> mode. If the pump light has random polarization, only a fraction of it couples efficiently into the waveguide. This leads to inefficient SFWM and reduced photon-pair generation.</p>

<h3>2. Polarization Controller (PC)</h3>
<p>A <strong>polarization controller (PC)</strong> is placed before the waveguide input. The PC adjusts the pump polarization so that all of the pump power is in the TE mode. This ensures maximum nonlinear interaction in the SFWM process.</p>


**Why is the Power Meter (PM) Placed After the Variable Optical Attenuator (VOA)?**\
The VOA (Variable Optical Attenuator) dynamically adjusts the power of the pump beam to maintain a stable intensity. The PM (Power Meter) continuously measures the output power after the VOA has adjusted the intensity. If the PM were placed before the VOA, it would only detect fluctuations, but not the final adjusted power.

<h2>F. HBT 1 in the Figure: Role and Purpose</h2>

<h3>1. What is HBT 1?</h3>
<p>In the experimental setup (<strong>Figure 2a</strong>), the Hanbury Brown and Twiss (HBT) setup at position 1 is used to characterize the intensity fluctuation of the pump laser before it enters the spiral waveguide.</p>

<h3>2. Why is HBT 1 Needed?</h3>
<p>HBT 1 measures the second-order correlation function <em>g<sup>(2)</sup>(0)</em> of the pump. This helps determine whether the pump exhibits intensity noise or fluctuations.</p>
<ul>
  <li>Since an incoherent pump degrades single-photon purity, it is critical to monitor how stable the pump is before it enters the nonlinear medium (the spiral waveguide).</li>
</ul>

<h3>3. How HBT 1 Works?</h3>
<p>The HBT interferometer consists of a <strong>50:50 beam splitter</strong> and two <strong>single-photon detectors (SNSPDs)</strong>. It records the time difference between detected photons and builds a correlation function <em>g<sup>(2)</sup>(τ)</em>.</p>
<ul>
  <li>If the pump is ideal (perfectly coherent CW light), <em>g<sup>(2)</sup>(0) ≈ 1</em> (Poissonian statistics).</li>
  <li>If there are intensity fluctuations, <em>g<sup>(2)</sup>(0) > 1</em>, indicating an incoherent pump, which would negatively affect the single-photon purity of the heralded source.</li>
</ul>

<h3>4. Key Purpose of HBT 1</h3>
<ul>
  <li>Verifies the coherence and stability of the pump before nonlinear interaction.</li>
  <li>Helps quantify the impact of intensity fluctuations on the single-photon purity of the heralded source.</li>
</ul>

## G. SPIRAL WAVEGUIDE

The spiral waveguide is a key component of the heralded single-photon source (HSPS) based on spontaneous four-wave mixing (SFWM).
1. Purpose of the Spiral Waveguide
Two pump photons (frequency &omega;<sub>p</sub>
​) are converted into a signal photon (&omega;<sub>s</sub>
​) and an idler photon (&omega;<sub>i</sub>
​), symmetrically spaced around &omega;<sub>p</sub>
​.  
#####Energy Conservation</h3>
    <p>In SFWM, <strong>two pump photons</strong> (each with frequency&omega;<sub>p</sub>
​) ) are annihilated, and <strong>two new photons</strong> are created:</p>
    <ul>
        <li><strong>A signal photon</strong> &omega;<sub>s</sub>
​
        <li><strong>An idler photon</strong> &omega;<sub>i</sub>
​
<p>Energy must be conserved, so:</p>

    <p style="text-align: center; font-size: 18px;">
    2&hbar;&omega;<sub>p</sub> = &hbar;&omega;<sub>s</sub> + &hbar;&omega;<sub>i</sub>
    &rArr; 2&omega;<sub>p</sub> = &omega;<sub>s</sub> + &omega;<sub>i</sub>
</p>

    <p>(where &hbar; is the reduced Planck constant).</p>
This means the total energy of the pump photons equals the combined energy of the signal and idler photons.

Symmetric Frequency Spacing

<p>The phrase <strong>"symmetrically spaced around &omega;<sub>p</sub>"</strong> means the signal and idler frequencies are equidistant from the pump frequency:</p>

<p style="text-align: center; font-size: 18px;">
    &omega;<sub>s</sub> = &omega;<sub>p</sub> + &Delta;&omega; &nbsp; and &nbsp;
    &omega;<sub>i</sub> = &omega;<sub>p</sub> - &Delta;&omega;
</p>

<p>(where &Delta;&omega; is the frequency detuning).</p>

<ul>
    <li><strong>Pump wavelength:</strong> 1550.52 nm (&omega;<sub>p</sub>)</li>
    <li><strong>Signal wavelength:</strong> 1546.92 nm (ITU channel C39, &omega;<sub>s</sub>)</li>
    <li><strong>Idler wavelength:</strong> 1554.94 nm (ITU channel C28, &omega;<sub>i</sub>)</li>
</ul>

<p>Here, &omega;<sub>s</sub> and &omega;<sub>i</sub> are symmetrically placed around &omega;<sub>p</sub> in the frequency spectrum:</p>

<ul>
    <li>Signal is shifted up by ~1.8 THz.</li>
    <li>Idler is shifted down by ~1.8 THz.</li>
</ul>
Why Symmetry Matters

<ul>
    <li><strong>Phase Matching:</strong> The symmetric spacing ensures momentum conservation
        (<span style="font-size: 18px;">2k<sub>p</sub> = k<sub>s</sub> + k<sub>i</sub></span>, where <i>k</i> is the wavevector),
        which is critical for efficient photon-pair generation.</li>
    <li><strong>Noise Reduction:</strong> The signal and idler photons are spectrally separated from the pump, allowing filters to block residual pump light,
        which is key for achieving high single-photon purity.</li>
</ul>

## H. Dense Wavelength Division Multiplexer (DWDM)

Dense Wavelength Division Multiplexer (DWDM) a critical optical filter used in the experiment to isolate signal (ωₛ) and idler (ωᵢ) photons from the pump (ωₚ) with high precision.

<h3>Key Features</h3>
            <ul>
                <li><strong>Channel Spacing:</strong> 0.8 nm (100 GHz ITU grid)</li>
                <li><strong>Passband Width:</strong> 0.6 nm (for signal/idler photons at 1546.92 nm and 1554.94 nm)</li>
                <li><strong>Isolation:</strong> &gt;105 dB (blocks residual pump light and noise)</li>
            </ul>
        

     
<h3>Role in the Experiment</h3>
            <ul>
                <li><strong>Noise Suppression:</strong> Rejects pump photons and spontaneous Raman scattering, enabling high coincidence-to-accidental ratio (CAR &gt; 2000).</li>
                <li><strong>Photon Routing:</strong> Directs signal/idler photons to separate superconducting nanowire single-photon detectors (SNSPDs).</li>
            </ul>
      
      
<h3>Advantage Over Other Filters</h3>
            <ul>
                <li><strong>Precision:</strong> Narrow passbands ensure minimal loss of correlated photon pairs.</li>
                <li><strong>Scalability:</strong> Compatible with on-chip integration for future quantum photonic circuits.</li>
