# Abstract

A quantum state is nonclassical if its Glauber-Sudarshan P function cannot be interpreted as a classical probability density. However, direct experimental verification of such nonclassicality is challenging due to the P function’s inherent singularities. This work demonstrates the experimental reconstruction of a well-behaved but nonclassical P function for a single-photon-added thermal state (SPATS), which exhibits statistically significant negativities. By conditionally preparing SPATS via parametric amplification of a pseudothermal source and employing balanced homodyne detection, the P function is determined through characteristic function sampling and Hankel transform. The results directly visualize nonclassicality via negativity. This approach highlights the robustness of P-function-based nonclassicality detection against experimental inefficiencies and thermal noise, providing a foundational validation of quantumness as originally defined.

# Classical Description of Light

In classical physics, light is described as a continuous electromagnetic wave characterized by its amplitude and phase. The intensity of light in classical terms is proportional to the square of the amplitude of the electric field:

$$
I \propto |E|^2.
$$

Classical light waves can interfere and diffract, exhibiting purely wave-like phenomena.

# Quantum Description of Light

Quantum mechanics introduces the concept of quantized electromagnetic fields. The modern quantum theory of light, quantum electrodynamics (QED), describes light as quantized excitations of the electromagnetic field—photons—where the electromagnetic field is quantized and the photon represents the smallest quantum of light. The quantum states of light are fundamentally different from classical descriptions and can exhibit non-classical behavior under certain conditions.

In a broader picture, we can classify the quantum states of light into three main categories:

- **Coherent States:** These states resemble classical waves and are often associated with laser light.
- **Thermal States:** These states describe light in thermal equilibrium.
- **Single-Photon States:** Also known as Fock states, these represent a fixed number of photons.

# Coherent States and its P function Representation

Coherent states, introduced by Roy Glauber in 1963 (for which he received the Nobel Prize), represent the quantum states that most closely resemble classical electromagnetic waves.

Quantum states of light can be represented in phase space using various quasi-probability distributions. A coherent state $ |\alpha\rangle $ is defined as a right-hand eigenstate of the annihilation operator $ \hat{a} $:

$$
\hat{a} |\alpha\rangle = \alpha |\alpha\rangle
$$

where $ \alpha $ is a complex number representing the amplitude and phase of the field. The coherent state can be expanded in the Fock (number) state basis as:

$$
|\alpha\rangle = e^{-|\alpha|^2/2} \sum_{n=0}^{\infty} \frac{\alpha^n}{\sqrt{n!}} |n\rangle.
$$

The Glauber-Sudarshan P-representation allows any quantum state $ \hat{\rho} $ to be expressed as:

$$
\hat{\rho} = \int P(\alpha) |\alpha\rangle \langle \alpha| d^2\alpha
$$

where $ P(\alpha) $ is the Glauber-Sudarshan P function, analogous to phase-space distributions in statistical mechanics. The normalization of the density matrix requires:

\begin{align}
\text{Tr} \hat{\rho} &= 1 \\
\text{Tr} \left[\int P(\alpha) |\alpha\rangle \langle \alpha| d^2\alpha \right] &= \int \left(\sum_n P(\alpha) \langle n|\alpha\rangle \langle \alpha|n\rangle \right) d^2\alpha \nonumber \\
&= \int P(\alpha) \langle \alpha|\alpha\rangle d^2\alpha \nonumber \\
&= \int P(\alpha) d^2\alpha = 1.
\end{align}

This shows that the $ P $ function integrates to unity, as expected for a probability distribution. 
## Derivation of P-function of coherent state
Let us look at what the P functions of a coherent state. Suppose you have a state $ \hat{\rho} $, you get the corresponding P function as follows:
Let $|\beta\rangle $ be a coherent state.

\begin{align*}
\langle \beta | \hat{\rho} | \beta' \rangle &= \int P(\alpha) \langle \beta | \alpha \rangle \langle \alpha | \beta' \rangle \, d^2\alpha \\
&= \int P(\alpha) \exp\left[ -\frac{1}{2} |\beta|^2 - \frac{1}{2} |\alpha|^2 - \beta^* \alpha \right] \\
&\quad \times \exp\left[ -\frac{1}{2} |\alpha|^2 - \frac{1}{2} |\beta|^2 + \alpha^* \beta \right] \, d^2\alpha \\
&= e^{-|\beta|^2} \int P(\alpha) e^{-|\alpha|^2} e^{\alpha^* \beta - \alpha \beta^*} \, d^2\alpha
\end{align*}

We see that $\langle \beta| \hat{\rho} |\beta'\rangle $ is related to $ P(\alpha)$ by a complex Fourier transform. So $ P(\alpha)$ is given by

$$
P(\alpha) = \frac{e^{|\alpha|^2}}{\pi^2} \int e^{|\beta|^2} \langle \beta | \hat{\rho} | \beta' \rangle \, e^{\beta^* \alpha - \beta \alpha^*} \, d^2\beta
$$

Let us consider the pure coherent state $ |\gamma\rangle $, where $ \hat{\rho} = |\gamma\rangle \langle \gamma| $.
$$
\langle \beta| \hat{\rho} |\beta \rangle = \langle \beta | \gamma \rangle \langle \gamma | \beta \rangle
= e^{-|\gamma|^2} e^{-|\beta|^2} e^{-\beta^* \gamma + \gamma^* \beta}
$$
$$
\implies P(\alpha) = e^{|\alpha|^2} e^{-|\gamma|^2} \frac{1}{\pi^2} \int e^{\beta^*(\alpha - \gamma) - \beta(\alpha^* - \gamma^*)} \, d^2\beta
$$
But the integral is just the two-dimensional form of the Dirac delta
function:
$$
\delta^{(2)}(\alpha - \gamma) = \frac{1}{\pi^2} \int e^{\beta^*(\alpha - \gamma) - \beta(\alpha^* - \gamma^*)} \, d^2\beta
$$

Therefore
$$
P(\alpha) = \delta^{(2)}(\alpha - \gamma)
$$
You can compare this to a classical oscillator occupying a precise point  $ \gamma $ in phase space.


However, for some quantum states, $P(\alpha)$ can have properties quite unlike any classical probability distribution: it can become negative or highly singular. States exhibiting these features are called *nonclassical states*.

When one calculates expectation values of normally ordered operators in this representation, the expressions can look formally identical to classical statistical averages. For an operator $\hat{O}$ that is in normal-ordered form (all creation operators to the left of annihilation operators), one can write:

$$
\langle \hat{O} \rangle \;=\; \mathrm{Tr}(\hat{\rho}\,\hat{O})
\;=\; \int d^2\alpha \; P(\alpha)\,\langle \alpha|\hat{O}|\alpha\rangle.
$$

If $P(\alpha)$ were a well-behaved, nonnegative function, it could be interpreted just like a classical probability density in the complex phase space. However, the key point is that, in many quantum states, the $P(\alpha)$ function becomes **negative** or even **singular**. Such behavior violates the usual requirements for a classical probability distribution. In classical probability theory, a valid probability density must be nonnegative everywhere and integrable to 1. Therefore, whenever $P(\alpha)$ fails to meet these criteria, we say that the state exhibits *nonclassical* properties.

For states showing classical behaviour (coherent and thermal), $P(\alpha)$ behaves like a true probability distribution:

- **Coherent states**: Resemble classical waves with Poissonian fluctuations where $P(\alpha) = \delta^2(\alpha - \beta)$.
- **Thermal states**: Exhibit enhanced (super-Poissonian) fluctuations characteristic of chaotic light where $P(\alpha)$ is a Gaussian function.

However, for states showing non-classical behaviour like Fock states, $P(\alpha)$ exhibits non-classical features:

- **Single-photon state (Fock states)** ($|1\rangle$): These demonstrate purely quantum mechanical behavior with sub-Poissonian statistics and photon antibunching where $P(\alpha)$ involves derivatives of delta functions.
- **Squeezed states**: $P(\alpha)$ takes negative values in some regions.

These distinctions are crucial for quantum technologies, where non-classical light sources like single-photon emitters enable applications such as quantum cryptography and quantum communication. The phase-space representations, particularly the Glauber-Sudarshan P-function, provide a powerful tool for characterizing the classical or non-classical nature of light states through their quasi-probability distributions.



# The P-Function for Single-Photon State: Derivation and Implications

The Glauber–Sudarshan $ P $ function for a single-photon state is a prime example of how quantum states can defy a classical description. For the one-photon Fock state $ |1\rangle $, the $ P $ function is highly singular and is commonly expressed as:

$$
P(\alpha) = \left(1 + \frac{\partial}{\partial \alpha} \frac{\partial}{\partial \alpha^*} \right) \delta(\alpha),
$$

where $ \delta(\alpha) $ is the two-dimensional Dirac delta function in the complex phase space, and the derivative operators account for the single quantum excitation relative to the vacuum state.

### Derivation:

The P-function is defined through the relation:

$$
|n\rangle \langle n| = \int d^2\alpha \, P_n(\alpha) |\alpha\rangle \langle \alpha|.
$$

For $ n=1 $, our goal is to derive $ P_1(\alpha)$. We start with the number state representation in terms of coherent states:

$$
|1\rangle \langle 1| = \hat{a}^\dagger |0\rangle \langle 0| \hat{a}.
$$

Using the resolution of identity in terms of coherent states:

$$
|0\rangle \langle 0| = \int d^2\alpha \, \delta(\alpha) |\alpha\rangle \langle \alpha|,
$$

we multiply both sides by $ \hat{a}^\dagger $ and $ \hat{a} $:

$$
\hat{a}^\dagger |0\rangle \langle 0| \hat{a} = \int d^2\alpha \, \delta(\alpha) \left(\alpha^* + \frac{\partial}{\partial \alpha} \right) |\alpha\rangle \langle \alpha| \left(\alpha + \frac{\partial}{\partial \alpha^*} \right).
$$

Expanding the terms inside:

$$
\left(\alpha^* + \frac{\partial}{\partial \alpha} \right) \left(\alpha + \frac{\partial}{\partial \alpha^*} \right).
$$

Expanding term-by-term:

$$
\alpha^* \alpha + \alpha^* \frac{\partial}{\partial \alpha^*} + \frac{\partial}{\partial \alpha} \alpha + \frac{\partial}{\partial \alpha} \frac{\partial}{\partial \alpha^*}.
$$
So we can rewrite:
$$
|1\rangle \langle 1| = \int d^2\alpha \,  \left(\alpha^* \alpha + \alpha^* \frac{\partial}{\partial \alpha^*} + \frac{\partial}{\partial \alpha} \alpha + \frac{\partial}{\partial \alpha} \frac{\partial}{\partial \alpha^*} + \frac{\partial}{\partial \alpha} \frac{\partial}{\partial \alpha^*} \right)\delta(\alpha) |\alpha\rangle \langle \alpha|.
$$
After simplifying, we get the expression as,
$$
|1\rangle \langle 1| = \int d^2\alpha \, \left( 1 + \frac{\partial}{\partial \alpha} \frac{\partial}{\partial \alpha^*} \right)\delta(\alpha)  |\alpha\rangle \langle \alpha|.
$$


Thus, we obtain the $ P_1(\alpha) $ after comparing with the general expression :

$$
P_1(\alpha) = \left(1+ \frac{\partial}{\partial \alpha} \frac{\partial}{\partial \alpha^*} \right) \delta(\alpha).
$$
As we can see  P function essentially tells us about the "quantumness" of the system, so we have to ask how can we measure  the properties of non-classical states which are directly related to the P function? We have to remember when we do a realistic experimental setup, demonstrating the non-classicality of a quantum state via its Glauber--Sudarshan $ P $ function is complicated by the presence of losses and imperfect detection. These effects can be modeled by treating the quantum state as if it undergoes a beam-splitter interaction, where only a fraction $ \eta $ of the field amplitude is transmitted to the detector. In phase space, this process manifests as a simple rescaling of the quadrature variable (or complex amplitude), $ \alpha \to \sqrt{\eta}\,\alpha$.

Consequently, the measured quasiprobability distribution, denoted $ P_{\eta}(\alpha)$, is
related to the ideal (lossless) distribution $P(\alpha)$ by
$$
P(\alpha) \;=\; \eta \; P_{\eta}\!\bigl(\sqrt{\eta}\,\alpha\bigr).
$$
The factor $\sqrt{\eta}$ in the argument compresses the phase-space distribution, while
the overall prefactor $\eta$ accounts for the reduced amplitude and ensures proper
normalization. Although this rescaling makes negative or singular features in the $ P $
function less pronounced (and thus more difficult to observe), it does not fundamentally erase them. If the efficiency $ \eta $ is known, one can in principle undo the rescaling to reconstruct the original $ P(\alpha)$, thereby recovering any nonclassical signatures present in the state.

## SPATS as a Tool for Experimental Determination of P-Functions



Many of the  nonclassical states exhibit highly singular $P$  functions (e.g., involving derivatives of delta functions). Such singular distributions are nearly impossible to reconstruct directly from experimental data because  experimental noise and finite resolution tend to obscure the sharp features or infinities.

One way to determine the P function experimentally is the use of single photon added thermal states (SPATS). By using a thermal state $ \hat{\rho}_{th}$ with mean photon number $\overline{n}$. By a photon creation, we would get the SPATS, $\hat{\rho} = \mathcal{N}\hat{a}^\dagger\hat{\rho}_{th}\hat{a}$. The $P $ function now although is well behaved but still violates the properties of classical probability distribution and is given by \begin{equation}
    P_1(\alpha) = \frac{1}{\pi \bar{n}^3} \left[ (1+\bar{n}) |\alpha|^2 - \bar{n} \right] e^{-|\alpha|^2 / \bar{n}}
\end{equation}
### Derivation:
for single photon states, the density function behaves as 
\begin{equation} 
\rho \propto \hat{a}^{\dagger} e^{-\beta \hat{a}^{\dagger} \hat{a}} \hat{a}, \quad \beta > 0, \quad m = \text{integer}
\end{equation}
We can write this in the context of thermal states as follows
\begin{equation}
    \rho \propto \int e^{-|\alpha|^2 / \bar{n}} a^{\dagger} |\alpha \rangle \langle \alpha | a d^2\alpha, \quad \bar{n} = (e^{\beta} -1)^{-1}.
\end{equation}

 
The derivation of the action of the creation operator $\hat{a}^\dagger$ on a coherent state $|\alpha\rangle$ is as follows:

$$
\begin{aligned}
\hat{a}^\dagger |\alpha\rangle e^{|\alpha|^2/2} &= \hat{a}^\dagger \left( e^{-|\alpha|^2/2} \sum_{n=0}^\infty \frac{\alpha^n}{\sqrt{n!}} |n\rangle \right) e^{|\alpha|^2/2} \\
&= \hat{a}^\dagger \sum_{n=0}^\infty \frac{\alpha^n}{\sqrt{n!}} |n\rangle = \sum_{n=0}^\infty \sqrt{n+1} \frac{\alpha^n}{\sqrt{n!}} |n+1\rangle \\
&= \sum_{m=1}^\infty \sqrt{m} \frac{\alpha^{m-1}}{\sqrt{(m-1)!}} |m\rangle = \sum_{m=0}^\infty \sqrt{m} \frac{\alpha^{m-1}}{\sqrt{(m-1)!}} |m\rangle \\
&= \sum_{m=0}^\infty m \frac{\alpha^{m-1}}{\sqrt{m!}} |m\rangle \\
&= \frac{\partial}{\partial \alpha} \left[ \sum_{m=0}^\infty \frac{\alpha^m}{\sqrt{m!}} |m\rangle e^{-|\alpha|^2/2} e^{|\alpha|^2/2}\right]  \\
&= \frac{\partial}{\partial \alpha}\left[ |\alpha\rangle e^{|\alpha|^2/2} \right].
\end{aligned}
$$
This shows that the coherent state $|\alpha\rangle$ is an eigenstate of the annihilation operator $\hat{a}$ with eigenvalue $\alpha$.
one can show that it can be reduced to
\begin{equation}
    \rho \propto \int \frac{1}{\pi \bar{n}} 
    \left( e^{|\alpha|^2} \frac{\partial^{2}}{\partial \alpha \partial \alpha^*} e^{-|\alpha|^2 (1+1/\bar{n})} \right)
    |\alpha \rangle \langle \alpha | d^2\alpha.
\end{equation}

The derivatives in  can be expressed in terms of the Laguerre polynomials \( L_1 \). The final result for the \( P \) function after normalization is found to be

\begin{equation}
    P_1(\alpha) = \frac{1}{\pi \bar{n}} e^{-|\alpha|^2 / \bar{n}} L_1 \left( \left[ 1+ \frac{1}{\bar{n}} \right] |\alpha|^2 \right) / D,
\end{equation}

where

\begin{equation}
    D = \left\{ \sum_{p=0}^{1} \frac{(-1)^p (\bar{n} +1)^p}{(1 - p)! p! 2^{2p}} 
    \left[ \sum_{q=0}^{p} \frac{2(p-q)! 2q!}{(p-q)! q!^2} \right] \right\}.
\end{equation}

The distribution $ P_1(\alpha) $ clearly becomes $\textbf{negative}$ as Laguerre polynomials oscillate between positive and negative values.
Hence, after simplifying the final expression becomes \begin{equation}
P_1(\alpha) = \frac{1}{\pi \bar{n}^3} \left[ (1+\bar{n}) |\alpha|^2 - \bar{n} \right] e^{-|\alpha|^2 / \bar{n}}
\end{equation}

```python

def L0(x):
    return np.ones_like(x) 

def L1(x):
    return -x + 1.0        

def L2(x):
    return 0.5 * (x**2 - 4*x + 2)

def L3(x):
    return (1.0 / 6.0) * (-x**3 + 9*x**2 - 18*x + 6)

# x from 0 to 5
x = np.linspace(-1, 5, 500)

# Evaluate each polynomial => each y has shape (500,)
y0 = L0(x)
y1 = L1(x)
y2 = L2(x)
y3 = L3(x)

plt.figure(figsize=(8, 5))

plt.plot(x, y1, label='L1')
plt.plot(x, y2, label='L2')
plt.plot(x, y3, label='L3')
plt.axhline(0, color='black', linestyle='-', linewidth=0.8)
plt.axvline(0, color='black', linestyle='-', linewidth=0.8)
# Optionally fill the negative areas:
plt.fill_between(x, y0, 0, where=(y0 < 0), alpha=0.2)
plt.fill_between(x, y1, 0, where=(y1 < 0), alpha=0.2)
plt.fill_between(x, y2, 0, where=(y2 < 0), alpha=0.2)
plt.fill_between(x, y3, 0, where=(y3 < 0), alpha=0.2)

plt.xlabel('x')
plt.ylabel('L(x)')
plt.title('Laguerre Polynomials L1, L2, L3')
plt.legend()
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D

def P_alpha(x, y, n_bar):
    """
    Compute P(alpha) = (1/(π n_bar^3)) * [ (1 + n_bar) * r^2 - n_bar ] * exp(-r^2 / n_bar)
    where r^2 = x^2 + y^2.
    """
    r2 = x**2 + y**2
    return (1.0 / (np.pi * n_bar**3)) * ((1.0 + n_bar) * r2 - n_bar) * np.exp(-r2 / n_bar)

# Parameters
n_bar = 1.11

# Grid setup for Re(alpha) and Im(alpha)
x_min, x_max = -2.3, 2.3
y_min, y_max = -2.3, 2.3
num_points = 200

x_vals = np.linspace(x_min, x_max, num_points)
y_vals = np.linspace(y_min, y_max, num_points)
X, Y = np.meshgrid(x_vals, y_vals)

# Compute P(alpha) on the grid
Z = P_alpha(X, Y, n_bar)

# Create the figure and 3D axis
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis',
                       edgecolor='blue',
                       alpha=0.8, linewidth=0.05, antialiased=True)
ax.view_init(elev=20, azim=60)
# Label axes
plt.tight_layout()
ax.set_xlabel(r'$\mathrm{Re}(\alpha)$')
ax.set_ylabel(r'$\mathrm{Im}(\alpha)$')
#ax.set_zlabel(r'$P(\alpha)$')

# Optionally set a title
ax.set_title(r'$P(\alpha)$ for $\bar{n} = 1.11$')

# Add a color bar
fig.colorbar(surf, shrink=0.6, aspect=15, label=r'$P(\alpha)$')

plt.show()
```

```python

# Define the P(α) function
def P_alpha(alpha, n_bar):
    """Calculate Glauber-Sudarshan P function"""
    return (1.0 / (np.pi * n_bar**3)) * ((1.0 + n_bar) * alpha**2 - n_bar) * np.exp(-alpha**2 / n_bar)
# Parameters
n_bar = 1.11  # Case (a)
eta = 0.6
alpha_range = np.linspace(-5, 5, 500)  # From -5 to 5

# Calculate P(α) for both cases
P = eta*P_alpha(np.sqrt(eta)*alpha_range, n_bar)

# Create the plot
#plt.figure(figsize=(8, 6))

# Plot both cases
plt.plot(alpha_range, P, 'b-', linewidth=2)

# Add horizontal line at P=0
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.8)

# Highlight negative regions (non-classicality)
plt.fill_between(alpha_range, P, where=(P<0), color='skyblue', alpha=0.2)

# Add labels and title
plt.xlabel('α', fontsize=12)
plt.ylabel('P(α)', fontsize=12)
plt.title(f'Glauber-Sudarshan P Function, n̄={n_bar}, $\eta$={eta}', fontsize=14)

# Set axis limits and ticks
plt.xlim(-5, 5)
plt.xticks(np.arange(-5, 6, 1))
plt.grid(True, linestyle='--', alpha=0.5)


plt.tight_layout()
plt.show()
```

## Spontaneous Parametric Down Conversion

To understand the experimental setup used in this paper, we first need to understand **Spontaneous Parametric Down Conversion (SPDC)**. SPDC is a quantum process in which a pump photon from a laser interacts with a nonlinear crystal and is converted into two photons of lower energy. These two photons are created simultaneously and conserve energy and momentum. They are emitted into two distinct spatial or spectral modes, referred to as the **signal** and **idler** modes.

## The Components & Procedure

The goal of this experiment is to create a **Single-Photon-Added Thermal State (SPATS)**. After creating the SPATS, we can measure its quadrature via homodyne detection to reconstruct its non-classical properties. Let's examine the components of the setup and their purposes. SPATS generation involves combining a classical thermal light source with a single-photon source, resulting in a quantum state with non-classical properties.

- A **continuous Laser light** is produced using a Ti-Sapphire mode-locked laser (mode-locked means the wavelength and phase of the beam are constant). The generated light will have pulses at fundamental wavelengths around $780-800 \, \text{nm}$. This laser provides light for pumping as well as serving as a local oscillator for homodyne detection.
  
- The laser light is incident on a **Lithium Triborate (LBO) crystal**, which converts the fundamental wavelength light ($800 \, \text{nm}$) into second harmonic light ($400 \, \text{nm}$). This is crucial for driving the parametric amplifier in the next stage.

- The $\sim 400 \, \text{nm}$ light is then incident on a **Barium Borate (BBO) crystal**, which induces parametric down-conversion (PDC), as discussed above.

- To **seed** thermal light into the pump, a portion of the laser beam is directed onto a **Rotating Ground Glass Disk (RD)** to randomize the phase and amplitude, mimicking a thermal light source. Later, photons are added by an **Optical Parametric Amplifier (OPA)**.

- After seeding thermal light into the pumping beam, the resultant beam passes through the **BBO**, which emits two photons of equal energy (called **idler** and **signal**). A **Signal-Photon-Counting Module (SPCM)** is placed in the idler beam path. When it "clicks," it implies the generation of a photon in the idler mode, which further implies that the signal mode now contains a photon-added state. A **filter** before the SPCM ensures that only the correct wavelength from the BBO reaches the detector, reducing noise.

- Finally, **Balanced Homodyne Detection (BHD)** is used. BHD is a method to extract information encoded as modulation of the phase or frequency of an oscillating signal by comparing it with a standard oscillation. Here, a **Local Oscillator (LO)** is used to measure the quadratures of the signal mode.

- A **beam splitter** is used to interfere the heralded single photons with the thermal light beam. Proper alignment ensures effective overlap and interaction between the two fields. The output modes are directed toward **photodetectors**. The currents from both photodetectors are subtracted, measuring different quantum state quadratures as the phase of the LO changes.

- For each laser pulse, the homodyne detector records a quadrature value. Many quadrature measurements are taken at random phase settings. The state and its quadrature distribution are assumed to be independent of the LO phase, simplifying the reconstruction of the $P$-function due to radial symmetry in phase space. A Hankel transform is then applied to reconstruct the Glauber–Sudarshan $P$ function.

The presence of negative values in the reconstructed $P$ function directly confirms the nonclassicality of the SPATS state.

## The Components and Procedure

The experimental setup used to create and analyze a SPAT state includes the following components:

### 1. Thermal Light Source:
A pseudothermal light source is generated by passing a coherent laser beam through a rotating ground-glass disk. This scattering introduces random amplitude and phase fluctuations, resulting in a speckle pattern of light with thermal statistics. The collected light will have a Bosonian nature with $\bar{n}$ as the average photon number.

### 2. Photon Addition via Parametric Down-Conversion:
The thermal light is sent into the signal mode of a nonlinear crystal pumped by a strong laser. Inside the crystal, spontaneous parametric down-conversion (SPDC) occurs, producing correlated photon pairs in the signal and idler modes.

### 3. Conditional Detection (Heralding):
A single-photon detector is placed in the idler mode. When it clicks, it heralds that a photon pair has been created. Because the idler mode started in a vacuum, a click indicates that a photon has been added to the thermal light in the signal mode, creating a SPAT state as follows:

$$
\boxed{\hat{\rho}_{SPATS} = \mathcal{N} \hat{a}^\dagger \hat{\rho}_{th}\hat{a}}
$$

### 4. Homodyne Detection:
The quadratures of the SPATS state are measured using balanced homodyne detection. In this process, the signal beam is mixed with a strong local oscillator (a coherent laser beam with known phase) at a 50:50 beam splitter. The output intensities are measured with two photodiodes, and the difference in their signals gives the quadrature value. Different quadratures can be measured by adjusting the relative phase between the SPATS state and the LO.

### 5. Data Acquisition and $P$-Function Reconstruction:
For each laser pulse, the homodyne detector records a quadrature value. Many quadrature measurements are taken at random phase settings. The state and its quadrature distribution are assumed to be independent of the LO phase, simplifying the reconstruction of the $P$-function due to radial symmetry in phase space. A Hankel transform is then applied to reconstruct the Glauber–Sudarshan $P$ function.

The presence of negative values in the reconstructed $P$ function directly confirms the nonclassicality of the SPATS state.

## Experimental Method

Suppose you have experimentally generated light in a certain state. How do you find the $P$ function corresponding to that state? The $P$ function of a quantum state is not directly accessible. Instead, we measure the field quadratures. To determine the $P$ function, we begin by determining the probability distribution of the quadratures using homodyne detection.

A schematic of a balanced homodyne setup is shown below. Mode $a$ is the quantum state you want to measure, and mode $b$ contains a strong coherent classical field, which may be taken as a coherent state of amplitude $\beta$. The beam splitters are assumed to be 50:50. The relation between the input ($\hat{a}, \hat{b}$) and output ($\hat{c}, \hat{d}$) operators is:

$$
\hat{c} = \frac{1}{\sqrt{2}} \left( \hat{a} + i\hat{b} \right)
$$
$$
\hat{d} = \frac{1}{\sqrt{2}} \left( \hat{b} + i\hat{a} \right)
$$

The detectors placed in the output beams measure the intensities $I_c = \langle \hat{c}^\dagger \hat{c} \rangle$ and $I_d = \langle \hat{d}^\dagger \hat{d} \rangle$, and the difference in these intensities is:

$$
I_c - I_d = \langle \hat{n}_{cd} \rangle = \langle \hat{c}^\dagger \hat{c} - \hat{d}^\dagger \hat{d} \rangle
= i \langle \hat{a}^\dagger \hat{b} - \hat{a} \hat{b}^\dagger \rangle \tag{7.91}
$$

Assuming the $b$ mode is in the coherent state $|\beta e^{-i\omega t} \rangle$, where $\beta = |\beta| e^{-i\psi}$, we have:

$$
\langle \hat{n}_{cd} \rangle = |\beta| \left\langle \hat{a} e^{i\omega t} e^{-i\theta} + \hat{a}^\dagger e^{-i\omega t} e^{i\theta} \right\rangle \tag{7.92}
$$

where $\theta = \psi + \frac{\pi}{2}$. Assuming that the $a$-mode light is also of frequency $\omega$ (and in practice, both the $a$ and $b$ mode light fields derive from the same laser), we can set $\hat{a} = \hat{a}_0 e^{-i\omega t}$ so that we may write:

$$
\langle \hat{n}_{cd} \rangle = 2|\beta| \langle \hat{X}(\theta) \rangle \tag{7.93}
$$

where:

$$
\hat{X}(\theta) = \frac{1}{2} \left( \hat{a}_0 e^{-i\theta} + \hat{a}_0^\dagger e^{i\theta} \right) \tag{7.94}
$$

is the field quadrature operator at the angle $\theta$. By changing $\theta$, which can be done by changing the phase $\psi$ of the local oscillator, we can measure an arbitrary quadrature of the signal field. This gives the probability distribution $P(x_\theta)$ of the quadrature. To obtain the $P$ function, we first calculate the symmetrically ordered characteristic function by taking the Fourier transform of the quadrature probability distribution.


# Characteristic function of SPATS
The characteristic function of a quantum state $\rho$ is defined as:
\begin{equation}
    \Phi(\beta) = \text{Tr} \left[ \rho D(\beta) \right],
\end{equation}
where $D(\beta)$ is the displacement operator:
\begin{equation}
    D(\beta) = \exp \left( \beta a^\dagger - \beta^* a \right).
\end{equation}
This function provides a phase-space description of the quantum state.
A thermal state $\rho_{\text{th}}$ at temperature $T$ (or mean photon number $\bar{n}$) is given by:
\begin{equation}
    \rho_{\text{th}} = \frac{1}{\bar{n}+1} \sum_{n=0}^{\infty} \left( \frac{\bar{n}}{\bar{n} + 1} \right)^n |n\rangle \langle n|,
\end{equation}
where the mean photon number is:
\begin{equation}
    \bar{n} = \frac{1}{e^{\hbar \omega / k_B T} - 1}.
\end{equation}
## Derivation:

A single-photon-added thermal state (SPATS) is constructed by applying the photon creation operator $a^\dagger$ to the thermal state:
\begin{equation}
    \rho_{\text{SPATS}} = \frac{a^\dagger \rho_{\text{th}} a}{\text{Tr} [a^\dagger \rho_{\text{th}} a]}.
\end{equation}


Since the thermal state is written as:
\begin{equation}
    \rho_{\text{th}} = \sum_{n=0}^{\infty} p_n |n\rangle \langle n|,
\end{equation}
where
\begin{equation}
    p_n = \frac{1}{\bar{n}+1} \left( \frac{\bar{n}}{\bar{n}+1} \right)^n,
\end{equation}
applying the creation operator $a^\dagger$ gives:
\begin{equation}
    a^\dagger |n\rangle = \sqrt{n+1} |n+1\rangle.
\end{equation}
Thus,
\begin{equation}
    a^\dagger \rho_{\text{th}} a = \sum_{n=0}^{\infty} p_n (n+1) |n+1\rangle \langle n+1|.
\end{equation}


The trace for normalization is:
\begin{equation}
    \text{Tr} [a^\dagger \rho_{\text{th}} a] = \sum_{n=0}^{\infty} p_n (n+1).
\end{equation}
Using the thermal distribution properties,
\begin{equation}
    \sum_{n=0}^{\infty} n p_n = \bar{n},
\end{equation}
which implies
\begin{equation}
    \sum_{n=0}^{\infty} (n+1) p_n = \bar{n} + 1.
\end{equation}
Thus, the normalized SPATS density matrix is:
\begin{equation}
    \rho_{\text{SPATS}} = \frac{1}{\bar{n} + 1} \sum_{n=0}^{\infty} p_n (n+1) |n+1\rangle \langle n+1|.
\end{equation}


The characteristic function is:
\begin{equation}
    \Phi_{\text{SPATS}}(\beta) = \text{Tr} \left[ \rho_{\text{SPATS}} D(\beta) \right].
\end{equation}
Expanding,
\begin{equation}
    \Phi_{\text{SPATS}}(\beta) = \frac{1}{\bar{n} + 1} \sum_{n=0}^{\infty} p_n (n+1) \langle n+1 | D(\beta) | n+1 \rangle.
\end{equation}
Using the result for displaced number states:
\begin{equation}
    \langle m | D(\beta) | m \rangle = e^{-\frac{1}{2} |\beta|^2} L_m(|\beta|^2),
\end{equation}
where $L_m(x)$ are Laguerre polynomials, we obtain:
\begin{equation}
    \langle n+1 | D(\beta) | n+1 \rangle = e^{-\frac{1}{2} |\beta|^2} L_{n+1}(|\beta|^2).
\end{equation}
Substituting,
\begin{equation}
    \Phi_{\text{SPATS}}(\beta) = \frac{e^{-\frac{1}{2} |\beta|^2}}{\bar{n} + 1} \sum_{n=0}^{\infty} p_n (n+1) L_{n+1}(|\beta|^2).
\end{equation}
Using the sum identity:
\begin{equation}
    \sum_{n=0}^{\infty} p_n (n+1) L_{n+1}(|\beta|^2) = (1 + \bar{n}) (1 - (1+\bar{n}) |\beta|^2) e^{-\bar{n} |\beta|^2},
\end{equation}
we obtain the final result:
\begin{equation}
    \Phi_{\text{SPATS}}(\beta) = \left[1 - (1+\bar{n}) |\beta|^2 \right] e^{-\bar{n} |\beta|^2}.
\end{equation}

# Reconstruction of the P-function

Using $GSP$-function, any quantum state $\hat{\rho}$ can be written in terms of coherent states as given by Eq.-4. We can find the $P(\alpha)$ function using its \textit{characteristic function} $\Phi(\beta)$. The characteristic function, in general, can be defined as:
\begin{equation} \label{eq:5}
    \Phi(\beta) = Tr[\hat{\rho} :e^{\beta \hat{a}^\dagger} e^{-\beta^* \hat{a}}:]
\end{equation}

In experiments, the states are phase independent, so the characteristic function depends only on the magnitude of $\beta$. 

Now, let's find the relation between the characteristic function and $P$-function.

\begin{equation}
    \Phi(\beta) = Tr\left[\int d^2\alpha P(\alpha) |\alpha\rangle \langle\alpha| :e^{\beta \hat{a}^\dagger} e^{-\beta^* \hat{a}}: \right]
\end{equation}

\begin{equation}
    \Phi(\beta) = \int d^2\alpha P(\alpha) 
 Tr \left[|\alpha\rangle \langle\alpha| :e^{\beta \hat{a}^\dagger} e^{-\beta^* \hat{a}}: \right]
\end{equation}

As the trace over an operator with $|\alpha\rangle\langle\alpha|$ is just the expectation value of the operator,

\begin{equation} 
    \Phi(\beta) = \int d^2 \alpha P(\alpha) \langle\alpha|:e^{\beta \hat{a}^\dagger} e^{-\beta^* \hat{a}}:|\alpha\rangle
\end{equation}

As the coherent states are the eigenstates of annihilation operator:

$$
    \hat{a}|\alpha\rangle = \alpha|\alpha\rangle
$$
and due to the normally ordered operator, we can write the above equation as

\begin{equation}
    \boxed{\Phi(\beta) = \int d^2 \alpha P(\alpha) e^{\beta \alpha^*} e^{-\beta^* \alpha}}
\end{equation}

So, $P(\alpha)$ and $\Phi(\beta)$ are related as fourier transform relation, and hence, the inverse fourier transform of $\Phi(\beta)$ yields the $P$-function. This does not exist as a well-behaved function. So the standard reconstruction methods like homodyne detection data or Fourier Transforms are impractical. But we can find the $P$-function using the $\textbf{Hankel transformation}$ with $|\beta| = b$, as:

$$
    P(\alpha) = \frac{2}{\pi} \int_0^\infty b J_0(2b|\alpha|) \Phi(b) db
$$

In practice, there is a cut-off value of $\beta$ beyond which the graph becomes small, hence can be taken as zero. So the truncated reconstruction of the $P$-function can be written as:
\begin{equation}
    \bar{P}(\alpha) = \frac{2}{\pi} \int_0^{|\beta_c|} b J_0(2b|\alpha|) \bar{\Phi}(b) db
\end{equation}



# Error in Characteristic functions
There is other ways of reconstructing the P function, namely, from its characteristic function, $\Phi(\beta)$ where $\Phi(\beta)$ can be obtained from the quadratures in the following way,
$$
\Phi(\beta) = \langle : \hat{D}(\beta) : \rangle = \left\langle e^{i |\beta| \hat{x} [\frac{\pi}{2} - \arg(\beta)]} \right\rangle e^{|\beta|^2/2},
$$ where, $ \hat{D}(\beta)$ denotes the displacement operator and : denotes the normal ordering.
We estimate the characteristic function $ \Phi(\beta) $ using the sample mean:

\begin{equation}
\bar{\Phi}(\beta) = \frac{1}{N} \sum_{j=1}^{N} e^{i |\beta| x_j}e^{|\beta|^2/2},
\end{equation}

where $ x_j $ are the measured quadrature values. Since each $ x_j $ is a random variable, $ \bar{\Phi}(\beta) $ is also a random variable.

## Derivation:
The variance of $ \bar{\Phi}(\beta) $ is given by:

\begin{equation}
\sigma^2 \{\bar{\Phi}(\beta)\} = \mathbb{E}[ |\bar{\Phi}(\beta)|^2 ] - |\mathbb{E}[ \bar{\Phi}(\beta) ]|^2.
\end{equation}


Taking expectation on both sides:

\begin{equation}
\mathbb{E}[ \bar{\Phi}(\beta) ] = \mathbb{E} \left[ \frac{1}{N} \sum_{j=1}^{N} e^{i |\beta| x_j} e^{|\beta|^2/2}\right].
\end{equation}

By linearity of expectation:

\begin{equation}
\mathbb{E}[ \bar{\Phi}(\beta) ] = \frac{1}{N}e^{|\beta|^2/2} \sum_{j=1}^{N} \mathbb{E}[ e^{i |\beta| x_j} ].
\end{equation}

Assuming independent and identically distributed $ x_j $, we get:

\begin{equation}
\mathbb{E}[ \bar{\Phi}(\beta) ] = \frac{1}{N}e^{|\beta|^2/2}[Ne^{-|\beta|^2/2}\bar{\Phi}(\beta)]
\implies \mathbb{E}[ \bar{\Phi}(\beta) ] = \bar{\Phi}(\beta)
\end{equation}

Thus,

\begin{equation}
|\mathbb{E}[ \bar{\Phi}(\beta) ]|^2 = |\bar{\Phi}(\beta)|^2.
\end{equation}


Expanding the squared magnitude:

\begin{equation}
\mathbb{E}[ |\bar{\Phi}(\beta)|^2 ] = \mathbb{E} \left[ \left( \frac{1}{N} \sum_{j=1}^{N} e^{i |\beta| x_j} e^{|\beta|^2/2}\right) \left( \frac{1}{N} \sum_{k=1}^{N} e^{-i |\beta| x_k}e^{|\beta|^2/2} \right) \right].
\end{equation}

Expanding the sum:

\begin{equation}
\mathbb{E}[ |\bar{\Phi}(\beta)|^2 ] = \frac{1}{N^2}e^{|\beta|^2} \sum_{j=1}^{N} \sum_{k=1}^{N} \mathbb{E} [ e^{i |\beta| (x_j - x_k)} ].
\end{equation}

For $ j = k $, each term contributes $ \mathbb{E} [ e^{i |\beta| x_j} e^{-i |\beta| x_j} ] = 1 $.

For $ j \neq k $, using independence:

\begin{equation}
\mathbb{E} [ e^{i |\beta| x_j} e^{-i |\beta| x_k} ] = \mathbb{E} [ e^{i |\beta| x_j} ] \mathbb{E} [ e^{-i |\beta| x_k} ] = |\bar{\Phi}(\beta)|^2.
\end{equation}

Summing these contributions:

\begin{equation}
\mathbb{E}[ |\bar{\Phi}(\beta)|^2 ] = \frac{1}{N^2} \left( N e^{|\beta|^2} + N(N-1) |\bar{\Phi}(\beta)|^2 \right).
\end{equation}

Simplifying,

\begin{equation}
\mathbb{E}[ |\bar{\Phi}(\beta)|^2 ] = \frac{1}{N} e^{|\beta|^2} + \left(1 - \frac{1}{N} \right) |\bar{\Phi}(\beta)|^2.
\end{equation}


\begin{align}
\sigma^2 \{\bar{\Phi}(\beta)\} &= \mathbb{E}[ |\bar{\Phi}(\beta)|^2 ] - |\mathbb{E}[ \bar{\Phi}(\beta) ]|^2 \\
&= \left( \frac{1}{N} e^{|\beta|^2} + \left(1 - \frac{1}{N} \right) |\bar{\Phi}(\beta)|^2 \right) - |\bar{\Phi}(\beta)|^2 \\
&= \frac{1}{N} e^{|\beta|^2} - \frac{1}{N} |\bar{\Phi}(\beta)|^2.
\end{align}
Thus, the final result is:

\begin{equation}
\sigma^2 \{\bar{\Phi}(\beta)\} = \frac{1}{N} [ e^{|\beta|^2} - |\bar{\Phi}(\beta)|^2 ].
\end{equation}
The inverse Fourier transform of $\Phi(\beta)$ yields the P function, which for many non-classical states does not exist as a well-behaved function. However, the sampled characteristic function converges stochastically toward the theoretical one. For radial symmetry of the state, the 2D Fourier transform reduces to the Hankel transform,

\begin{equation}
P(\alpha) = \frac{2}{\pi} \int_{0}^{\infty} b J_0(2b|\alpha|) \Phi(b) \, db.
\end{equation}

Due to the experimental conditions, the disturbing sampling noise gets limited on the reconstructed function

\begin{equation}
\bar{P}(\alpha) = \frac{2}{\pi} \int_{0}^{|\beta|_c} b J_0(2b|\alpha|) \bar{\Phi}(b) \, db
\end{equation}

to a reasonable level. The corresponding variance has been calculated as

\begin{equation}
\sigma^2\{\bar{P}(\alpha)\} = \frac{1}{N} \left( \frac{4}{\pi^2} \int_{0}^{|\beta|_c} \int_{0}^{|\beta|_c} 
bb' J_0(2b|\alpha|) J_0(2b'|\alpha|) \bar{\Phi}(b - b') e^{b b'} \, db \, db' - \bar{P}(\alpha)^2 \right).
\end{equation}

The systematic error

\begin{equation}
\Delta_P(\alpha) = \frac{2}{\pi} \int_{|\beta|_c}^{\infty} b J_0(2b|\alpha|) \Phi(b) \, db
\end{equation}

is estimated with the help of the fitted theoretical function.



```python
import numpy as np
import matplotlib.pyplot as plt

# Function definitions
def Phi_SPATS(beta, n_bar):
    """Characteristic function for pure SPATS state"""
    return (1 - (1 + n_bar) * beta**2) * np.exp(-n_bar * beta**2)

def Phi_thermal(beta, n_bar):
    """Characteristic function for thermal state"""
    return np.exp(-n_bar * beta**2)

def sigma2_Phi(beta, phi_func, n_bar, N):
    """Variance of the characteristic function"""
    phi_beta = phi_func(beta, n_bar)
    return (1/N) * (np.exp(beta**2) - np.abs(phi_beta)**2)

def mixed_state_Phi(beta, n_bar, thermal_fraction=0.19):
    """Characteristic function for mixed SPATS + thermal state"""
    phi_spats = Phi_SPATS(beta, n_bar)
    phi_thermal = Phi_thermal(beta, n_bar)
    return (1 - thermal_fraction)*phi_spats + thermal_fraction*phi_thermal

# Parameters
n_bar_a = 1.11    # Case (a) SPATS
eta_a = 0.60
n_bar_b = 3.71    # Case (b) SPATS with thermal background
eta_b = 0.62
thermal_fraction = 0.19  # 19% thermal background
N = 1000          # Number of samples

# Define the range for |β|
beta_mag = np.linspace(0, 3, 300)

# Calculate both cases using the functions
Phi_a = Phi_SPATS(beta_mag, n_bar_a)
sigma_a = np.sqrt(sigma2_Phi(beta_mag, Phi_SPATS, n_bar_a, N))

Phi_b = mixed_state_Phi(beta_mag, n_bar_b, thermal_fraction)
sigma_b = np.sqrt(sigma2_Phi(beta_mag, lambda b, n: mixed_state_Phi(b, n, thermal_fraction), n_bar_b, N))

# Create the plot
#plt.figure(figsize=(8, 6))
plt.axhline(y=0, color='black', linestyle='-', linewidth=0.8)

# Plot both cases
plt.plot(beta_mag, Phi_a, 'b-', linewidth=2, label=f'(a) SPATS (n̄={n_bar_a}, η={eta_a})')
plt.plot(beta_mag, Phi_b, 'r-', linewidth=2, 
         label=f'(b) SPATS + {thermal_fraction*100:.0f}% thermal (n̄={n_bar_b}, η={eta_b})')

# Fill the standard deviation regions
plt.fill_between(beta_mag, Phi_a - 0.2*sigma_a, Phi_a + 0.2*sigma_a, color='blue', alpha=0.1, label='±0.2σ (a)')
plt.fill_between(beta_mag, Phi_b - 0.2*sigma_b, Phi_b + 0.2*sigma_b, color='red', alpha=0.1, label='±0.2σ (b)')

# Add labels and title
plt.xlabel('|β|', fontsize=12)
plt.ylabel(r'$\Phi$(β)', fontsize=12)
plt.title(r'Characteristic Function $\Phi(\beta)$ Comparison', fontsize=14)

# Set axis limits and ticks
plt.xlim(0, 2.5)
plt.ylim(-1, 1)
plt.xticks(np.arange(0, 3, 0.5))
plt.yticks(np.arange(-1, 1.2, 0.2))

# Add grid, legend and annotations
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(loc='upper right', fontsize=10)

# Add (a) and (b) labels near the curves
plt.text(0.25, 0.85, '(a)', color='blue', fontsize=12)
plt.text(0.5, -0.2, '(b)', color='red', fontsize=12)

plt.tight_layout()
plt.show()
```

# Conclusion and Discussion

In this work, we have successfully demonstrated the experimental reconstruction of a nonclassical Glauber-Sudarshan P-function with measurable negativities, thereby providing a direct visualization of nonclassicality as originally formulated by Glauber and Sudarshan. 
We derived the P-function of a single-photon state, given by
$$
P(\alpha) = \left(1 + \frac{\partial}{\partial \alpha}\frac{\partial}{\partial \alpha^*}\right) \delta(\alpha)
$$

We also derived the P-function for SPATS as:
$$
    P_1(\alpha) = \frac{1}{\pi \bar{n}^3} \left[ (1+\bar{n}) |\alpha|^2 - \bar{n} \right] e^{-|\alpha|^2 / \bar{n}}
$$

By deriving the SPATS characteristic function 
$$
\Phi_{\text{SPATS}}(\beta) = \left[1-(1+\bar{n})|\beta|^2\right]e^{-\bar{n}|\beta|^2}
$$
and performing its Hankel transform, we established a robust theoretical framework that bridges the gap between theoretical predictions and experimental observations. Our approach, involving the photon-added thermal state construction $ \rho_{\text{SPATS}} \propto \hat{a}^\dagger \rho_{\text{th}} \hat{a}$ and the normally-ordered characteristic function formalism, enabled us to reconstruct the P-function and observe negativities, fulfilling our objective of experimentally visualizing nonclassicality.


The inverse Fourier-Hankel transform used to reconstruct the P-function from the characteristic function is given by:
\begin{equation}
    \bar{P}(\alpha) = \frac{2}{\pi} \int_0^{|\beta_c|} b J_0(2b|\alpha|) \bar{\Phi}(b) db
\end{equation}

The appearance of negative values in the reconstructed P-function is deeply rooted in the underlying Laguerre polynomial structure within the characteristic function, resulting from the expression 
$$
\langle n+1 | D(\beta) | n+1 \rangle = e^{-|\beta|^2/2}L_{n+1}(|\beta|^2)
$$
This mathematical structure plays a critical role in generating nonclassical features, which we successfully observed through our combined homodyne-Hankel transform technique. The agreement between our theoretical predictions and experimental findings underscores the validity of our approach and its potential applicability to various quantum states exhibiting nonclassicality.
