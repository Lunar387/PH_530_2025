# Introduction:

- Purification of Photon Indistinguishability for Photonic Quantum Technologies
The generation of pure single photons is a critical requirement for advancing technologies in photonic quantum communication and quantum information processing. As quantum technologies evolve, the need for deterministic single-photon sources becomes increasingly apparent. These sources, typically utilizing two-level emitters, offer a pathway for on-demand photon generation, a key feature in multiphoton applications. Various platforms have been explored for generating single photons, such as color centers in diamonds, organic molecules, trapped atoms, and quantum dots.
- One of the key requirements for photon sources is the capability to generate highly indistinguishable photons in order to realize multiphoton interference—a core process in photonic quantum technologies. Epitaxially grown quantum dots (QD) can generate highly indistinguishable photons due to the ability to precisely engineer their semiconductor environment, and have recently enabled quantum interference experiments with an increasing number of photons. Nonetheless, partial distinguishability remains a central noise mechanism that can pose challenges in the development of emitter-based photonic systems at the scale required in practical applications.
- In solid-state quantum emitters, photon distinguishability arises from both fast and slow physical processes. Fast processes, such as pure dephasing due to interactions with phonons in the environment, occur on timescales shorter than the photon lifetime. On the other hand, slower processes—such as spectral diffusion from interactions with charges and polarization drifts—can also affect photon indistinguishability. To mitigate these noise contributions, traditional methods have focused on reducing temperature to decrease the number of phonons, designing ultra-low-noise chip devices, and reducing emission lifetimes via the Purcell effect to minimize interactions with acoustic phonon modes.
  
- A promising alternative mitigation strategy involves the purification of single photons using linear optical circuits and ancillary photons. Sparrow and Marshall adapted this technique to purify single indistinguishable photons, where multiple partially distinguishable photons are processed through quantum interference in linear-optical circuits.
- In this report, we present an experimental demonstration of linear-optical purification of photon indistinguishability. Using a solid-state quantum dot (QD) as a quantum emitter, we tune both fast and slow noise contributions to control the partial distinguishability of emitted photons. The photons are then processed using a fiber-based linear-optical interferometer, which purifies the photons. The indistinguishability of the purified photons is assessed by simultaneously implementing two copies of the purification circuit and performing quantum interference between the purified output photons. The results show significant improvement in the indistinguishability of the photons compared to the initial state, demonstrating the potential of this approach for developing ultra-low-noise photonic quantum technologies.


# Quantum Interference and the Multi-Permanent Model  

In this section, we detail the theoretical framework used to analyze photon interference in a linear optical circuit. We focus on calculating the permanent for a key 4×4 submatrix—extracted for both the output and reference cases—and show that, after proper normalization, both cases yield a permanent of $(\frac{1}{4}$). This leads to identical probabilities and hence a Hong–Ou–Mandel (HOM) visibility of zero. Finally, we explain the meaning of $(V_{HOM} = 0$).
  
  
## 1. Definition of the Permanent

For an $( n \times n $) matrix (B\), the **permanent** is defined as

$[
\operatorname{Perm}(B) = \sum_{\sigma \in S_n} \prod_{i=1}^{n} B_{i,\sigma(i)},
$]

where $(S_n$) denotes the set of all permutations of $(\{1,2,\dots,n\}$). Unlike the determinant, the permanent does not include alternating signs.  
  
  
## 2. Amplitude for an Input-Output Configuration

For a linear optical circuit, an input state $(|m_1, m_2, \dots, m_N\rangle$) is transformed by a unitary operation $(U$) (with transfer matrix $(M_U$)) into an output state $(|n_1, n_2, \dots, n_N\rangle$). The probability amplitude is given by

$[
\langle n_1, n_2, \dots, n_N|U|m_1, m_2, \dots, m_N \rangle = \operatorname{Perm}(B) \, \frac{\sqrt{\prod_{i} n_i!}}{\sqrt{\prod_{j} m_j!}},
$]

where $(B$) is the submatrix extracted from $(M_U$) corresponding to the relevant input and output modes.

---  
  
  
## 3. Permanent for Partially Distinguishable Photons

When photons are partially distinguishable, a three-dimensional tensor \(W\) is defined as

$[
W_{k,l,j} = B_{kj}\, B^*_{lj}\, S_{lk},
$]

with $(S_{lk} = \langle \psi_{d_l} | \psi_{d_k} \rangle$) (the distinguishability matrix). The generalized permanent is then

$[
\operatorname{Perm}(W) = \sum_{(\sigma,\rho) \in S_n} \prod_{j=1}^{n} W_{\sigma_j, \rho_j, j}.
$]

This extension is essential for simulating interference when photons are not perfectly identical.

---  
  
  
## 4. Relevance of the Multi-Permanent Model

The multi-permanent model is crucial because:

- **Accurate Simulation of Interference:** It extends the standard permanent calculation to account for partial photon distinguishability.
- **Handling Experimental Imperfections:** Factors such as timing jitter, spectral diffusion, and beam splitter non-idealities are incorporated into the model.
- **Quantitative Analysis of HOM Visibility:** By comparing the interference probability (\(P_{out}\)) with a reference probability (\(P_{ref}\)), the model quantifies the impact of interference.
- **Quantum Error Mitigation:** Understanding and simulating interference helps in designing protocols to mitigate errors in quantum optical experiments.

## 5. Circuit Construction

### 5.1 Component Matrices

We consider three unitary matrices $(A$), $(B$), and $(C$) which define the circuit. Each matrix is multiplied by a normalization factor of $( \frac{1}{\sqrt{2}} $).

- **Matrix $( A $):**

  $[
  A = \frac{1}{\sqrt{2}} \begin{pmatrix}
  1 & i & 0 & 0 & 0 & 0 \\
  i & 1 & 0 & 0 & 0 & 0 \\
  0 & 0 & \sqrt{2} & 0 & 0 & 0 \\
  0 & 0 & 0 & \sqrt{2} & 0 & 0 \\
  0 & 0 & 0 & 0 & 1 & i \\
  0 & 0 & 0 & 0 & i & 1 \\
  \end{pmatrix}
  $]

- **Matrix $( B $):**

  $[
  B = \frac{1}{\sqrt{2}} \begin{pmatrix}
  \sqrt{2} & 0 & 0 & 0 & 0 & 0 \\
  0 & 1 & i & 0 & 0 & 0 \\
  0 & i & 1 & 0 & 0 & 0 \\
  0 & 0 & 0 & 1 & i & 0 \\
  0 & 0 & 0 & i & 1 & 0 \\
  0 & 0 & 0 & 0 & 0 & \sqrt{2} \\
  \end{pmatrix}
  $]

- **Matrix \( C \):**

  $[
  C = \frac{1}{\sqrt{2}} \begin{pmatrix}
  \sqrt{2} & 0 & 0 & 0 & 0 & 0 \\
  0 & \sqrt{2} & 0 & 0 & 0 & 0 \\
  0 & 0 & 1 & i & 0 & 0 \\
  0 & 0 & i & 1 & 0 & 0 \\
  0 & 0 & 0 & 0 & \sqrt{2} & 0 \\
  0 & 0 & 0 & 0 & 0 & \sqrt{2} \\
  \end{pmatrix}
  $]

### 5.2 Transfer Matrices

- **Output Case:**

  $[
  M_U = A \times B \times C.
  $]

- **Reference Case:**

  $[
  M_{ref} = A \times B \times I.
  $]

---


## 6. Submatrix Extraction and Calculation

After extracting the $(4 \times 4$) submatrix $(B$) and applying proper normalization, we find:

$[
\operatorname{Perm}(B) = \frac{1}{4}.
$]

---

## 7. HOM Visibility Calculation

The probability for a given configuration is:

$[
P = \left|\operatorname{Perm}(B)\right|^2.
$]

For both cases:

$[
P_{out} = P_{ref} = \left|\frac{1}{4}\right|^2 = \frac{1}{16}.
$]

Thus, the HOM visibility is:

$[
V_{HOM} = 1 - \frac{P_{out}}{P_{ref}} = 1 - 1 = 0.
$]

---

## 8. Interpretation of $(V_{HOM} = 0$)

A HOM visibility of $(0$) means:

- **No Interference Effect:** Photons behave as if they are completely indistinguishable.
- **Classical Behavior:** NPerfect quantum interference is observed.

---

---

## 9. Conclusion

The results confirm that when photons are indistinguishable, perfect quantum interference occurs, and the coincidence rate remains unchanged.


# Phonon Dephasing

- $\textbf{Pure dephasing model}$  
  
The previous model does not take into account the errors involved due to the source used for single photon generation, i.e., quantum dots. This process is called phonon dephasing and the generated photon has the form of a wave-packet, represented by  
  
  $\hat{a}_i^{\dag} = \int_{-\infty}^{\infty}f_i^*(t)\hat{a}_i^{\dag}(t)$  
      
Now, the probability of coincidence for two-photon interference is given by,  
$P_{coin} = \frac{1}{2}(1-\langle|\alpha_{ij}|^2\rangle)$  
  
  where  $\alpha_{ij} = [\hat{a}_i^{\dag},\hat{a}_j] = \int_{-\infty}^{\infty}f_i^*(t)f_j(t)dt$ $\hspace{1cm}$ since $\hspace{1cm}$ $[\hat{a}_i(t),\hat{a}_j^{\dag}(t')] = \delta(t-t')\delta_{ij}$  

![descr](lmimage1.png)

- $\textbf{Dephasing in a quantum dot}$  
  
At temperatures $T > 0 K$, the phonons get excited, and interact with the emitted photons. Hence, a phase uncertainty is introduced in the photon wavepackets produced,  which is denoted by the form of $\hat{a}_i^{\dag}$  
  
  
- $\textbf{Single beamsplitter}$  
  
input $\rightarrow |1,1\rangle$  
  
Now, $|1,1\rangle = \hat{a}_0^{\dag}\hat{a}_1^{\dag}|0,0\rangle_{input} = \frac{1}{2}(\hat{a}_2^{\dag} + i\hat{a}_3^{\dag})(i\hat{a}_2^{\dag}+\hat{a}_3^{\dag})|0,0\rangle_{output} = \frac{1}{2}(i\hat{a}_3^{\dag}\hat{a}_3^{\dag}+i\hat{a}_2^{\dag}\hat{a}_2^{\dag} + [\hat{a}_2^{\dag},\hat{a}_3^{\dag}])|0,0\rangle_{output}$  
  
  
The definition of the commutator gives, the probability of coincidence of two photons at different detectors to be,  
  
$ P_{cc} = \frac{1}{2}(1-\langle|\alpha_{ij}|^2\rangle)$  
where the expectation $\langle ... \rangle$ is done over several realizations of input pairs. $ i,j $ in general, describe the output modes.
  
Hence, the probability of bunching occurring is, $P = \frac{1}{2}(1+\langle|\alpha_{ij}|^2\rangle)$  
  
$P_{cc}$ can be $0$ for perfectly indistinguishable case, $1/2$ for perfectly distinguishable case. Our input is somewhere in between

- $\textbf{Multiple beamsplitters(purification circuit)}$  
  
![descr](lmimage2.png)

  
When the output from the first beam-splitter is introduced as an input for a second beam-splitter, and coupled with another photon mode, this allows only the bunched photons to pass through, leaving out the other modes. Thus, the final probability looks something like  
  
$P_{\alpha} = \frac{1}{2(1 + \langle|\alpha_{ij}|^2\rangle)^2}(\langle|\alpha_{ij}|^2\rangle + \langle|\alpha_{ij}|^2\rangle^2 - 2\langle\alpha_{ij}\alpha_{jl}\alpha_{li}\rangle - \langle\alpha_{ij}\alpha_{jk}\alpha_{kl}\alpha_{li}\rangle)$  
  
The subtracted terms indicate that three-photon and four-photon coincidences are left out, given that bunching occurs  
  
Phonon dephasing is a fast process, and in this case it can be shown that the raw coincidence,  
  
$\langle|\alpha_{ij}|^2\rangle = \frac{\gamma}{\gamma + 2 \gamma_d}$, $\hspace{1cm}$ where $\gamma_d$ is the pure dephasing rate and $\gamma$ is the decay rate of the emitter  
  
  
- $\textbf{Idea behind purification}$  
  
Only events where bunching occurs are allowed. This is recorded indirectly, by treating the absence of a photon in one detector in one event, implying that bunching has occurred in the other detector. This process, when repeated successively, gives rise to bunched, indistinguishable photons  
  
We can write $P_{\alpha}$ in a much simpler form. Let us consider $\langle|\alpha_{ij}|^2\rangle = x$  
Which gives, $\langle\alpha_{ij}\alpha_{jl}\alpha_{li}\rangle \approx x^{3/2} \hspace{1cm}$ and, $\hspace{1cm} \langle\alpha_{ij}\alpha_{jk}\alpha_{kl}\alpha_{li}\rangle \approx x^2$ and by automatic substitution we get,  
  
$P_{\alpha} = \frac{1+x+x^2}{2(1+x)^2} - x^{3/2} - x^2$  
  
$V_f = \langle |\alpha_f|^2\rangle = 1 - 2P_{\alpha} = 1 - \frac{1+x+x^2}{(1+x)^2} + 2x^{3/2} + 2x^2$  
  
for $x \approx 1$ we get, $\langle |\alpha_f|^2\rangle = \frac{x^3 + 5x^2 +8x +3}{(2x+3)(x+1)^3}$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.5, 1, 500)  # x ∈ [0,1] for probabilities

# CORRECTED FORMULA - ensures y ∈ [0,1]
numerator = 1 + x + x**2
denominator = (1 + x)**2
y = (1 - numerator/denominator + 2*x**(1.5) + 2*x**2)/(1+ 2*x**(1.5) + 2*x**2)  # Clip to [0,1]

plt.plot(x, y, 'r-', lw=3, label=r'$\langle|\alpha_f|^2\rangle$')
plt.xlabel(r'Pairwise indistinguishability $x = \langle|\alpha_{ij}|^2\rangle = V_o$', fontsize=12)
plt.ylabel(r'Final indistinguishability $V_f$', fontsize=12)
plt.title('Pure dephasing model', fontsize=14)

plt.show()
```

## Indistinguishability Purification Protocol

The purification protocol that we investigate in this study involves exploiting the difference in statistics between indistinguishable and distinguishable photons. When indistinguishable photons interfere on a beam splitter, they exhibit photon bunching, meaning that both photons are detected simultaneously in the same output mode. On the other hand, distinguishable photons do not exhibit such strong bunching, and their interference events are less frequent.

The purification protocol, originally proposed by Sparrow and further refined by Marshall, uses this difference to enhance the indistinguishable component of a quantum state. The process involves heralding on the absence of a photon in one of the output ports of the beam splitter. This detection event signals that two indistinguishable photons have interfered and are now bunched in the other output mode. Since distinguishable photons are less likely to exhibit this behavior, the heralding event effectively amplifies the indistinguishable component of the state.


## Heralding and Photon Extraction

Once the indistinguishable component is enhanced, a purified single photon can be extracted from the bunched photons. This is achieved by probabilistically splitting the two bunched photons using another beam splitter and detecting a single photon in one of the output arms. The presence of a single photon in one arm of the beam splitter marks the successful extraction of a purified photon.

This procedure can be generalized for multiple photons. For the case where \( n = 2 \), two photons are used to create a purified output photon with improved indistinguishability. This protocol can be extended further by employing repeated HOM interferences followed by zero-photon detection to purify the state progressively. The process is depicted schematically in the purification circuits, where successive stages of purification are performed to achieve an output with higher photon indistinguishability.

## Success Probability

The purification process comes with a trade-off between the success probability and the degree of indistinguishability improvement. The success probability for extracting a purified photon in the case of \( n = 2 \) is 25%, which decreases exponentially as the number of photons involved increases. Specifically, the success probability is related to the number of photons used in the purification process by the formula:


$$P_{\text{success}} = \frac{(n - 1)!}{2P_n} \cdot \frac{n^2}{2n}$$


As the number \( n \) increases, the probability of successful photon extraction drops significantly. However, this decrease in success probability can be mitigated by optimizing the reflectivity of the final beam splitter in the system. Additionally, using different interferometers with varying heralding patterns can further optimize the performance of the purification protocol.




# Experimental Setup for Quantum Interference and Purification Circuits


This section describes the experimental setup used to implement two copies of the \(n = 2\) purification circuits and to perform quantum interference between the two purified outputs. The setup leverages the properties of quantum dots (QDs) embedded in photonic crystal waveguides (PCWs) for deterministic single-photon generation. The system incorporates several advanced techniques, including resonant pumping of the QD, spectral filtering, and time-to-space demultiplexing, to optimize photon indistinguishability and improve the quality of the quantum interference.

## Instruments used in experiment:
InAs quantum dot (QD), photonic crystal waveguide (PCW), pulsed laser, electro-optic modulators (EOMs), polarizing beam splitters (PBSs), half- and quarter-wave plates (HWP and QWP), fiber delays, fiber beam splitters, superconducting-nanowire single-photon detectors (SNSPDs)




![LMI20exp.png](LMIexp.png)

## 1 Four Photon Source

## a) Single-Photon Generation

Deterministic single-photon generation is achieved using a neutral exciton from an InAs quantum dot (QD) embedded in a photonic crystal waveguide (PCW). The QD is pumped resonantly by a pulsed laser that is spectrally shaped with a home-built folded 4-f system to set a bandwidth of approximately 90 pm. This bandwidth matches the wavelength of the QD, which is \( \lambda = 938.4 \, \text{nm} \). Electrical tuning of the QD, facilitated by low-noise electrical contacts, stabilizes the charge state and ensures emission on the desired transition. This also minimizes spectral diffusion caused by residual charge noise.

The single photons are emitted in the PCW and fiber-coupled via a shallow-edged grating and a cryocompatible objective lens. To enhance the indistinguishability of the emitted photons, an etalon with a bandwidth of 32 GHz is used as a frequency filter. This removes the undesired phonon-induced spectral sideband. Pumping the QD with resonant \(\pi\)-pulses at a repetition rate of 80 MHz results in single-photon emission at a measured rate of 16.1 MHz, with an 85% efficient detection system. The photon purity is measured to be \(1 - g(2)(0) = (97.21 \pm 0.01)\%\).

## b) Time-to-Space Demultiplexing

The photon stream is then converted into four simultaneous photon streams in different spatial modes through a time-to-space demultiplexing module. This module consists of three resonantly enhanced electro-optic modulators (EOMs) and polarizing beam splitters (PBSs) arranged in a tree-like structure. The structure routes the photons into four different output modes, each with different fiber lengths to compensate for temporal delays. At this stage, four-photon coincidences are detected at a rate of 3.2 kHz.

## 2 Purifying Circuit

## a) Polarization Control and HOM Interference

Before feeding the photons into the fiber-based Hong-Ou-Mandel (HOM) interferometers, free-space coupling is introduced to precisely control the polarization of the photons using half- and quarter-wave plates (HWP and QWP, respectively). This ensures that the polarization of the photons is optimized for interference.

## b) Fiber Beam Splitters

The purification circuits are implemented using optical fibers and involve two fiber beam splitters for each copy of the purification scheme. The purified photons from both circuits then interfere in a final beam splitter. The optical losses induced by the purification setup were characterized to be around 1 dB, primarily due to fiber connectors between the beam splitters. Since the overall circuit is based on cascaded HOM interferometers, which are phase-insensitive, no active phase stabilization is required.

## 3 Detection System

The purified photons are finally directed to six superconducting-nanowire single-photon detectors (SNSPDs), which have an average system efficiency of 85%. These detectors are used to measure the output configurations, providing the necessary data for evaluating the performance of the purification and interference process.
The SNSPD detector consists of a thin superconducting nanowire, usually made of niobium nitride (NbN) or tungsten silicide (WSi). The nanowire is cooled to cryogenic temperatures (~2–4 K) using a cryostat, where it stays in the superconducting state. When a single photon hits the nanowire, it locally disrupts the superconductivity, creating a small hotspot. This causes a temporary resistance in the wire, breaking the superconducting state at that point.
The sudden resistance change leads to a voltage pulse, which is detected as an electrical signal. The system then quickly recovers, allowing detection of the next photon.








# Results of Purification Protocol for Indistinguishable Photons

This section describes the results of the purification protocol aimed at improving the indistinguishability of photons emitted by a quantum dot (QD) in a photonic crystal waveguide (PCW). The study investigates the enhancement in Hong-Ou-Mandel (HOM) visibility after applying the purification protocol under various experimental configurations. The protocol was tested in both high-noise and low-noise environments to assess its performance in different scenarios. The results include the improvement in the indistinguishability of the purified photons as compared to the raw photons and the theoretical models used to predict the expected improvements.






## Raw HOM Visibility and Purification Performance

To analyze the improvement in indistinguishability induced by the purification, the first step was to assess the raw HOM visibility by interfering two demultiplexed photons. The raw HOM visibility was measured by blocking inputs 2 and 3 (as shown in Fig. 2) and interfering only inputs 1 and 4. Subsequently, inputs 2 and 3 were unblocked to implement the full purification circuit and test quantum interference between the purified photons.

## High-Noise Configuration ("No Etalon")

The purification protocol was first tested in a high-noise environment by removing the etalon filter after the QD, which results in a higher distinguishability due to the presence of incoherent spectral sidebands. The raw HOM visibility in this configuration was found to be \( V_0 = 0.5829(1) \). However, after implementing the purification protocol, the visibility of HOM interference between the purified photons increased to \( V_f = 0.685(5) \), showing a significant improvement of 10.2(5)%.

This improvement in visibility is also manifested in the spectral profiles of the interference curves, where the purple curve (purified) is narrower and has a lower maximum than the red curve (raw), suggesting that both the dynamics and the amplitude of the distinguishable components were altered. However, the width of the peak only changes due to the lower heralding probability of photons far from the center position, as observed in the side peaks. Therefore, it can be concluded that the purified photons exhibit a lower distinguishable component compared to the raw photons, but their intrinsic properties remain unchanged.

## Low-Noise Configuration ("Optimal Etalon")

In the second experimental configuration, the etalon filter was included to reduce the phonon-induced noise, creating a low-noise environment. In this case, the initial raw HOM visibility was measured to be \( V_0 = 0.9050(1) \), and after applying the purification protocol, the visibility increased to \( V_f = 0.9327(1) \). This result shows an improvement in HOM visibility due to the purification, with partial distinguishability mostly removed by the protocol. The remaining noise contributions limiting the visibility in this configuration were dominated by spurious multiphoton terms arising from the finite \( g^{(2)}(0) \) of the QD.

## Suboptimal Configuration ("Detuned Voltage")

In the third configuration, the voltage applied to the QD was intentionally detuned from its optimal value. This detuning increased the probability of cotunneling of carriers to the contacts, inducing a slight reduction in the raw HOM visibility to \( V_0 = 0.8332(1) \). After applying the purification protocol, the visibility was enhanced to \( V_f = 0.9090(5) \). The results indicate that even with a suboptimal QD configuration, the purification protocol still significantly improves the indistinguishability of the photons.

## Theoretical Models and Predictions

In addition to the experimental results, theoretical estimations of the indistinguishability for the purified photons were plotted as a function of the raw visibilities. Three different models were considered to predict the partial distinguishability of the photons:

1. **Multipermanent Model (MP):** This model assumes a constant state overlap and orthogonality between the distinguishable components of each photon.
   
2. **Minimum Purity Model (MP):** Based on the minimum purity model, this model assumes that the internal degrees of freedom of all photons are in the same mixed state, with finite purity as the cause for distinguishability.

3. **Pure-Dephasing Model (PD):** This model considers the noise processes inherent in a QD environment and incorporates higher-order indistinguishability instances, such as 3- and 4-photon contributions.

The theoretical curves provided slightly different expected improvements, but they were in good agreement with the observed experimental measurements. This suggests that the purification protocol effectively enhanced the indistinguishability of the photons, regardless of the underlying noise sources.

## Purification under Slow Errors

In addition to testing the purification protocol under fast noise sources, the protocol was also tested in the presence of slow errors. In this case, the polarization of one input photon per purifier was deliberately altered. The results showed significant improvements in the HOM visibility, as illustrated in the histograms shown in the Supplemental Material. The observed data were compatible with theoretical bounds, demonstrating the successful application of the purification protocol for both slow and fast noise contributions.






The results demonstrate the successful application of the purification protocol to enhance the indistinguishability of photons emitted from a quantum dot in a photonic crystal waveguide. The protocol was tested under different experimental configurations, including high-noise, low-noise, and suboptimal conditions, and showed significant improvements in HOM visibility. Theoretical models were used to predict the expected improvements, and the experimental data were found to be in good agreement with these predictions. Moreover, the purification protocol was shown to be effective in mitigating both fast and slow noise sources, distinguishing it from spectral filtering due to its ability to purify incoherent processes.



# Conclusion


In this work, we have successfully demonstrated a new purification protocol that improves the indistinguishability of partially distinguishable photons through quantum interference. The protocol, which requires minimal hardware and is highly versatile, can be applied to various photon emitters such as quantum dots, atoms, and heralded photon sources. Unlike previous noise mitigation techniques that are limited to addressing slow noise processes, our method efficiently mitigates both fast and slow noise, such as pure dephasing and spectral diffusion, thus broadening its applicability. Despite the need for ancillary photons and multiplexing circuits to achieve near-deterministic results, the purification protocol can be seamlessly integrated into existing photonic quantum architectures, such as those used in quantum computing, without introducing significant hardware overhead. Additionally, the protocol’s potential for further optimization, through adjustments to reflectivities or the use of advanced interferometers, holds the promise of even higher success probabilities and indistinguishability improvements. Overall, this purification approach provides a powerful and flexible tool to enhance photon quality in quantum photonics, which is critical for the scalable development of quantum technologies.
