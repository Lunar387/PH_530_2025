<h1 style="color:blue;">Cavity-enhanced Photon Indistinguishability at Room Temperature and Telecom Wavelengths : Nature Communications volume <b>15</b> 3989 (2024)</h1>

<p>
Indistinguishable single photons within the telecom range of optical fibers play a critical role in long-distance quantum communications. While solid-state emitters have made significant progress, achieving photon indistinguishability at ambient temperatures remains challenging. This paper demonstrates a significant enhancement in photon indistinguishability at room temperature by coupling individual nanotube defects (NTDs) within a fiber-based microcavity operating in the incoherent good cavity-coupling regime. The developed system exhibits superior efficiency compared to traditional spectral or temporal filtering, increasing photon indistinguishability by over two orders of magnitude relative to free-space emitters. This approach highlights the potential for creating optimized non-classical photon sources for quantum technologies.

# Introduction
By: Anurag Verma(23n0317)

**The ability of two indistinguishable single photons to interfere at a balanced beam splitter, resulting in their simultaneous exit through one of its output ports, is fundamental to quantum photonic technologies, including quantum teleportation, quantum computing, and quantum optical metrology.**


*   Solid-state-based sources of indistinguishable single photons have
witnessed tremendous progress in the past decades
* Semiconductor quantum dots stand out as they enable the generation
of pure and indistinguishable single photons when coupled to optical microcavities

The main drawback is:- **Restricted to cryogenic temeratures and weavelengths in near-infrared** but we want to operate at ambient conditions and telecom wavelengths to faciliate long-distance quantum communication. Various quantum emiiter are there for emitting pure telecom-band single photons at room temperature for example:


*   Color centers in silicon carbide
*   Gallium nitride

*   Luminescent nanotube defects (NTDs) : sp3-functionalized single-wall carbon nanotubes. These allow precise control unlike others over the emission wavelengths via covalent side-wall chemistry. These are easy to integrated with gated structures, microcavity or plasmonic cavities, all of these make NTDs excellent candidates for the development of source of quantum light.

<b>NTDs</b>
As we see in common solid-state quantum emitters, NTDS also are subject to strong dephasing at room temperatures, due to which the coherence time of the emitted photons is orders of magnitude smaller than the population time, which leads to the small values of photon indistinguishability.

<center>Coherence time $\tau = T_1$</center>
<center>Photon indistinguishability $=\frac{T_2}{2T_1}$</center>
<p>Previously</p>

*  Reducing $T_1$ to enhance the photon indistinguishability via Purcell enhancement (successfully applied to quantum dots and Erbium ions in various cavity geometries)
*  But the above tech were operated in the regimes of coherent or incoherent bad cavity coupling, and relied on operation at cryogenic temperatures.

In this paper they have demonstrated enhancement of photon indistinguishability for telecom band single photons from individual NTDs coupled to an optical microcavity, where the photon coherence time is determined by the cavity linewidth and by choosing cavity with a spectrally narrow linewidth, we enhance $T_2$ and thus photon indistinguishability of the coupled NTD-cavity system. And at the same time the cavity enahnce the emission via Purcell effect, thus we get incraese in the efficiency of our system by atleast a factor of 4 and with an estimated increase of photon indistinguishability by two orders of magnitude as compared to free space NTDs.
Their results experimentally establish the regime of incoherent good cavity-coupling as a powerful strategy for optimized
sources of quantum light.

## Methods
By Khush Soni(23n0290)

1.Sample Prepration

A) Preparation of NTD-Functionalized Carbon Nanotubes
* Dissolution of Carbon Nanotubes
  * Raw HiPco single-walled carbon nanotubes (SWCNTs) (from NoPo Nanotechnologies, India) were dissolved in chlorosulfonic acid (99%, Sigma-Aldrich) at 0.5 mg/mL concentration.This helps in dispersing nanotubes without causing bundling.
* Addition of Functionalizing Agents
  * 2-amino-4,5,6-trifluorobenzenesulfonyl chloride was added.
  * It was synthesized from 3,4,5-trifluoroaniline and sodium nitrite (NaNO₂, ReagentPlus® >99.0%, Sigma-Aldrich).
  * Their concentrations were 0.24 mg/mL and 0.2 mg/mL, respectively.
* Mixing and Precipitation
  * The acid mixture was gradually added drop by drop to Nanopure® water while stirring vigorously.
  * This led to the formation of NTD-functionalized carbon nanotubes, which precipitated as black solids.
* Purification
     * The precipitates were filtered and rinsed multiple times with Nanopure® water to remove excess reagents.
* Dispersion and Centrifugation
     * The purified NTDs were dissolved in 2% (wt/v) sodium deoxycholate (DOC, Sigma-Aldrich, ≥97%) solution.
     * The solution was centrifuged at 16,400 rpm for 1 hour to remove nanotube bundles.
* Chirality Sorting
     * The NTDs were sorted to obtain (8,6) chirality-enriched nanotubes using aqueous two-phase extraction.
     * The extraction solution contained 2% (w/v) DOC in deuterium oxide (D₂O, Cambridge Isotope Laboratories, Inc. 99.8%).

B) Deposition of NTDs on a Planner Mirror

 * Coating the Mirror with a Spacer Layer
      * A macroscopic planar mirror was spin-coated with 10 μL of a 3% (wt/v) polystyrene/toluene solution.
      * The spin-coating was performed at 2000 RPM for 1 minute, forming a 150 nm-thick polystyrene spacer layer.
     * The coated mirror was vacuum-dried at room temperature for 24 hours.

* Deposition of NTDs on the Mirror
     * 5 μL of the NTD-containing solution was spin-coated onto the mirror at 3000 RPM for 1 minute.
     * This step completed the deposition process.

2. Fiber based cavity

*   Experimental Setup: An ultra-stable fiber-based open cavity platform (Qlibri Quantum, Qlibri GmbH) was used for the study


A) Cavity Structure

*   Comprised of a macroscopic planar mirror with a 150 nm thick polystyrene spacer layer and functionalized carbon nanotubes (NTDs) on top.
*   A concave fiber mirror (radius of curvature: 25 μm) was created using CO₂ laser ablation.


B)Spacer Layer Purpose

*   The 150 nm polystyrene layer helped position the nanotubes at the antinode of the cavity mode for maximum interaction.

C) Precision Control

*   Piezoelectric positioners allowed sub-nanometer precision tuning of cavity length and lateral scans with three translational degrees of freedom.

D) Mirror Coatings

*   Both fiber and sample mirrors had DBR coatings made via ion beam sputtering (Laseroptik GmbH).
*   Designed for high reflectivity at telecom wavelengths, with a minimum transmission of 95.2 ppm at 1535.4 nm.

E) Cavity Performance

*   **Highest finesse recorded:** 3010 ± 10 at **1468 nm**, near the E*₁₁ peak maximum.
*   Lowest accessible mode order: q = 4, corresponding to a mirror distance of **Lc = 2.6 μm**.
*Computed cavity mode waist: ω₀ = 2 μm.

















3.  Photoluminescence and photon correlation experiments

A) Excitation Source

* Pulsed supercontinuum white light source (NKT SuperK Extreme).

* Operates at a 78 MHz repetition rate.

* Spectrally filtered to a 2 nm linewidth using a home-built monochromator.


B) Cavity Tuning

* Adjusted by modifying mirror distance to match E*₁₁ transition.


C) Emission Collection & Filtering

* PL emitted through the planar mirror was collimated using an achromatic doublet lens (Thorlabs AC127-019-CML).

* Filtered using two longpass filters:

    a. Thorlabs FEL1400 (band edge 1400 nm).

    b. Semrock BLP02-1319R-25 (band edge 1320 nm).

* The filtered signal was coupled into a single-mode fiber for further analysis.


D) Detection System

* Superconducting nanowire single-photon detectors (Scontel TCOPRS-CCR-SW-85).

* Photon timing recorded using a TCSPC module (Swabian Instruments Time Tagger Ultra and PicoQuant PicoHarp300).




4. Photon Correlation & Interference Experiments:



A) Second-order photon correlation measured using a Hanbury–Brown–Twiss (HBT) setup
* Used to measure single-photon purity of emitted light.


B) Two-Photon Interference (Hong-Ou-Mandel Effect) via Mach–Zehnder Interferometer

* Mechanical delay stage enabled sub-picosecond precision tuning of the interferometer delay.

* Polarization control achieved using fiber polarization controllers (Thorlabs FPC562).




5. Data Processing & Uncertainty Considerations

A) Histogram Generation 

* Detection events were integrated into 2.5 ns wide time windows.
* Histograms displayed peaks separated by excitation pulse delays.

B) Correlation Function Calculation
* g^(2)_HBT and g^(2)_HOM were obtained by normalizing histograms to the average peak height N_∞ at large time delays.

** Uncertainty Considerations:**
* The main uncertainty in peak height measurement N_0 at ̕τ = 0 was $\sqrt{N_0}$
* Gaussian error propagation was applied for quantities derived from measured peak heights.

** Second-order Correlation Function at Zero Time Delay:**

* The normalized second-order correlation function was computed as:
$
  g^{(2)}(0) = \frac{N_0}{N_1} \left( 1 \pm \frac{1}{\sqrt{N_0}} \right)
   $


C) Error Estimation 
* Dark counts and background noice were corrected
* Uncertainities in $ N_{\infty}\ $ and background were negligible compared to the domain uncertainity from $ N_0 $

Fig 1c- Effect of Cavity Coupling on Spectral Bandwidth
* Figure 1c illustrates how cavity coupling influences the spectral bandwidth of the emitted photons from nanotube defects (NTDs).
* Under ambient conditions, the spectral width of the NTD emission is primarily dictated by pure dephasing at a rate γ*. This results in a linewidth (Γ = 2γ*) that spans several nanometers (~10 meV), which is significantly broader than the cavity linewidth.
* The experimentally determined cavity linewidth for the lowest longitudinal mode was κ = 35.4 ± 0.1 μeV, translating to 61.7 ± 0.2 pm in the wavelength domain.
* The small value of κ enables operation in the incoherent good cavity coupling regime, characterized by the conditions:

 $ 2g << γ + γ* + κ $\
$ κ < γ + γ* $

where g is the light-matter coupling strength, γ is the population decay rate, and γ* represents the pure dephasing rate.

* In this regime, the cavity acts as an incoherent energy reservoir, where photons are incoherently pumped at a rate R ≈ 4g²/γ*.
* This rate is lower than the population decay rate, ensuring that any photon coupled into the resonator is emitted through the cavity mode on a timescale 1/κ.
* Since the cavity emission is coherent, this leads to a significant increase in photon coherence time compared to the free-space case, where coherence time is limited to 1/γ*.
* Spectral purification: The impact of this process is evident in the spectral domain, where cavity coupling drastically narrows the emission spectrum.
* This effect functions similarly to spectral filtering, ensuring that emitted photons exhibit high spectral purity and coherence.
* The observed spectral narrowing in Figure 1c is a defining feature of the incoherent good cavity coupling regime and is crucial for achieving enhanced photon indistinguishability, which is essential for quantum photonic applications.

**Effect of Cavity Coupling on Emmision Spectral Density**

* Beyond improving coherence time, the cavity also enhances the emission spectral density, which is characterized by the Purcell factor (Fₚ).
* The Purcell factor follows the relation:

  $F_p ∝ g^2$ \
  where g is the light matter coupling strength.

  Enhanshing single-photon efficiency:
* As g increases probability of photon being emitted into the cavity mode is also increases.
* However, in the incoherent good cavity regime, this probability remains lower than the free-space quantum yield because of the large spectral mismatch between the emitter's natural linewidth and the cavity linewidth.

Optimizing photon collection efficiency:
* Maximizing g (achieved by reducing the microcavity mode volume) leads to a substantial increase in photon emission efficiency.
* This enhancement significantly surpasses the efficiency obtained using spectral filtering at a bandwidth κ or an equivalent temporal bandwidth.

About Fig. 2b
* Figure 2b shows the normalized photoluminescence (PL) intensity of NTD 1 as the cavity length is adjusted.
* The cavity is tuned over three different longitudinal mode orders (q = 7, 8, and 9).
* For each mode order, the emission spectrum appears asymmetric due to the presence of higher-order transverse electromagnetic (TEM) modes.
* The cavity's linewidth (κ) is much narrower than the natural emission linewidth (Γ) of the NTD.
* Because of this, the NTD emission is detected only at the resonance wavelengths of specific TEM modes.
* The experimental data was fitted using three Lorentzian curves for each longitudinal mode order.
* Higher-order modes (TEM modes with n+m > 2) were ignored since their contribution was negligible.

Key Results:

* The NTD emission wavelength was determined to be 1465 ± 3 nm.
* The full width at half maximum (FWHM) linewidth was measured as 28 ± 5 nm, which corresponds to a dephasing rate (γ) of 8 ± 2 meV*.

* The intensity values were normalized to the highest value and adjusted for variations in fiber coupling efficiency.

* At this mode order, the mirror separation was 2.6 μm, which was mainly limited by the 2 μm depth of the fiber mirror profile.
* The increase in intensity is due to a stronger light-matter interaction (coupling strength, g) as the cavity length and mode volume (Vc) decrease.
* A smaller cavity mode volume (Vc) enhances the coupling strength g, which in turn increases emission efficiency.

Mathematical Relationship:
* In the low Purcell enhancement regime, the NTD's population lifetime remains mostly unchanged by the cavity.
* The PL intensity is proportional to g², and g² is inversely proportional to the cavity mode volume (Vc).
* A mathematical fit using αVc⁻¹, where Vc is calculated from the cavity length Lc = qλ/2, closely matches the experimental data (solid line in Fig. 2c).

Fig:4a
The indistinguishability of photons was measured using a Hong-Ou-Mandel (HOM) interference experiment with an imbalanced Mach–Zehnder interferometer

Experimental Setup
* A train of single-photon pulses was split using a fiber beamsplitter.
* The time delay (Δt) was adjusted by varying the path difference (Δz) using a delay stage.

* This allowed two-photon interference between consecutively emitted photons at the second beamsplitter.
* A zero delay means the photons are separated by just one excitation pulse.

Observations and Results:
* The polarization of the interferometer arms was adjusted using fiber polarization controllers.
* Detection events at the output ports were analyzed to obtain the HOM autocorrelation function g^2_HOM(τ).
* Experiments were conducted for co-polarized and cross-polarized configurations on NTD 3.

Key Findings:
* In the co-polarized case, there was a reduction in correlations at zero time delay, indicating quantum-coherent two-photon interference.
* This means that indistinguishable photons arriving at separate input ports of the beamsplitter tend to exit through the same output port, lowering correlation values.

Photon Indistinguishability Visibility:
* The two-photon interference visibility was measured as:
v = 0.51 ± 0.21
* This accounts for factors like non-identical reflection and transmission coefficients of the beamsplitters and the finite single-photon purity of NTD 3.

## 1. Proof of  regime of incoherent good NTD-cavity coupling

In this regime of incoherent good cavity coupling, the conditions **2g ≪ γ + γ∗ + κ and κ < γ + γ∗** hold. Here:

g = Light-matter coupling strength

γ = Population decay rate (which includes radiative decay rate γr and nonradiative decay rate γnr)

γ∗ = Pure dephasing rate (loss of coherence without energy loss)

κ = Cavity linewidth (related to how fast the cavity leaks photons)

Key Findings in This Work:
* The cavity linewidth was measured as **κ = 35.4 ± 0.1 µeV.**

* The pure dephasing rate **γ∗ = 8 ± 2 meV** was determined using cavity length sweep experiments.

* The population decay follows a biexponential trend, meaning it has two different timescales:
    1. Fast decay: τfast = 2 ps, corresponding to γfast = 330 µeV.

    2. Slow decay: τslow = 91 ps, corresponding to γslow = 7.3 µeV.

The condition **κ < γ + γ∗** holds true for both fast and slow decay components, confirming that our system operates in the good cavity regime.

Estimation of Light-Matter Coupling Strength (g)
The coupling strength g is estimated using the equation:

**$$ g = \sqrt\frac{3\lambda^2 c}{8\pi n^3 V_c \tau_{\text{rad}}} $$**

where:

λ = Wavelength of light

c = Speed of light

Vc = Cavity mode volume (calculated as 8.2 µm³)

n = Refractive index (taken as 1, though the NTDs are on a polystyrene spacer with n = 1.57)

τrad = Radiative lifetime (varies between 1 and 15 ns)

By varying τrad between 1 and 15 ns, we estimate that g ranges between 17 and 64 µeV.

Final Conclusion:
* Since **2g ≪ γ + γ∗ + κ**, this confirms that the system operates in the regime of incoherent good cavity coupling, meaning that: Light-matter interaction is relatively weak.

* The emitter (NTD) decays mostly through non-radiative and dephasing processes rather than strong coherent coupling.

* The cavity enhances the emission but does not reach the strong coupling regime where quantum effects like Rabi oscillations would be dominant.

## 2. Regime of Low Purcell Enhancement

### A. Purcell Factor and Photoluminescence (PL) Lifetime

The **PL lifetime** $(\tau_{pl})$ of an emitter in a cavity is determined by the **free-space lifetime** $(\tau_{fs})$ and the **effective Purcell factor** $(F_{p^*})$.

The relationship is:

$$
\tau_{pl} = \frac{\tau_{fs}}{1 + F_{p^*}}
$$

where $
F_{p^*} = \frac{R}{\gamma}
$

with:

-  R being the **emitter-cavity coupling rate**,  
- $\gamma$ the **spontaneous emission rate**.

In the study, for **slow decay** $(\tau_{fs} = 91 ps)$, $F_{p^*}$ varies from **0.018 to 0.28**, implying **minimal influence of cavity coupling**.

For **fast decay**,$ F_{p^*}$ is even lower (**between $4.1 * 10^{-4}$ and $ 6.2 * 10^{-3} $**).

These **small values indicate weak Purcell enhancement**, meaning the **cavity has little effect on emission rate enhancement**.

### Radiative Decay Enhancement

The **Purcell factor** $(F_p)$ quantifies how much the cavity boosts the radiative decay rate:

$$
F_p = \frac{3 \lambda^3 Q_{\text{eff}}}{4 \pi^2 n^3 V_c}$$

where:

- $Q_{\text{eff}}$ is the **effective quality factor**,
- $ \lambda $ is the **emission wavelength**,
- $ n$ is the **refractive index**,
- $V_c $ is the **cavity mode volume**.

For the given system, **$F_p = 1.6 $**, indicating **limited enhancement**.

In the **bad cavity regime** $( \kappa \gg \gamma )$, $F_p$ follows **Purcell’s original formula**:

$$
F_p = \frac{4g^2}{\gamma_{\text{rad}} \kappa}
$$

where $ g$ is the **light-matter coupling strength**.

The **effective Purcell factor** can also be written as $F_{p^*} = \eta_Q F_p$

where $ \eta_Q$  (**quantum yield**) is the **fraction of spontaneous emission contributing to useful radiation**.

In this system, $ \eta_Q $ is estimated between **0.006 and 0.09**, aligning with **previous experimental values**.

### B. Single-Photon Efficiency

The **single-photon emission efficiency** $beta_c$ measures the probability that an emitted photon remains within the cavity linewidth $kappa$.

The **photon detection rate** $I_{em}$ follows the equation:

$$I_{em} = f_{exc} \eta_{out} \eta_{sys} \eta_{em} \beta_c$$

where:

- $f_{exc}$ is the **laser repetition rate**,
- $\eta_{out}$ is the **probability of a photon exiting the cavity**,
- $\eta_{sys}$ is the **setup efficiency**,
- $\eta_{em}$ is the **free-space emission efficiency**.

The **maximum observed detection rate** is **1840 counts/s**.

The **minimum estimated** $\beta_c$ is:

$\beta_c = 3.9 \times 10^{-3}$

while the **theoretical value** is:

$\beta_c = 6.6 \times 10^{-3}$

**Cavity coupling** improves **single-photon emission efficiency** by **at least 44 times** compared to free-space emission.

## C. Cavity-Enhanced PL Intensity

The **single-photon emission efficiency** $\beta_c$ is given by:

$$
\beta_c = \frac{\kappa R}{\kappa R + \gamma (\kappa + R)}$$

which, in the **Purcell regime** $( R \ll \kappa) $, simplifies to:

$$
\beta_c \approx \frac{F_{p^*}}{1 + F_{p^*}}
$$

The **PL intensity** (Iem)is proportional to $V_c^{-1}$,confirming that **smaller cavity volumes enhance light-matter interaction**.

Thus, **cavity coupling significantly boosts photon extraction efficiency**, but the **Purcell enhancement remains in a low regime**.
